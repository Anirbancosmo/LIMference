```python
import numpy as np
from scipy.stats import qmc
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec

class LIMpyGenerator:
    """Generate Line Intensity Maps with flexible parameter combinations"""
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, noise_level=1e1):
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        
        # Parameter configuration
        self.param_names = param_names
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        self.astrophysical_params = [p for p in param_names if p not in self.cosmological_params]
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        #noise
        self.noise_level = noise_level
        
    def setup_parameter_ranges(self, param_ranges=None):
        """
        Setup parameter ranges for astrophysical parameters
        
        param_ranges : dict
            e.g., {'a_off': (-1.0, 1.0), 'b_off': (-0.5, 0.5)}
        """
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
    
    def generate_latin_hypercube_samples(self, n_samples):
        """
        Generate Latin hypercube samples for astrophysical parameters
        Fixed seed ensures reproducibility
        """
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        # Create Latin hypercube sampler with fixed seed
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        
        # Generate samples in [0,1]
        samples_unit = sampler.random(n=n_samples)
        
        # Scale to parameter ranges
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            # Keep the column as 2D array for qmc.scale
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        for i, param in enumerate(self.astrophysical_params):
            print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
        
        return samples_scaled
    
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process single halocat with parameter combination"""
        
        # Load cosmological parameters from halocat
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters from halocat
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters from Latin hypercube
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            # Initialize LIMpy simulator
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)

    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None):
        """Save results with proper parameter structure including seed and noise"""

        if seed is None:
            seed = self.seed

        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)

        # Create structured array based on parameter names
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)

        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)

        # Save files with metadata
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"

        all_k = np.array([self.k_values] * len(self.power_spectra))

        # Save with additional metadata
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            seed=seed,
            noise_level=self.noise_level
        )

        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            seed=seed,
            noise_level=self.noise_level
        )

        # Enhanced parameter configuration
        self.save_parameter_configuration(output_dir)

        # Save parameter array with full metadata
        param_file = output_path / f"parameters_{seed}.npz"
        np.savez_compressed(
            param_file,
            param_names=self.param_names,
            param_array=self.params_array,
            latin_hypercube_samples=self.latin_hypercube_samples if self.latin_hypercube_samples is not None else [],
            seed=seed,
            noise_level=self.noise_level,
            boxsize=self.boxsize,
            ngrid=self.ngrid,
            redshift=self.redshift,
            line_name=self.line_name,
            model_name=self.model_name
        )

        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Parameters: {param_file}")
        print(f"\nMetadata:")
        print(f"  Seed: {seed}")
        print(f"  Noise level: {self.noise_level}")
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")

    def save_parameter_configuration(self, output_dir):
        """Save parameter configuration for reproducibility"""
        config = {
            'param_names': self.param_names,
            'param_ranges': self.param_ranges,
            'seed': self.seed,
            'noise_level': self.noise_level,
            'n_samples': len(self.latin_hypercube_samples) if self.latin_hypercube_samples is not None else 0,
            'cosmological_params': self.cosmological_params,
            'astrophysical_params': self.astrophysical_params,
            'boxsize': self.boxsize,
            'ngrid': self.ngrid,
            'redshift': self.redshift,
            'line_name': self.line_name,
            'model_name': self.model_name
        }

        output_path = Path(output_dir)
        config_file = output_path / 'parameter_config.json'

        with open(config_file, 'w') as f:
            json.dump(config, f, indent=2)

        print(f"Parameter configuration saved to {config_file}")

    def process_batch_with_params(self, halocat_files, store_maps=False):
        """
        Process batch with parameter combinations

        Parameters:
        -----------
        noise_level : float
            Standard deviation of Gaussian noise to add
        store_maps : bool
            Whether to store intensity maps for visualization
        """

        n_files = len(halocat_files)

        # Generate Latin hypercube samples if needed
        if self.astrophysical_params:
            if self.latin_hypercube_samples is None:
                self.generate_latin_hypercube_samples(n_files)

        # Set random seed for reproducible noise
        np.random.seed(self.seed)

        all_params = []
        all_pk = []
        all_pdf = []

        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                # Get astrophysical parameters for this file
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]

                # Process
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )

                # Add noise with fixed seed for reproducibility
                noise = np.random.normal(0, self.noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise

                # Store map if requested
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)

                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)

                # Store parameters in correct order
                param_values = []
                for param_name in self.param_names:
                    param_values.append(params[param_name])

                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)

                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins

            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue

        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf

        return len(all_params)
    
    def visualize_parameter_dependence(self, param_to_vary, fixed_params=None, n_samples=5):
        """
        Visualize how LIM observables change with one parameter
        
        Parameters:
        -----------
        param_to_vary : str
            Name of parameter to vary ('sigma_8', 'omega_m', 'a_off', 'b_off')
        fixed_params : dict or None
            Values to fix other parameters at. If None, uses median values.
        n_samples : int
            Number of samples to show
        """
        
        if self.params_array is None:
            print("No data processed yet")
            return
        
        if len(self.intensity_maps) == 0:
            print("No intensity maps stored. Process with store_maps=True")
            return
        
        # Get parameter index
        if param_to_vary not in self.param_names:
            print(f"Parameter {param_to_vary} not in {self.param_names}")
            return
        
        param_idx = self.param_names.index(param_to_vary)
        
        # Only use data with stored maps
        n_stored_maps = len(self.intensity_maps)
        params_with_maps = self.params_array[:n_stored_maps]
        
        # Set fixed values for other parameters
        if fixed_params is None:
            fixed_params = {}
            for i, name in enumerate(self.param_names):
                if name != param_to_vary:
                    fixed_params[name] = np.median(params_with_maps[:, i])
        
        # Find samples matching fixed parameters
        tolerance = 0.1  # Relative tolerance
        mask = np.ones(len(params_with_maps), dtype=bool)
        
        for name, value in fixed_params.items():
            if name != param_to_vary and name in self.param_names:
                idx = self.param_names.index(name)
                param_range = params_with_maps[:, idx].max() - params_with_maps[:, idx].min()
                tol = tolerance * param_range if param_range > 0 else 0.1
                mask &= np.abs(params_with_maps[:, idx] - value) < tol
        
        filtered_indices = np.where(mask)[0]
        
        if len(filtered_indices) < n_samples:
            print(f"Only found {len(filtered_indices)} samples matching criteria")
            if len(filtered_indices) == 0:
                return
        
        # Sort by varying parameter and select samples
        sorted_idx = filtered_indices[np.argsort(params_with_maps[filtered_indices, param_idx])]
        
        if len(sorted_idx) > n_samples:
            step = len(sorted_idx) // n_samples
            selected_idx = sorted_idx[::step][:n_samples]
        else:
            selected_idx = sorted_idx
        
        # Create visualization
        actual_n_samples = len(selected_idx)
        fig = plt.figure(figsize=(20, 14))
        gs = gridspec.GridSpec(3, actual_n_samples + 1, 
                              width_ratios=[1]*actual_n_samples + [0.05],
                              height_ratios=[1.2, 1, 1],
                              hspace=0.3, wspace=0.2)
        
        # Color mapping
        param_values = params_with_maps[selected_idx, param_idx]
        norm = plt.Normalize(param_values.min(), param_values.max())
        colors = cm.viridis(norm(param_values))
        
        # Print info
        print(f"Showing {param_to_vary} variations")
        print("Fixed parameters:")
        for name, value in fixed_params.items():
            if name != param_to_vary:
                print(f"  {name} = {value:.3f}")
        
        # Row 1: LIM maps
        all_intensities = np.concatenate([self.intensity_maps[idx].flatten() for idx in selected_idx])
        vmin, vmax = np.percentile(all_intensities, [5, 95])
        
        for i, idx in enumerate(selected_idx):
            ax = fig.add_subplot(gs[0, i])
            
            im = ax.imshow(self.intensity_maps[idx], 
                          cmap='hot', aspect='equal',
                          vmin=vmin, vmax=vmax,
                          origin='lower',
                          extent=[0, self.boxsize, 0, self.boxsize])
            
            ax.set_title(f'{param_to_vary}={param_values[i]:.3f}', 
                        fontsize=11, fontweight='bold')
            
            if i == 0:
                ax.set_ylabel('LIM Maps\ny [Mpc/h]', fontsize=12)
                ax.set_xlabel('x [Mpc/h]', fontsize=10)
            else:
                ax.set_xticks([])
                ax.set_yticks([])
        
        # Row 2: Power Spectra
        ax_ps = fig.add_subplot(gs[1, :actual_n_samples])
        
        for i, idx in enumerate(selected_idx):
            ax_ps.loglog(self.k_values, self.power_spectra[idx], 
                        color=colors[i], alpha=0.8, linewidth=2.5,
                        label=f'{param_to_vary}={param_values[i]:.3f}')
        
        ax_ps.set_xlabel('k [h/Mpc]', fontsize=12)
        ax_ps.set_ylabel('P(k) [(Mpc/h)³]', fontsize=12)
        ax_ps.set_title(f'Power Spectrum Response to {param_to_vary}', 
                       fontsize=14, fontweight='bold')
        ax_ps.grid(True, alpha=0.3, which='both')
        ax_ps.legend(loc='best', fontsize=10, framealpha=0.9)
        
        # Row 3: PDFs
        ax_pdf = fig.add_subplot(gs[2, :actual_n_samples])
        
        bin_centers = (self.pdf_bins[:-1] + self.pdf_bins[1:]) / 2
        
        for i, idx in enumerate(selected_idx):
            ax_pdf.semilogy(bin_centers, self.pdfs[idx],
                           color=colors[i], alpha=0.8, linewidth=2.5,
                           label=f'{param_to_vary}={param_values[i]:.3f}')
        
        ax_pdf.set_xlabel('Line Intensity [Jy/sr]', fontsize=12)
        ax_pdf.set_ylabel('PDF', fontsize=12)
        ax_pdf.set_title(f'Intensity PDF Response to {param_to_vary}', 
                        fontsize=14, fontweight='bold')
        ax_pdf.grid(True, alpha=0.3, which='both')
        ax_pdf.legend(loc='best', fontsize=10, framealpha=0.9)
        
        # Colorbar
        cax = fig.add_subplot(gs[:, -1])
        sm = plt.cm.ScalarMappable(cmap=cm.viridis, norm=norm)
        sm.set_array([])
        cbar = plt.colorbar(sm, cax=cax)
        cbar.set_label(f'{param_to_vary}', fontsize=14, fontweight='bold')
        
        # Title with all parameters
        title = f'LIMpy: {self.line_name} at z={self.redshift} ({self.model_name})\n'
        title += f'Parameter: {param_to_vary}'
        fig.suptitle(title, fontsize=16, fontweight='bold', y=0.98)
        
        plt.show()

# Main processing function
def process_with_parameter_combination(
    input_dir,
    output_dir,
    param_names=['sigma_8', 'omega_m'],
    param_ranges=None,
    seed=42,
    max_files=None,
    noise_level=1e3,
    store_maps=True
):
    """
    Process halocats with specified parameter combination
    
    Parameters:
    -----------
    param_names : list
        Names of parameters to use
    param_ranges : dict
        Ranges for astrophysical parameters
    seed : int
        Random seed for reproducibility
    max_files : int or None
        Maximum number of files to process
    store_maps : bool
        Whether to store intensity maps for visualization
    """
    
    generator = LIMpyGenerator(param_names=param_names, seed=seed, noise_level=noise_level)
    generator.setup_parameter_ranges(param_ranges)
    
    # Get halocats
    input_path = Path(input_dir)
    halocat_files = sorted(input_path.glob("halocat_*.npz"))
    
    if max_files:
        halocat_files = halocat_files[:max_files]
    
    print(f"Processing {len(halocat_files)} files with parameters: {param_names}")
    
    # Process with option to store maps
    n_processed = generator.process_batch_with_params(halocat_files, store_maps=store_maps)
    
    # Save
    generator.save_results(output_dir, seed=seed)
    
    return generator

# Example usage
if __name__ == "__main__":
    # Example configurations
    configs = [
        # 2 parameters (cosmological only)
        #{'param_names': ['sigma_8', 'omega_m']},
        
        # 3 parameters (cosmological + 1 astrophysical)
        #{'param_names': ['sigma_8', 'omega_m', 'a_off'], 
        # 'param_ranges': {'a_off': (2, 10)}},
        
        # 4 parameters (cosmological + 2 astrophysical)
        {'param_names': ['sigma_8', 'omega_m', 'a_off', 'b_off'], 
         'param_ranges': {'a_off': (2, 10), 'b_off': (0, 2)}}
    ]
    
    # Process with 4 parameters as in your example
    config = configs[0]  # Use the 4-parameter configuration
    
    print(f"\n{'='*60}")
    generator = process_with_parameter_combination(
        input_dir="/Users/anirbanroy/Desktop/halocats/",
        output_dir="/Users/anirbanroy/Desktop/processed_halos_4p/",
        seed=52,  # Fixed seed for reproducibility
        max_files=3000,
        noise_level=1e-5,
        store_maps=True,  # Store maps for visualization
        **config
    )
    
    # Visualize parameter effects if generator was successful
    if generator and len(generator.intensity_maps) > 0:
        # For 4-parameter case, need to fix other parameters
        generator.visualize_parameter_dependence(
            param_to_vary='sigma_8',
            fixed_params={'omega_m': 0.3, 'a_off': 7.0, 'b_off': 1.0},
            n_samples=5
        )
        
        generator.visualize_parameter_dependence(
            param_to_vary='a_off',
            fixed_params={'sigma_8': 0.8, 'omega_m': 0.3, 'b_off': 1.0},
            n_samples=5
        )
```

```python
import numpy as np
from scipy.stats import qmc
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec

class LIMpyGenerator:
    """Generate Line Intensity Maps with flexible parameter combinations"""
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, noise_level=1e1,
                 fixed_astro_params_file=None):
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        
        # Parameter configuration
        self.param_names = param_names
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        self.astrophysical_params = [p for p in param_names if p not in self.cosmological_params]
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # Noise
        self.noise_level = noise_level
        
        # Fixed parameters file
        self.fixed_astro_params_file = fixed_astro_params_file
    
    def setup_parameter_ranges(self, param_ranges=None):
        """
        Setup parameter ranges for astrophysical parameters
        
        param_ranges : dict
            e.g., {'a_off': (4, 10), 'b_off': (0.0, 2.0)}
        """
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
    
    def generate_latin_hypercube_samples(self, n_samples):
        """
        Generate Latin hypercube samples for astrophysical parameters
        Fixed seed ensures reproducibility
        """
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        # Create Latin hypercube sampler with fixed seed
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        
        # Generate samples in [0,1]
        samples_unit = sampler.random(n=n_samples)
        
        # Scale to parameter ranges
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        for i, param in enumerate(self.astrophysical_params):
            print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
        
        return samples_scaled
    
    def load_fixed_latin_hypercube(self, filepath):
        """Load pre-generated Latin hypercube samples"""
        data = np.load(filepath, allow_pickle=True)
        self.latin_hypercube_samples = data['latin_hypercube_samples']
        
        # Verify parameters match if stored
        if 'param_names' in data:
            expected_params = data['param_names'].tolist()
            if isinstance(expected_params[0], bytes):
                expected_params = [p.decode() if isinstance(p, bytes) else p for p in expected_params]
            if expected_params != self.astrophysical_params:
                print(f"Warning: Parameter mismatch - expected {self.astrophysical_params}, got {expected_params}")
        
        # Load parameter ranges if available
        if 'param_ranges' in data:
            stored_ranges = data['param_ranges'].item() if hasattr(data['param_ranges'], 'item') else data['param_ranges']
            self.param_ranges.update(stored_ranges)
        
        print(f"Loaded {len(self.latin_hypercube_samples)} fixed Latin hypercube samples from {filepath}")
        for i, param in enumerate(self.astrophysical_params):
            print(f"  {param}: [{self.latin_hypercube_samples[:, i].min():.3f}, "
                  f"{self.latin_hypercube_samples[:, i].max():.3f}]")
        
        return self.latin_hypercube_samples
    
    def save_latin_hypercube(self, filepath):
        """Save Latin hypercube samples for future reuse"""
        if self.latin_hypercube_samples is None:
            raise ValueError("No Latin hypercube samples to save")
        
        filepath = Path(filepath)
        np.savez_compressed(
            filepath,
            latin_hypercube_samples=self.latin_hypercube_samples,
            param_names=self.astrophysical_params,
            param_ranges=self.param_ranges,
            seed=self.seed
        )
        print(f"Saved Latin hypercube samples to {filepath}")
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process single halocat with parameter combination"""
        
        # Load cosmological parameters from halocat
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters from halocat
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters from Latin hypercube
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            # Initialize LIMpy simulator
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def process_batch_with_params(self, halocat_files, store_maps=False):
        """
        Process batch with parameter combinations
        
        Parameters:
        -----------
        halocat_files : list
            List of halocat files to process
        store_maps : bool
            Whether to store intensity maps for visualization
        """
        n_files = len(halocat_files)
        
        # Load fixed parameters if specified
        if self.fixed_astro_params_file and self.latin_hypercube_samples is None:
            self.load_fixed_latin_hypercube(self.fixed_astro_params_file)
        
        # Generate Latin hypercube samples only if not already loaded
        if self.astrophysical_params:
            if self.latin_hypercube_samples is None:
                self.generate_latin_hypercube_samples(n_files)
        
        # Verify we have enough samples
        if self.latin_hypercube_samples is not None:
            if len(self.latin_hypercube_samples) < n_files:
                raise ValueError(f"Not enough Latin hypercube samples: have {len(self.latin_hypercube_samples)}, need {n_files}")
        
        # Set random seed for reproducible noise
        np.random.seed(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                # Get astrophysical parameters for this file
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                # Process
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise with fixed seed for reproducibility
                noise = np.random.normal(0, self.noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                # Store map if requested
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters in correct order
                param_values = []
                for param_name in self.param_names:
                    param_values.append(params[param_name])
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def save_results(self, output_dir, seed=None):
        """Save results with proper parameter structure including seed and noise"""
        
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array based on parameter names
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # Save files with metadata
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Save with additional metadata
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            seed=seed,
            noise_level=self.noise_level
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            seed=seed,
            noise_level=self.noise_level
        )
        
        # Save parameter configuration
        self.save_parameter_configuration(output_dir)
        
        # Save parameter array with full metadata
        param_file = output_path / f"parameters_{seed}.npz"
        np.savez_compressed(
            param_file,
            param_names=self.param_names,
            param_array=self.params_array,
            latin_hypercube_samples=self.latin_hypercube_samples if self.latin_hypercube_samples is not None else [],
            seed=seed,
            noise_level=self.noise_level,
            boxsize=self.boxsize,
            ngrid=self.ngrid,
            redshift=self.redshift,
            line_name=self.line_name,
            model_name=self.model_name,
            fixed_astro_params_file=str(self.fixed_astro_params_file) if self.fixed_astro_params_file else ""
        )
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Parameters: {param_file}")
        print(f"\nMetadata:")
        print(f"  Seed: {seed}")
        print(f"  Noise level: {self.noise_level}")
        if self.fixed_astro_params_file:
            print(f"  Fixed params from: {self.fixed_astro_params_file}")
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def save_parameter_configuration(self, output_dir):
        """Save parameter configuration for reproducibility"""
        config = {
            'param_names': self.param_names,
            'param_ranges': self.param_ranges,
            'seed': self.seed,
            'noise_level': self.noise_level,
            'n_samples': len(self.latin_hypercube_samples) if self.latin_hypercube_samples is not None else 0,
            'cosmological_params': self.cosmological_params,
            'astrophysical_params': self.astrophysical_params,
            'boxsize': self.boxsize,
            'ngrid': self.ngrid,
            'redshift': self.redshift,
            'line_name': self.line_name,
            'model_name': self.model_name,
            'fixed_astro_params_file': str(self.fixed_astro_params_file) if self.fixed_astro_params_file else None
        }
        
        output_path = Path(output_dir)
        config_file = output_path / 'parameter_config.json'
        
        with open(config_file, 'w') as f:
            json.dump(config, f, indent=2)
        
        print(f"Parameter configuration saved to {config_file}")
    
    def visualize_parameter_dependence(self, param_to_vary, fixed_params=None, n_samples=5):
        """
        Visualize how LIM observables change with one parameter
        """
        # [Keep existing visualization code unchanged]
        # ... (same as your original implementation) ...
        pass


# Utility functions for generating and managing fixed parameters

def generate_and_save_fixed_latin_hypercube(n_samples, param_ranges, seed, output_file):
    """
    Generate Latin hypercube samples once and save them
    
    Parameters:
    -----------
    n_samples : int
        Number of samples to generate
    param_ranges : dict
        Parameter ranges, e.g., {'a_off': (2, 10), 'b_off': (0, 2)}
    seed : int
        Random seed for reproducibility
    output_file : str
        Path to save the samples
    """
    from scipy.stats import qmc
    
    param_names = list(param_ranges.keys())
    n_params = len(param_names)
    
    # Create sampler with fixed seed
    sampler = qmc.LatinHypercube(d=n_params, seed=seed)
    samples_unit = sampler.random(n=n_samples)
    
    # Scale to parameter ranges
    samples_scaled = np.zeros((n_samples, n_params))
    for i, param in enumerate(param_names):
        low, high = param_ranges[param]
        samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
    
    # Save
    output_path = Path(output_file)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    np.savez_compressed(
        output_path,
        latin_hypercube_samples=samples_scaled,
        param_names=param_names,
        param_ranges=param_ranges,
        seed=seed
    )
    
    print(f"Generated and saved {n_samples} Latin hypercube samples to {output_file}")
    print(f"Parameters: {param_names}")
    for i, param in enumerate(param_names):
        print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
    
    return samples_scaled


def process_with_fixed_parameters(
    input_dir,
    output_dir,
    fixed_params_file,
    param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],
    seed=42,
    max_files=None,
    noise_level=1e3,
    store_maps=False
):
    """
    Process halocats with fixed astrophysical parameters
    
    Parameters:
    -----------
    fixed_params_file : str
        Path to pre-generated Latin hypercube samples
    Other parameters same as original
    """
    
    generator = LIMpyGenerator(
        param_names=param_names, 
        seed=seed, 
        noise_level=noise_level,
        fixed_astro_params_file=fixed_params_file
    )
    
    # Get halocats
    input_path = Path(input_dir)
    halocat_files = sorted(input_path.glob("halocat_*.npz"))
    
    if max_files:
        halocat_files = halocat_files[:max_files]
    
    print(f"Processing {len(halocat_files)} files with parameters: {param_names}")
    print(f"Using fixed parameters from: {fixed_params_file}")
    
    # Process
    n_processed = generator.process_batch_with_params(halocat_files, store_maps=store_maps)
    
    # Save
    generator.save_results(output_dir, seed=seed)
    
    return generator


# Main execution
if __name__ == "__main__":
    # Step 1: Generate fixed Latin hypercube samples ONCE
    fixed_params_file = "/Users/anirbanroy/Desktop/fixed_latin_hypercube_3000.npz"
    
    # Only run this once to generate the fixed parameters
    if not Path(fixed_params_file).exists():
        param_ranges = {'a_off': (2, 10), 'b_off': (0, 2)}
        generate_and_save_fixed_latin_hypercube(
            n_samples=3000,
            param_ranges=param_ranges,
            seed=999,  # Fixed seed for Latin hypercube generation
            output_file=fixed_params_file
        )
    
    # Step 2: Process with different seeds but same astrophysical parameters
    seeds_to_process = [50, 51, 52, 53]  # Different noise realizations
    noise_level = [1e-5, 1e1, 1e2, 1e3]
    
    for i, seed in enumerate(seeds_to_process):
        print(f"\n{'='*60}")
        print(f"Processing with seed {seed}")
        print(f"{'='*60}")
        
        generator = process_with_fixed_parameters(
            input_dir="/Users/anirbanroy/Desktop/halocats/",
            output_dir=f"/Users/anirbanroy/Desktop/processed_halos_cosmology_2p_seed{seed}/",
            fixed_params_file=fixed_params_file,
            param_names=['sigma_8', 'omega_m'],
            seed=seed,  # Different seed for noise
            max_files=3000,
            noise_level=noise_level[i],
            store_maps=(seed == seed)  # Only store maps for first seed to save memory
        )
        
        print(f"Completed processing for seed {seed}")
```

```python
import numpy as np
from scipy.stats import qmc
from scipy.spatial import KDTree  # Correct import for KDTree
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec
import pandas as pd  # If you're using the catalog indexing


class HalocatMatcher:
    """Find and match halo catalogs based on cosmological parameters"""
    
    def __init__(self, halocat_dir):
        self.halocat_dir = Path(halocat_dir)
        self.catalog_index = None
        self.kdtree = None
        
    def build_catalog_index(self, max_files=None):
        """Build index of all halocats with their cosmological parameters"""
        halocat_files = sorted(self.halocat_dir.glob("halocat_*.npz"))
        
        if max_files:
            halocat_files = halocat_files[:max_files]
        
        catalog_data = []
        
        print(f"Indexing {len(halocat_files)} halo catalogs...")
        for halocat_file in tqdm(halocat_files):
            try:
                data = np.load(halocat_file, allow_pickle=True)
                
                if 'params' in data:
                    params = data['params']
                    if hasattr(params, 'item'):
                        params = params.item()
                else:
                    params = {}
                
                sigma_8 = params.get('sigma_8', params.get('sigma8', None))
                omega_m = params.get('omega_m', params.get('Omega_m', None))
                h = params.get('h', None)
                
                if sigma_8 is not None and omega_m is not None:
                    catalog_data.append({
                        'file': halocat_file,
                        'sigma_8': float(sigma_8),
                        'omega_m': float(omega_m),
                        'h': float(h) if h else 0.7,
                        'index': len(catalog_data)
                    })
                    
            except Exception as e:
                print(f"Error reading {halocat_file.name}: {e}")
                continue
        
        self.catalog_index = pd.DataFrame(catalog_data) if catalog_data else None
        
        if self.catalog_index is not None:
            # Build KDTree for fast nearest neighbor search
            coords = self.catalog_index[['sigma_8', 'omega_m']].values
            self.kdtree = KDTree(coords)
            
            print(f"Indexed {len(self.catalog_index)} catalogs")
            print(f"  sigma_8 range: [{self.catalog_index['sigma_8'].min():.3f}, "
                  f"{self.catalog_index['sigma_8'].max():.3f}]")
            print(f"  omega_m range: [{self.catalog_index['omega_m'].min():.3f}, "
                  f"{self.catalog_index['omega_m'].max():.3f}]")
        
        return self.catalog_index
    
    def find_closest_halocat(self, sigma_8, omega_m, return_distance=False):
        """Find halocat with closest cosmological parameters"""
        if self.kdtree is None:
            raise ValueError("Build catalog index first with build_catalog_index()")
        
        query_point = np.array([sigma_8, omega_m])
        distance, index = self.kdtree.query(query_point)
        
        closest = self.catalog_index.iloc[index]
        
        if return_distance:
            return closest['file'], distance, closest
        return closest['file']
    
    def find_multiple_closest(self, sigma_8, omega_m, k=5):
        """Find k closest halocats"""
        if self.kdtree is None:
            raise ValueError("Build catalog index first")
        
        query_point = np.array([sigma_8, omega_m])
        distances, indices = self.kdtree.query(query_point, k=k)
        
        results = []
        for dist, idx in zip(distances, indices):
            catalog = self.catalog_index.iloc[idx]
            results.append({
                'file': catalog['file'],
                'distance': dist,
                'sigma_8': catalog['sigma_8'],
                'omega_m': catalog['omega_m']
            })
        
        return results
    
    def save_index(self, filepath):
        """Save catalog index for future use"""
        if self.catalog_index is not None:
            self.catalog_index.to_pickle(filepath)
            print(f"Saved catalog index to {filepath}")
    
    def load_index(self, filepath):
        """Load pre-built catalog index"""
        import pandas as pd
        self.catalog_index = pd.read_pickle(filepath)
        coords = self.catalog_index[['sigma_8', 'omega_m']].values
        self.kdtree = KDTree(coords)
        print(f"Loaded catalog index with {len(self.catalog_index)} entries")


# Usage example
if __name__ == "__main__":
    import pandas as pd
    
    # Step 1: Build catalog index (run once)
    matcher = HalocatMatcher("/Users/anirbanroy/Desktop/halocats/")
    catalog_index = matcher.build_catalog_index()
    
    # Optional: View catalog statistics
    if catalog_index is not None:
        print("\nCatalog statistics:")
        print(catalog_index[['sigma_8', 'omega_m']].describe())
    
    # Step 2: Find specific cosmology
    target_sigma8 = 0.82
    target_omega_m = 0.31
    
    closest_file = matcher.find_closest_halocat(target_sigma8, target_omega_m)
    print(f"\nClosest halocat to σ8={target_sigma8}, Ωm={target_omega_m}:")
    print(f"  {closest_file.name}")
    
```

```python
import numpy as np
from scipy.stats import qmc
from scipy.spatial import KDTree
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec
import pandas as pd

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    - If only astrophysical parameters: uses one fixed halo catalog
    - If cosmological + astrophysical: uses different halo catalogs
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, noise_level=1e1,
                 fixed_astro_params_file=None):
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        self.noise_level = noise_level
        
        # Parameter configuration
        self.param_names = param_names
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        
        # Determine mode based on parameters
        self.has_cosmo_params = any(p in param_names for p in self.cosmological_params)
        self.astrophysical_params = [p for p in param_names if p not in self.cosmological_params]
        
        # Mode flags
        self.fixed_cosmology_mode = not self.has_cosmo_params and len(self.astrophysical_params) > 0
        
        print(f"Mode: {'Fixed cosmology' if self.fixed_cosmology_mode else 'Varying cosmology'}")
        print(f"Parameters to vary: {param_names}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # Fixed astrophysical parameters file
        self.fixed_astro_params_file = fixed_astro_params_file
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        
        # Halocat matcher for finding specific cosmologies
        self.halocat_matcher = None
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
    
    def load_fixed_astro_params(self, filepath):
        """Load pre-generated astrophysical parameters"""
        data = np.load(filepath)
        self.latin_hypercube_samples = data['latin_hypercube_samples']
        
        if 'param_names' in data:
            expected_params = data['param_names'].tolist()
            if isinstance(expected_params[0], bytes):
                expected_params = [p.decode() if isinstance(p, bytes) else p for p in expected_params]
            
            # Check consistency
            if set(expected_params) != set(self.astrophysical_params):
                print(f"Warning: Parameter mismatch")
                print(f"  Expected: {self.astrophysical_params}")
                print(f"  Got: {expected_params}")
        
        print(f"Loaded {len(self.latin_hypercube_samples)} fixed astrophysical parameter sets")
        for i, param in enumerate(self.astrophysical_params):
            if i < self.latin_hypercube_samples.shape[1]:
                print(f"  {param}: [{self.latin_hypercube_samples[:, i].min():.3f}, "
                      f"{self.latin_hypercube_samples[:, i].max():.3f}]")
        
        return self.latin_hypercube_samples
    
    def load_and_fix_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        data = np.load(halocat_file, allow_pickle=True)

        # Extract and fix cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}

        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }

        self.fixed_halocat = {key: data[key].copy() for key in data.keys()}
        self.fixed_cosmology_mode = True

        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")

        return self.fixed_halocat, self.fixed_cosmo_params

    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """Main processing method that handles both modes"""
        # Load fixed astrophysical parameters if specified
        if self.fixed_astro_params_file and self.latin_hypercube_samples is None:
            self.load_fixed_astro_params(self.fixed_astro_params_file)

        # Check if we're in fixed cosmology mode
        if hasattr(self, 'fixed_halocat') and self.fixed_halocat is not None:
            # Fixed cosmology mode - use single halocat
            return self._process_fixed_cosmology_batch(n_samples, store_maps)

        # Varying cosmology mode
        if halocat_files is None:
            raise ValueError("Provide halocat files for varying cosmology mode")

        return self._process_varying_cosmology_batch(halocat_files, store_maps)

    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with ONE fixed cosmology, varying only astrophysics"""
        if n_samples is None:
            n_samples = len(self.latin_hypercube_samples)

        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)

        all_params = []
        all_pk = []
        all_pdf = []

        print(f"Processing {n_samples} astrophysical variations with fixed cosmology")
        print(f"Each sample will have unique noise realization")

        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Get astrophysical parameters
                astro_params = self.latin_hypercube_samples[i]

                # Build parameters with fixed cosmology
                params_internal = self.fixed_cosmo_params.copy()
                for j, param in enumerate(self.astrophysical_params):
                    params_internal[param] = astro_params[j]

                # Create temporary file for the fixed halocat
                import tempfile
                temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
                np.savez(temp_file.name, **self.fixed_halocat)

                try:
                    # Process using LIMpy
                    lim_sim = ll.lim_sims(
                        temp_file.name,
                        self.redshift,
                        model_name=self.model_name,
                        line_name=self.line_name,
                        halo_cutoff_mass=self.mmin,
                        halocat_type="input_cat",
                        parameters=params_internal,
                        ngrid_x=self.ngrid,
                        ngrid_y=self.ngrid,
                        ngrid_z=self.ngrid,
                        boxsize_x=self.boxsize,
                        boxsize_y=self.boxsize,
                        boxsize_z=self.boxsize,
                        nu_obs=220,
                        theta_fwhm=1,
                        dnu_obs=2.2
                    )

                    intensity_3d = lim_sim.make_intensity_grid()
                    intensity_2d = np.mean(intensity_3d, axis=2)

                finally:
                    import os
                    os.unlink(temp_file.name)

                # Add unique noise for each sample using the RNG
                # This ensures each sample gets different noise but is reproducible
                noise = rng.normal(0, self.noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise

                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)

                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)

                # Store only astrophysical parameters
                param_values = [params_internal[p] for p in self.param_names]
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)

                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins

            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                import traceback
                traceback.print_exc()
                continue

        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf

        return len(all_params)

    def generate_latin_hypercube_samples(self, n_samples):
        """Generate Latin hypercube samples for astrophysical parameters"""
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        samples_unit = sampler.random(n=n_samples)
        
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        return samples_scaled
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process halocat for varying cosmology mode"""
        # Load halocat
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters from halocat
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions and process
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology"""
        n_files = len(halocat_files)
        
        # Verify we have enough astrophysical parameters
        if self.astrophysical_params:
            if self.latin_hypercube_samples is None:
                raise ValueError("No astrophysical parameters loaded")
            elif len(self.latin_hypercube_samples) < n_files:
                raise ValueError(f"Not enough parameter samples: have {len(self.latin_hypercube_samples)}, need {n_files}")
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"Processing {n_files} halocats with varying cosmology")
        print(f"Using seed {self.seed} for noise generation")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add unique noise for each sample using the RNG
                noise = rng.normal(0, self.noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store all parameters
                param_values = [params[p] for p in self.param_names]
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def save_results(self, output_dir, seed=None):
        """Save results with appropriate metadata"""
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # Standard naming for compatibility
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Save with metadata
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            seed=seed,
            noise_level=self.noise_level
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            seed=seed,
            noise_level=self.noise_level
        )
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Seed: {seed}")
        print(f"  Noise level: {self.noise_level}")
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")


def generate_and_save_fixed_astro_params(n_samples, param_ranges, seed, output_file):
    """Generate and save fixed astrophysical parameters"""
    param_names = list(param_ranges.keys())
    n_params = len(param_names)
    
    sampler = qmc.LatinHypercube(d=n_params, seed=seed)
    samples_unit = sampler.random(n=n_samples)
    
    samples_scaled = np.zeros((n_samples, n_params))
    for i, param in enumerate(param_names):
        low, high = param_ranges[param]
        samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
    
    output_path = Path(output_file)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    np.savez_compressed(
        output_path,
        latin_hypercube_samples=samples_scaled,
        param_names=param_names,
        param_ranges=param_ranges,
        seed=seed
    )
    
    print(f"Generated and saved {n_samples} parameter sets to {output_file}")
    for i, param in enumerate(param_names):
        print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
    
    return samples_scaled


def process_with_fixed_parameters(
    input_dir,
    output_dir,
    fixed_params_file,
    param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],
    seed=42,
    max_files=None,
    noise_level=1e3,
    store_maps=False
):
    """Process halocats with fixed astrophysical parameters"""
    generator = LIMpyGenerator(
        param_names=param_names, 
        seed=seed,
        noise_level=noise_level,
        fixed_astro_params_file=fixed_params_file
    )
    
    # Setup parameter ranges
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Get halocats
    input_path = Path(input_dir)
    halocat_files = sorted(input_path.glob("halocat_*.npz"))
    
    if max_files:
        halocat_files = halocat_files[:max_files]
    
    print(f"Processing {len(halocat_files)} files")
    print(f"Parameters: {param_names}")
    print(f"Fixed astro params from: {fixed_params_file}")
    print(f"Seed for noise: {seed}")
    
    # Process batch
    n_processed = generator.process_batch(
        halocat_files=halocat_files, 
        store_maps=store_maps
    )
    
    # Save results
    generator.save_results(output_dir, seed=seed)
    
    print(f"Processed {n_processed} simulations")
    
    return generator


# Main execution for FIXED COSMOLOGY
if __name__ == "__main__":
    # Step 1: Generate fixed astrophysical parameters (run ONCE)
    fixed_params_file = "/Users/anirbanroy/Desktop/fixed_astro_params_3000.npz"
    
    if not Path(fixed_params_file).exists():
        print("Generating fixed astrophysical parameters...")
        generate_and_save_fixed_astro_params(
            n_samples=3000,
            param_ranges={'a_off': (4, 10), 'b_off': (0, 2)},
            seed=999,
            output_file=fixed_params_file
        )
    else:
        print(f"Using existing fixed parameters from {fixed_params_file}")
    
    # Step 2: Choose ONE halo catalog for fixed cosmology
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    
    # Step 3: Process with different seeds, fixed cosmology, varying only a_off and b_off
    seeds_to_process = [50, 51, 52, 53]
    noise_levels = [1e-5, 1e1, 1e2, 1e3]
    
    for i, seed in enumerate(seeds_to_process):
        print(f"\n{'='*60}")
        print(f"FIXED COSMOLOGY - Processing with seed {seed}")
        print(f"Noise level: {noise_levels[i]}")
        print(f"{'='*60}")
        
        # Create generator for FIXED cosmology mode
        generator = LIMpyGenerator(
            param_names=['a_off', 'b_off'],  # ONLY astrophysical parameters
            seed=seed,
            noise_level=noise_levels[i],
            fixed_astro_params_file=fixed_params_file
        )
        
        # Load the SINGLE halo catalog for fixed cosmology
        generator.load_and_fix_halocat(single_halocat)
        
        # Process using the fixed cosmology
        n_samples = 10  # Will use 3000 different a_off, b_off combinations
        
        print(f"Processing {n_samples} variations of a_off, b_off with fixed cosmology")
        
        # Process batch with fixed cosmology
        n_processed = generator.process_batch(
            n_samples=n_samples,
            store_maps=False  # Set to True if you want to save maps
        )
        
        # Save results
        output_dir = f"/Users/anirbanroy/Desktop/processed_halos_cosmology_fixed_seed{seed}/"
        generator.save_results(output_dir, seed=seed)
        
        print(f"Completed processing for seed {seed}")
        print(f"Processed {n_processed} simulations with fixed cosmology")
        print(f"Each simulation has unique noise realization")
    
    print(f"\n{'='*60}")
    print("Fixed cosmology processing complete!")
    print(f"{'='*60}")
```

```python
import numpy as np
from scipy.stats import qmc
from scipy.spatial import KDTree
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec
import pandas as pd

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    - If only astrophysical parameters: uses one fixed halo catalog
    - If cosmological + astrophysical: uses different halo catalogs
    - Noise can be fixed or treated as a varying parameter
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, 
                 noise_level=1e1, noise_mode='fixed',
                 noise_range=(1e-5, 1e3),
                 fixed_astro_params_file=None):
        """
        Initialize generator
        
        Args:
            noise_level: Fixed noise level (used when noise_mode='fixed')
            noise_mode: 'fixed' or 'varying' 
            noise_range: (min, max) range for noise when noise_mode='varying'
        """
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        
        # Noise configuration
        self.noise_mode = noise_mode
        self.noise_level = noise_level
        self.noise_range = noise_range
        
        # Check if noise_level is in param_names
        self.noise_as_param = 'noise_level' in param_names
        if self.noise_as_param:
            self.noise_mode = 'varying'
            # Remove noise_level from param_names for internal processing
            param_names = [p for p in param_names if p != 'noise_level']
        
        # Parameter configuration
        self.param_names_internal = param_names  # Without noise_level
        self.param_names = param_names if not self.noise_as_param else param_names + ['noise_level']
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        
        # Determine mode based on parameters
        self.has_cosmo_params = any(p in param_names for p in self.cosmological_params)
        self.astrophysical_params = [p for p in param_names if p not in self.cosmological_params]
        
        # Mode flags
        self.fixed_cosmology_mode = not self.has_cosmo_params and len(self.astrophysical_params) > 0
        
        print(f"Mode: {'Fixed cosmology' if self.fixed_cosmology_mode else 'Varying cosmology'}")
        print(f"Noise mode: {self.noise_mode}")
        if self.noise_mode == 'varying':
            print(f"Noise range: {self.noise_range}")
        else:
            print(f"Fixed noise level: {self.noise_level}")
        print(f"Parameters to vary: {self.param_names}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.noise_values = None  # Store generated noise levels
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # Fixed astrophysical parameters file
        self.fixed_astro_params_file = fixed_astro_params_file
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        
        # Halocat matcher for finding specific cosmologies
        self.halocat_matcher = None
    
    def generate_noise_levels(self, n_samples, method='uniform'):
        """
        Generate noise levels for varying noise mode
        
        Args:
            n_samples: Number of noise levels to generate
            method: 'uniform', 'log_uniform', or 'latin_hypercube'
        """
        if self.noise_mode == 'fixed':
            return np.full(n_samples, self.noise_level)
        
        min_noise, max_noise = self.noise_range
        rng = np.random.RandomState(self.seed + 1000)  # Different seed for noise generation
        
        if method == 'uniform':
            noise_levels = rng.uniform(min_noise, max_noise, n_samples)
        elif method == 'log_uniform':
            log_min = np.log10(min_noise)
            log_max = np.log10(max_noise)
            noise_levels = 10**rng.uniform(log_min, log_max, n_samples)
        elif method == 'latin_hypercube':
            sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
            samples = sampler.random(n=n_samples)
            log_min = np.log10(min_noise)
            log_max = np.log10(max_noise)
            noise_levels = 10**qmc.scale(samples, log_min, log_max).flatten()
        else:
            raise ValueError(f"Unknown method: {method}")
        
        self.noise_values = noise_levels
        return noise_levels
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
        
        # Add noise range if treating as parameter
        if self.noise_as_param:
            self.param_ranges['noise_level'] = self.noise_range
    
    def load_and_fix_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        data = np.load(halocat_file, allow_pickle=True)

        # Extract and fix cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}

        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }

        self.fixed_halocat = {key: data[key].copy() for key in data.keys()}
        self.fixed_cosmology_mode = True

        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")

        return self.fixed_halocat, self.fixed_cosmo_params

    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with ONE fixed cosmology, varying only astrophysics (and possibly noise)"""
        if n_samples is None:
            n_samples = len(self.latin_hypercube_samples) if self.latin_hypercube_samples is not None else 100

        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_samples, method='log_uniform')
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)

        all_params = []
        all_pk = []
        all_pdf = []

        print(f"Processing {n_samples} astrophysical variations with fixed cosmology")
        if self.noise_mode == 'varying':
            print(f"Using varying noise levels from {self.noise_range[0]:.2e} to {self.noise_range[1]:.2e}")
        else:
            print(f"Using fixed noise level: {self.noise_level:.2e}")

        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Get astrophysical parameters
                if self.latin_hypercube_samples is not None and i < len(self.latin_hypercube_samples):
                    astro_params = self.latin_hypercube_samples[i]
                else:
                    # Generate random parameters if needed
                    astro_params = []
                    for param in self.astrophysical_params:
                        low, high = self.param_ranges.get(param, (0, 1))
                        astro_params.append(rng.uniform(low, high))
                    astro_params = np.array(astro_params)

                # Build parameters with fixed cosmology
                params_internal = self.fixed_cosmo_params.copy()
                for j, param in enumerate(self.astrophysical_params):
                    params_internal[param] = astro_params[j]

                # Create temporary file for the fixed halocat
                import tempfile
                temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
                np.savez(temp_file.name, **self.fixed_halocat)

                try:
                    # Process using LIMpy
                    lim_sim = ll.lim_sims(
                        temp_file.name,
                        self.redshift,
                        model_name=self.model_name,
                        line_name=self.line_name,
                        halo_cutoff_mass=self.mmin,
                        halocat_type="input_cat",
                        parameters=params_internal,
                        ngrid_x=self.ngrid,
                        ngrid_y=self.ngrid,
                        ngrid_z=self.ngrid,
                        boxsize_x=self.boxsize,
                        boxsize_y=self.boxsize,
                        boxsize_z=self.boxsize,
                        nu_obs=220,
                        theta_fwhm=1,
                        dnu_obs=2.2
                    )

                    intensity_3d = lim_sim.make_intensity_grid()
                    intensity_2d = np.mean(intensity_3d, axis=2)

                finally:
                    import os
                    os.unlink(temp_file.name)

                # Add noise with appropriate level
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise

                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)

                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)

                # Store parameters (including noise level if varying)
                param_values = [params_internal[p] for p in self.param_names_internal]
                if self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)

                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins

            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                import traceback
                traceback.print_exc()
                continue

        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf

        return len(all_params)
    
    def load_fixed_astro_params(self, filepath):
        """Load pre-generated astrophysical parameters"""
        data = np.load(filepath)
        self.latin_hypercube_samples = data['latin_hypercube_samples']
        
        if 'param_names' in data:
            expected_params = data['param_names'].tolist()
            if isinstance(expected_params[0], bytes):
                expected_params = [p.decode() if isinstance(p, bytes) else p for p in expected_params]
            
            # Check consistency
            if set(expected_params) != set(self.astrophysical_params):
                print(f"Warning: Parameter mismatch")
                print(f"  Expected: {self.astrophysical_params}")
                print(f"  Got: {expected_params}")
        
        print(f"Loaded {len(self.latin_hypercube_samples)} fixed astrophysical parameter sets")
        for i, param in enumerate(self.astrophysical_params):
            if i < self.latin_hypercube_samples.shape[1]:
                print(f"  {param}: [{self.latin_hypercube_samples[:, i].min():.3f}, "
                      f"{self.latin_hypercube_samples[:, i].max():.3f}]")
        
        return self.latin_hypercube_samples
    
    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """Main processing method that handles both modes"""
        # Load fixed astrophysical parameters if specified
        if self.fixed_astro_params_file and self.latin_hypercube_samples is None:
            self.load_fixed_astro_params(self.fixed_astro_params_file)

        # Check if we're in fixed cosmology mode
        if hasattr(self, 'fixed_halocat') and self.fixed_halocat is not None:
            # Fixed cosmology mode - use single halocat
            return self._process_fixed_cosmology_batch(n_samples, store_maps)

        # Varying cosmology mode
        if halocat_files is None:
            raise ValueError("Provide halocat files for varying cosmology mode")

        return self._process_varying_cosmology_batch(halocat_files, store_maps)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None):
        """Save results with appropriate metadata"""
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # Standard naming for compatibility
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Save with metadata
        metadata = {
            'seed': seed,
            'noise_mode': self.noise_mode,
        }
        
        if self.noise_mode == 'fixed':
            metadata['noise_level'] = self.noise_level
        else:
            metadata['noise_range'] = self.noise_range
            metadata['noise_values'] = self.noise_values
        
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            **metadata
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            **metadata
        )
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Seed: {seed}")
        print(f"  Noise mode: {self.noise_mode}")
        
        if self.noise_mode == 'varying':
            print(f"  Noise range: {self.noise_range}")
            print(f"  Actual noise values: [{self.noise_values.min():.2e}, {self.noise_values.max():.2e}]")
        else:
            print(f"  Noise level: {self.noise_level}")
        
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology - updated to handle varying noise"""
        n_files = len(halocat_files)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_files, method='log_uniform')
        
        # Verify we have enough astrophysical parameters
        if self.astrophysical_params:
            if self.latin_hypercube_samples is None:
                raise ValueError("No astrophysical parameters loaded")
            elif len(self.latin_hypercube_samples) < n_files:
                raise ValueError(f"Not enough parameter samples: have {len(self.latin_hypercube_samples)}, need {n_files}")
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"Processing {n_files} halocats with varying cosmology")
        print(f"Using seed {self.seed} for noise generation")
        if self.noise_mode == 'varying':
            print(f"Using varying noise levels")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise with appropriate level
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters (including noise level if varying)
                param_values = [params[p] for p in self.param_names_internal]
                if self.noise_as_param:
                    param_values.append(current_noise_level)
                    
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process halocat for varying cosmology mode"""
        # Load halocat
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters from halocat
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions and process
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)


# Example usage
if __name__ == "__main__":
    
    # Example 2: Varying noise as a parameter
    print("\nExample 2: Varying noise as a parameter")
    generator_varying = LIMpyGenerator(
        param_names=['a_off', 'b_off', 'noise_level'],  # noise_level included as parameter
        seed=42,
        noise_mode='varying',  # Will be set automatically when noise_level is in param_names
        noise_range=(1e0, 1e4)
    )
    
    # Process with fixed cosmology
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    
    # Load fixed astro paramsimport numpy as np
from scipy.stats import qmc
from scipy.spatial import KDTree
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec
import pandas as pd

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    - If only astrophysical parameters: uses one fixed halo catalog
    - If cosmological + astrophysical: uses different halo catalogs
    - Noise can be fixed or treated as a varying parameter (in linear or log space)
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, 
                 noise_level=1e1, noise_mode='fixed',
                 noise_range=(1e-5, 1e3),
                 log_noise_range=(0, 4),  # Range for log10(noise_level)
                 fixed_astro_params_file=None):
        """
        Initialize generator
        
        Args:
            noise_level: Fixed noise level (used when noise_mode='fixed')
            noise_mode: 'fixed' or 'varying' 
            noise_range: (min, max) range for noise when noise_mode='varying' (linear space)
            log_noise_range: (min, max) range for log10(noise) when using log_noise_level parameter
        """
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        
        # Noise configuration
        self.noise_mode = noise_mode
        self.noise_level = noise_level
        self.noise_range = noise_range
        self.log_noise_range = log_noise_range
        
        # Check if noise is in param_names (either linear or log)
        self.noise_as_param = 'noise_level' in param_names
        self.log_noise_as_param = 'log_noise_level' in param_names
        
        if self.noise_as_param or self.log_noise_as_param:
            self.noise_mode = 'varying'
            # Remove noise parameters from internal param_names for processing
            param_names_filtered = [p for p in param_names if p not in ['noise_level', 'log_noise_level']]
        else:
            param_names_filtered = param_names
        
        # Parameter configuration
        self.param_names_internal = param_names_filtered  # Without noise parameters
        self.param_names = param_names  # Original, including noise if specified
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        
        # Determine mode based on parameters
        self.has_cosmo_params = any(p in param_names_filtered for p in self.cosmological_params)
        self.astrophysical_params = [p for p in param_names_filtered if p not in self.cosmological_params]
        
        # Mode flags
        self.fixed_cosmology_mode = not self.has_cosmo_params and len(self.astrophysical_params) > 0
        
        print(f"Mode: {'Fixed cosmology' if self.fixed_cosmology_mode else 'Varying cosmology'}")
        print(f"Noise mode: {self.noise_mode}")
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"Log noise range: {self.log_noise_range}")
            else:
                print(f"Noise range: {self.noise_range}")
        else:
            print(f"Fixed noise level: {self.noise_level}")
        print(f"Parameters to vary: {self.param_names}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.noise_values = None  # Store generated noise levels (linear)
        self.log_noise_values = None  # Store log10(noise_levels)
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # Fixed astrophysical parameters file
        self.fixed_astro_params_file = fixed_astro_params_file
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        
        # Halocat matcher for finding specific cosmologies
        self.halocat_matcher = None
    
    def generate_noise_levels(self, n_samples, method='uniform'):
        """
        Generate noise levels for varying noise mode
        
        Args:
            n_samples: Number of noise levels to generate
            method: 'uniform', 'log_uniform', or 'latin_hypercube'
        """
        if self.noise_mode == 'fixed':
            self.noise_values = np.full(n_samples, self.noise_level)
            self.log_noise_values = np.log10(self.noise_values)
            return self.noise_values
        
        rng = np.random.RandomState(self.seed + 1000)  # Different seed for noise generation
        
        if self.log_noise_as_param:
            # Generate in log space directly
            log_min, log_max = self.log_noise_range
            
            if method == 'uniform' or method == 'log_uniform':
                # For log parameter, uniform sampling in log space
                log_noise_levels = rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_noise_levels = qmc.scale(samples, log_min, log_max).flatten()
            else:
                raise ValueError(f"Unknown method: {method}")
            
            self.log_noise_values = log_noise_levels
            self.noise_values = 10**log_noise_levels
            
        else:
            # Generate in linear space
            min_noise, max_noise = self.noise_range
            
            if method == 'uniform':
                noise_levels = rng.uniform(min_noise, max_noise, n_samples)
            elif method == 'log_uniform':
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**qmc.scale(samples, log_min, log_max).flatten()
            else:
                raise ValueError(f"Unknown method: {method}")
            
            self.noise_values = noise_levels
            self.log_noise_values = np.log10(noise_levels)
        
        return self.noise_values
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
        
        # Add noise range if treating as parameter
        if self.noise_as_param:
            self.param_ranges['noise_level'] = self.noise_range
        elif self.log_noise_as_param:
            self.param_ranges['log_noise_level'] = self.log_noise_range
    
    def load_fixed_astro_params(self, filepath):
        """Load pre-generated astrophysical parameters"""
        data = np.load(filepath)
        self.latin_hypercube_samples = data['latin_hypercube_samples']
        
        if 'param_names' in data:
            expected_params = data['param_names'].tolist()
            if isinstance(expected_params[0], bytes):
                expected_params = [p.decode() if isinstance(p, bytes) else p for p in expected_params]
            
            # Check consistency
            if set(expected_params) != set(self.astrophysical_params):
                print(f"Warning: Parameter mismatch")
                print(f"  Expected: {self.astrophysical_params}")
                print(f"  Got: {expected_params}")
        
        print(f"Loaded {len(self.latin_hypercube_samples)} fixed astrophysical parameter sets")
        for i, param in enumerate(self.astrophysical_params):
            if i < self.latin_hypercube_samples.shape[1]:
                print(f"  {param}: [{self.latin_hypercube_samples[:, i].min():.3f}, "
                      f"{self.latin_hypercube_samples[:, i].max():.3f}]")
        
        return self.latin_hypercube_samples
    
    def load_and_fix_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        data = np.load(halocat_file, allow_pickle=True)

        # Extract and fix cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}

        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }

        self.fixed_halocat = {key: data[key].copy() for key in data.keys()}
        self.fixed_cosmology_mode = True

        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")

        return self.fixed_halocat, self.fixed_cosmo_params

    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """Main processing method that handles both modes"""
        # Load fixed astrophysical parameters if specified
        if self.fixed_astro_params_file and self.latin_hypercube_samples is None:
            self.load_fixed_astro_params(self.fixed_astro_params_file)

        # Check if we're in fixed cosmology mode
        if hasattr(self, 'fixed_halocat') and self.fixed_halocat is not None:
            # Fixed cosmology mode - use single halocat
            return self._process_fixed_cosmology_batch(n_samples, store_maps)

        # Varying cosmology mode
        if halocat_files is None:
            raise ValueError("Provide halocat files for varying cosmology mode")

        return self._process_varying_cosmology_batch(halocat_files, store_maps)

    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with ONE fixed cosmology, varying only astrophysics (and possibly noise)"""
        if n_samples is None:
            n_samples = len(self.latin_hypercube_samples) if self.latin_hypercube_samples is not None else 100

        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_samples, method='latin_hypercube')
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)

        all_params = []
        all_pk = []
        all_pdf = []

        print(f"Processing {n_samples} astrophysical variations with fixed cosmology")
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"Using varying log_noise_level from {self.log_noise_range[0]} to {self.log_noise_range[1]}")
                print(f"Actual noise levels: [{self.noise_values.min():.2e}, {self.noise_values.max():.2e}]")
            else:
                print(f"Using varying noise levels from {self.noise_range[0]:.2e} to {self.noise_range[1]:.2e}")
        else:
            print(f"Using fixed noise level: {self.noise_level:.2e}")

        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Get astrophysical parameters
                if self.latin_hypercube_samples is not None and i < len(self.latin_hypercube_samples):
                    astro_params = self.latin_hypercube_samples[i]
                else:
                    # Generate random parameters if needed
                    astro_params = []
                    for param in self.astrophysical_params:
                        low, high = self.param_ranges.get(param, (0, 1))
                        astro_params.append(rng.uniform(low, high))
                    astro_params = np.array(astro_params)

                # Build parameters with fixed cosmology
                params_internal = self.fixed_cosmo_params.copy()
                for j, param in enumerate(self.astrophysical_params):
                    params_internal[param] = astro_params[j]

                # Create temporary file for the fixed halocat
                import tempfile
                temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
                np.savez(temp_file.name, **self.fixed_halocat)

                try:
                    # Process using LIMpy
                    lim_sim = ll.lim_sims(
                        temp_file.name,
                        self.redshift,
                        model_name=self.model_name,
                        line_name=self.line_name,
                        halo_cutoff_mass=self.mmin,
                        halocat_type="input_cat",
                        parameters=params_internal,
                        ngrid_x=self.ngrid,
                        ngrid_y=self.ngrid,
                        ngrid_z=self.ngrid,
                        boxsize_x=self.boxsize,
                        boxsize_y=self.boxsize,
                        boxsize_z=self.boxsize,
                        nu_obs=220,
                        theta_fwhm=1,
                        dnu_obs=2.2
                    )

                    intensity_3d = lim_sim.make_intensity_grid()
                    intensity_2d = np.mean(intensity_3d, axis=2)

                finally:
                    import os
                    os.unlink(temp_file.name)

                # Add noise with appropriate level
                current_noise_level = noise_levels[i]  # Linear noise level
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise

                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)

                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)

                # Store parameters
                param_values = [params_internal[p] for p in self.param_names_internal]
                
                # Add noise parameter (log or linear as specified)
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)

                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins

            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                import traceback
                traceback.print_exc()
                continue

        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf

        return len(all_params)
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology"""
        n_files = len(halocat_files)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_files, method='latin_hypercube')
        
        # Verify we have enough astrophysical parameters
        if self.astrophysical_params:
            if self.latin_hypercube_samples is None:
                raise ValueError("No astrophysical parameters loaded")
            elif len(self.latin_hypercube_samples) < n_files:
                raise ValueError(f"Not enough parameter samples: have {len(self.latin_hypercube_samples)}, need {n_files}")
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"Processing {n_files} halocats with varying cosmology")
        print(f"Using seed {self.seed} for noise generation")
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"Using varying log_noise_level")
            else:
                print(f"Using varying noise levels")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise with appropriate level
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters (including noise level if varying)
                param_values = [params[p] for p in self.param_names_internal]
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                    
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def generate_latin_hypercube_samples(self, n_samples):
        """Generate Latin hypercube samples for astrophysical parameters"""
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        samples_unit = sampler.random(n=n_samples)
        
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        return samples_scaled
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process halocat for varying cosmology mode"""
        # Load halocat
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters from halocat
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions and process
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None):
        """Save results with appropriate metadata"""
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # Standard naming for compatibility
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Save with metadata
        metadata = {
            'seed': seed,
            'noise_mode': self.noise_mode,
        }
        
        if self.noise_mode == 'fixed':
            metadata['noise_level'] = self.noise_level
        else:
            metadata['noise_range'] = self.noise_range if not self.log_noise_as_param else self.log_noise_range
            metadata['noise_values'] = self.noise_values
            if self.log_noise_as_param:
                metadata['log_noise_values'] = self.log_noise_values
        
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            **metadata
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            **metadata
        )
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Seed: {seed}")
        print(f"  Noise mode: {self.noise_mode}")
        
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"  Log noise range: {self.log_noise_range}")
                print(f"  Actual noise values: [{self.noise_values.min():.2e}, {self.noise_values.max():.2e}]")
            else:
                print(f"  Noise range: {self.noise_range}")
        else:
            print(f"  Noise level: {self.noise_level}")
        
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            values = self.params_array[:, i]
            if 'log' in name:
                print(f"  {name}: [{values.min():.3f}, {values.max():.3f}] (10^x)")
            else:
                print(f"  {name}: [{values.min():.3f}, {values.max():.3f}]")


def generate_and_save_fixed_astro_params(n_samples, param_ranges, seed, output_file):
    """Generate and save fixed astrophysical parameters"""
    param_names = list(param_ranges.keys())
    n_params = len(param_names)
    
    sampler = qmc.LatinHypercube(d=n_params, seed=seed)
    samples_unit = sampler.random(n=n_samples)
    
    samples_scaled = np.zeros((n_samples, n_params))
    for i, param in enumerate(param_names):
        low, high = param_ranges[param]
        samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
    
    output_path = Path(output_file)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    np.savez_compressed(
        output_path,
        latin_hypercube_samples=samples_scaled,
        param_names=param_names,
        param_ranges=param_ranges,
        seed=seed
    )
    
    print(f"Generated and saved {n_samples} parameter sets to {output_file}")
    for i, param in enumerate(param_names):
        print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
    
    return samples_scaled


# Main execution examples
if __name__ == "__main__":
    # Example 1: Fixed cosmology with log_noise_level as parameter
    print("="*70)
    print("EXAMPLE 1: Fixed cosmology with log_noise_level as parameter")
    print("="*70)
    
    # Step 1: Generate fixed astrophysical parameters (run ONCE)
    fixed_params_file = "/Users/anirbanroy/Desktop/fixed_astro_params_3000.npz"
    
    if not Path(fixed_params_file).exists():
        print("Generating fixed astrophysical parameters...")
        generate_and_save_fixed_astro_params(
            n_samples=3000,
            param_ranges={'a_off': (4, 10), 'b_off': (0, 2)},
            seed=999,
            output_file=fixed_params_file
        )
    else:
        print(f"Using existing fixed parameters from {fixed_params_file}")
    
    # Step 2: Process with log_noise_level as parameter
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    
    generator = LIMpyGenerator(
        param_names=['a_off', 'b_off', 'log_noise_level'],  # log_noise_level as parameter
        seed=42,
        log_noise_range=(0, 4),  # log10(noise) from 0 to 4 (1 to 10^4)
        fixed_astro_params_file=fixed_params_file
    )
    
    # Load the single halo catalog for fixed cosmology
    generator.load_and_fix_halocat(single_halocat)
    generator.setup_parameter_ranges()
    
    # Process
    n_processed = generator.process_batch(
        n_samples=100,  # Will use 100 different combinations
        store_maps=False
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_with_log_noise/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nCompleted processing {n_processed} simulations")
    print("Parameters varied: a_off, b_off, log_noise_level")
    
    # Example 2: Fixed noise level (traditional approach)
    print("\n" + "="*70)
    print("EXAMPLE 2: Fixed noise level (traditional)")
    print("="*70)
    
    generator_fixed = LIMpyGenerator(
        param_names=['a_off', 'b_off'],  # No noise in parameters
        seed=43,
        noise_mode='fixed',
        noise_level=1e2,  # Fixed at 100
        fixed_astro_params_file=fixed_params_file
    )
    
    generator_fixed.load_and_fix_halocat(single_halocat)
    generator_fixed.setup_parameter_ranges()
    
    n_processed_fixed = generator_fixed.process_batch(
        n_samples=100,
        store_maps=False
    )
    
    output_dir_fixed = "/Users/anirbanroy/Desktop/processed_fixed_noise/"
    generator_fixed.save_results(output_dir_fixed, seed=43)
    
    print(f"\nCompleted processing {n_processed_fixed} simulations with fixed noise")
    fixed_params_file = "/Users/anirbanroy/Desktop/fixed_astro_params_3000.npz"
    
    generator_varying.fixed_astro_params_file = fixed_params_file
    generator_varying.load_and_fix_halocat(single_halocat)
    generator_varying.setup_parameter_ranges()
    
    # Process
    n_processed = generator_varying.process_batch(
        n_samples=3000,  # Will generate 100 samples with different noise levels
        store_maps=False
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_with_fixed_cosmology_varying_noise/"
    generator_varying.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} simulations")
    print("Each simulation has:")
    print("  - Fixed cosmology")
    print("  - Varying a_off and b_off")
    print("  - Varying noise level (as a parameter)")
```

```python
import numpy as np
from scipy.stats import qmc
from scipy.spatial import KDTree
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec
import pandas as pd

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    - If only astrophysical parameters: uses one fixed halo catalog
    - If cosmological + astrophysical: uses different halo catalogs
    - Noise can be fixed or treated as a varying parameter (in linear or log space)
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, 
                 noise_level=1e1, noise_mode='fixed',
                 noise_range=(1e-5, 1e3),
                 log_noise_range=(0, 4),  # Range for log10(noise_level)
                 fixed_astro_params_file=None):
        """
        Initialize generator
        
        Args:
            noise_level: Fixed noise level (used when noise_mode='fixed')
            noise_mode: 'fixed' or 'varying' 
            noise_range: (min, max) range for noise when noise_mode='varying' (linear space)
            log_noise_range: (min, max) range for log10(noise) when using log_noise_level parameter
        """
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        
        # Noise configuration
        self.noise_mode = noise_mode
        self.noise_level = noise_level
        self.noise_range = noise_range
        self.log_noise_range = log_noise_range
        
        # Check if noise is in param_names (either linear or log)
        self.noise_as_param = 'noise_level' in param_names
        self.log_noise_as_param = 'log_noise_level' in param_names
        
        if self.noise_as_param or self.log_noise_as_param:
            self.noise_mode = 'varying'
            # Remove noise parameters from internal param_names for processing
            param_names_filtered = [p for p in param_names if p not in ['noise_level', 'log_noise_level']]
        else:
            param_names_filtered = param_names
        
        # Parameter configuration
        self.param_names_internal = param_names_filtered  # Without noise parameters
        self.param_names = param_names  # Original, including noise if specified
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        
        # Determine mode based on parameters
        self.has_cosmo_params = any(p in param_names_filtered for p in self.cosmological_params)
        self.astrophysical_params = [p for p in param_names_filtered if p not in self.cosmological_params]
        
        # Mode flags
        self.fixed_cosmology_mode = not self.has_cosmo_params and len(self.astrophysical_params) > 0
        
        print(f"Mode: {'Fixed cosmology' if self.fixed_cosmology_mode else 'Varying cosmology'}")
        print(f"Noise mode: {self.noise_mode}")
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"Log noise range: {self.log_noise_range}")
            else:
                print(f"Noise range: {self.noise_range}")
        else:
            print(f"Fixed noise level: {self.noise_level}")
        print(f"Parameters to vary: {self.param_names}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.noise_values = None  # Store generated noise levels (linear)
        self.log_noise_values = None  # Store log10(noise_levels)
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # Fixed astrophysical parameters file
        self.fixed_astro_params_file = fixed_astro_params_file
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        
        # Halocat matcher for finding specific cosmologies
        self.halocat_matcher = None
    
    def generate_noise_levels(self, n_samples, method='uniform'):
        """
        Generate noise levels for varying noise mode
        
        Args:
            n_samples: Number of noise levels to generate
            method: 'uniform', 'log_uniform', or 'latin_hypercube'
        """
        if self.noise_mode == 'fixed':
            self.noise_values = np.full(n_samples, self.noise_level)
            self.log_noise_values = np.log10(self.noise_values)
            return self.noise_values
        
        rng = np.random.RandomState(self.seed + 1000)  # Different seed for noise generation
        
        if self.log_noise_as_param:
            # Generate in log space directly
            log_min, log_max = self.log_noise_range
            
            if method == 'uniform' or method == 'log_uniform':
                # For log parameter, uniform sampling in log space
                log_noise_levels = rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_noise_levels = qmc.scale(samples, log_min, log_max).flatten()
            else:
                raise ValueError(f"Unknown method: {method}")
            
            self.log_noise_values = log_noise_levels
            self.noise_values = 10**log_noise_levels
            
        else:
            # Generate in linear space
            min_noise, max_noise = self.noise_range
            
            if method == 'uniform':
                noise_levels = rng.uniform(min_noise, max_noise, n_samples)
            elif method == 'log_uniform':
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**qmc.scale(samples, log_min, log_max).flatten()
            else:
                raise ValueError(f"Unknown method: {method}")
            
            self.noise_values = noise_levels
            self.log_noise_values = np.log10(noise_levels)
        
        return self.noise_values
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
        
        # Add noise range if treating as parameter
        if self.noise_as_param:
            self.param_ranges['noise_level'] = self.noise_range
        elif self.log_noise_as_param:
            self.param_ranges['log_noise_level'] = self.log_noise_range
    
    def load_fixed_astro_params(self, filepath):
        """Load pre-generated astrophysical parameters"""
        data = np.load(filepath)
        self.latin_hypercube_samples = data['latin_hypercube_samples']
        
        if 'param_names' in data:
            expected_params = data['param_names'].tolist()
            if isinstance(expected_params[0], bytes):
                expected_params = [p.decode() if isinstance(p, bytes) else p for p in expected_params]
            
            # Check consistency
            if set(expected_params) != set(self.astrophysical_params):
                print(f"Warning: Parameter mismatch")
                print(f"  Expected: {self.astrophysical_params}")
                print(f"  Got: {expected_params}")
        
        print(f"Loaded {len(self.latin_hypercube_samples)} fixed astrophysical parameter sets")
        for i, param in enumerate(self.astrophysical_params):
            if i < self.latin_hypercube_samples.shape[1]:
                print(f"  {param}: [{self.latin_hypercube_samples[:, i].min():.3f}, "
                      f"{self.latin_hypercube_samples[:, i].max():.3f}]")
        
        return self.latin_hypercube_samples
    
    def load_and_fix_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        data = np.load(halocat_file, allow_pickle=True)

        # Extract and fix cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}

        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }

        self.fixed_halocat = {key: data[key].copy() for key in data.keys()}
        self.fixed_cosmology_mode = True

        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")

        return self.fixed_halocat, self.fixed_cosmo_params

    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """Main processing method that handles both modes"""
        # Load fixed astrophysical parameters if specified
        if self.fixed_astro_params_file and self.latin_hypercube_samples is None:
            self.load_fixed_astro_params(self.fixed_astro_params_file)

        # Check if we're in fixed cosmology mode
        if hasattr(self, 'fixed_halocat') and self.fixed_halocat is not None:
            # Fixed cosmology mode - use single halocat
            return self._process_fixed_cosmology_batch(n_samples, store_maps)

        # Varying cosmology mode
        if halocat_files is None:
            raise ValueError("Provide halocat files for varying cosmology mode")

        return self._process_varying_cosmology_batch(halocat_files, store_maps)

    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with ONE fixed cosmology, varying only astrophysics (and possibly noise)"""
        if n_samples is None:
            n_samples = len(self.latin_hypercube_samples) if self.latin_hypercube_samples is not None else 100

        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_samples, method='latin_hypercube')
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)

        all_params = []
        all_pk = []
        all_pdf = []

        print(f"Processing {n_samples} astrophysical variations with fixed cosmology")
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"Using varying log_noise_level from {self.log_noise_range[0]} to {self.log_noise_range[1]}")
                print(f"Actual noise levels: [{self.noise_values.min():.2e}, {self.noise_values.max():.2e}]")
            else:
                print(f"Using varying noise levels from {self.noise_range[0]:.2e} to {self.noise_range[1]:.2e}")
        else:
            print(f"Using fixed noise level: {self.noise_level:.2e}")

        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Get astrophysical parameters
                if self.latin_hypercube_samples is not None and i < len(self.latin_hypercube_samples):
                    astro_params = self.latin_hypercube_samples[i]
                else:
                    # Generate random parameters if needed
                    astro_params = []
                    for param in self.astrophysical_params:
                        low, high = self.param_ranges.get(param, (0, 1))
                        astro_params.append(rng.uniform(low, high))
                    astro_params = np.array(astro_params)

                # Build parameters with fixed cosmology
                params_internal = self.fixed_cosmo_params.copy()
                for j, param in enumerate(self.astrophysical_params):
                    params_internal[param] = astro_params[j]

                # Create temporary file for the fixed halocat
                import tempfile
                temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
                np.savez(temp_file.name, **self.fixed_halocat)

                try:
                    # Process using LIMpy
                    lim_sim = ll.lim_sims(
                        temp_file.name,
                        self.redshift,
                        model_name=self.model_name,
                        line_name=self.line_name,
                        halo_cutoff_mass=self.mmin,
                        halocat_type="input_cat",
                        parameters=params_internal,
                        ngrid_x=self.ngrid,
                        ngrid_y=self.ngrid,
                        ngrid_z=self.ngrid,
                        boxsize_x=self.boxsize,
                        boxsize_y=self.boxsize,
                        boxsize_z=self.boxsize,
                        nu_obs=220,
                        theta_fwhm=1,
                        dnu_obs=2.2
                    )

                    intensity_3d = lim_sim.make_intensity_grid()
                    intensity_2d = np.mean(intensity_3d, axis=2)

                finally:
                    import os
                    os.unlink(temp_file.name)

                # Add noise with appropriate level
                current_noise_level = noise_levels[i]  # Linear noise level
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise

                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)

                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)

                # Store parameters
                param_values = [params_internal[p] for p in self.param_names_internal]
                
                # Add noise parameter (log or linear as specified)
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)

                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins

            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                import traceback
                traceback.print_exc()
                continue

        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf

        return len(all_params)
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology"""
        n_files = len(halocat_files)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_files, method='latin_hypercube')
        
        # Verify we have enough astrophysical parameters
        if self.astrophysical_params:
            if self.latin_hypercube_samples is None:
                raise ValueError("No astrophysical parameters loaded")
            elif len(self.latin_hypercube_samples) < n_files:
                raise ValueError(f"Not enough parameter samples: have {len(self.latin_hypercube_samples)}, need {n_files}")
        
        # Create random number generator with base seed
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"Processing {n_files} halocats with varying cosmology")
        print(f"Using seed {self.seed} for noise generation")
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"Using varying log_noise_level")
            else:
                print(f"Using varying noise levels")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise with appropriate level
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters (including noise level if varying)
                param_values = [params[p] for p in self.param_names_internal]
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                    
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def generate_latin_hypercube_samples(self, n_samples):
        """Generate Latin hypercube samples for astrophysical parameters"""
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        samples_unit = sampler.random(n=n_samples)
        
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        return samples_scaled
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process halocat for varying cosmology mode"""
        # Load halocat
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters from halocat
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions and process
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None):
        """Save results with appropriate metadata"""
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # Standard naming for compatibility
        ps_file = output_path / f"ps_80_256.npz"
        pdf_file = output_path / f"pdf_80_256.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Save with metadata
        metadata = {
            'seed': seed,
            'noise_mode': self.noise_mode,
        }
        
        if self.noise_mode == 'fixed':
            metadata['noise_level'] = self.noise_level
        else:
            metadata['noise_range'] = self.noise_range if not self.log_noise_as_param else self.log_noise_range
            metadata['noise_values'] = self.noise_values
            if self.log_noise_as_param:
                metadata['log_noise_values'] = self.log_noise_values
        
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            **metadata
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            **metadata
        )
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Seed: {seed}")
        print(f"  Noise mode: {self.noise_mode}")
        
        if self.noise_mode == 'varying':
            if self.log_noise_as_param:
                print(f"  Log noise range: {self.log_noise_range}")
                print(f"  Actual noise values: [{self.noise_values.min():.2e}, {self.noise_values.max():.2e}]")
            else:
                print(f"  Noise range: {self.noise_range}")
        else:
            print(f"  Noise level: {self.noise_level}")
        
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            values = self.params_array[:, i]
            if 'log' in name:
                print(f"  {name}: [{values.min():.3f}, {values.max():.3f}] (10^x)")
            else:
                print(f"  {name}: [{values.min():.3f}, {values.max():.3f}]")


def generate_and_save_fixed_astro_params(n_samples, param_ranges, seed, output_file):
    """Generate and save fixed astrophysical parameters"""
    param_names = list(param_ranges.keys())
    n_params = len(param_names)
    
    sampler = qmc.LatinHypercube(d=n_params, seed=seed)
    samples_unit = sampler.random(n=n_samples)
    
    samples_scaled = np.zeros((n_samples, n_params))
    for i, param in enumerate(param_names):
        low, high = param_ranges[param]
        samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
    
    output_path = Path(output_file)
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    np.savez_compressed(
        output_path,
        latin_hypercube_samples=samples_scaled,
        param_names=param_names,
        param_ranges=param_ranges,
        seed=seed
    )
    
    print(f"Generated and saved {n_samples} parameter sets to {output_file}")
    for i, param in enumerate(param_names):
        print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
    
    return samples_scaled


# Main execution examples
if __name__ == "__main__":
    # Example 1: Fixed cosmology with log_noise_level as parameter
    print("="*70)
    print("EXAMPLE 1: Fixed cosmology with log_noise_level as parameter")
    print("="*70)
    
    # Step 1: Generate fixed astrophysical parameters (run ONCE)
    fixed_params_file = "/Users/anirbanroy/Desktop/fixed_astro_params_3000.npz"
    
    if not Path(fixed_params_file).exists():
        print("Generating fixed astrophysical parameters...")
        generate_and_save_fixed_astro_params(
            n_samples=3000,
            param_ranges={'a_off': (4, 10), 'b_off': (0, 2)},
            seed=999,
            output_file=fixed_params_file
        )
    else:
        print(f"Using existing fixed parameters from {fixed_params_file}")
    
    # Step 2: Process with log_noise_level as parameter
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    
    generator = LIMpyGenerator(
        param_names=["sigma_8", "omega_m", 'a_off', 'b_off'],  # log_noise_level as parameter
        seed=42,
        log_noise_range=(0, 4),  # log10(noise) from 0 to 4 (1 to 10^4)
        fixed_astro_params_file=fixed_params_file,
        noise_level=1e2, noise_mode='fixed',
    )
    
    # Load the single halo catalog for fixed cosmology
    #generator.load_and_fix_halocat(single_halocat)
    generator.setup_parameter_ranges()
    
    # Process
    n_processed = generator.process_batch(
        n_samples=10,  # Will use 100 different combinations
        store_maps=False
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_4p_with_noise/"
    generator.save_results(output_dir, seed=42)
    
```

```python
import numpy as np
from scipy.stats import qmc
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    
    Two operational modes:
    1. VARYING COSMOLOGY: Different halo catalogs for each parameter set (cosmo + astro params)
    2. FIXED COSMOLOGY: Single halo catalog with varying astrophysical parameters only
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, 
                 noise_level=1e1, noise_mode='fixed',
                 noise_range=(1e-5, 1e3),
                 log_noise_range=(0, 4),
                 mode='varying_cosmology'):
        """
        Initialize generator
        
        Args:
            mode: 'varying_cosmology' or 'fixed_cosmology'
            param_names: List of parameters to vary
                - For varying_cosmology: can include cosmo + astro params
                - For fixed_cosmology: should only include astro params (cosmo will be fixed from halocat)
        """
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        self.mode = mode
        
        # Noise configuration
        self.noise_mode = noise_mode
        self.noise_level = noise_level
        self.noise_range = noise_range
        self.log_noise_range = log_noise_range
        
        # Check if noise is in param_names
        self.noise_as_param = 'noise_level' in param_names
        self.log_noise_as_param = 'log_noise_level' in param_names
        
        if self.noise_as_param or self.log_noise_as_param:
            self.noise_mode = 'varying'
            param_names_filtered = [p for p in param_names if p not in ['noise_level', 'log_noise_level']]
        else:
            param_names_filtered = param_names
        
        # Parameter configuration
        self.param_names_internal = param_names_filtered
        self.param_names = param_names
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        self.astrophysical_params = [p for p in param_names_filtered if p not in self.cosmological_params]
        
        # Validate mode and parameters
        if mode == 'fixed_cosmology':
            cosmo_in_params = [p for p in param_names_filtered if p in self.cosmological_params]
            if cosmo_in_params:
                print(f"WARNING: In fixed_cosmology mode, cosmological parameters {cosmo_in_params} will be ignored")
                print("They will be taken from the fixed halo catalog instead")
        
        print(f"Mode: {mode}")
        print(f"Parameters to vary: {self.param_names}")
        print(f"  Cosmological: {[p for p in param_names_filtered if p in self.cosmological_params]}")
        print(f"  Astrophysical: {self.astrophysical_params}")
        print(f"Noise mode: {self.noise_mode}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.noise_values = None
        self.log_noise_values = None
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        self.fixed_halocat_file = None
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
        
        # Add noise range if treating as parameter
        if self.noise_as_param:
            self.param_ranges['noise_level'] = self.noise_range
        elif self.log_noise_as_param:
            self.param_ranges['log_noise_level'] = self.log_noise_range
    
    def generate_latin_hypercube_samples(self, n_samples):
        """Generate Latin hypercube samples for astrophysical parameters"""
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        samples_unit = sampler.random(n=n_samples)
        
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        for i, param in enumerate(self.astrophysical_params):
            print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
        
        return samples_scaled
    
    def generate_noise_levels(self, n_samples, method='uniform'):
        """Generate noise levels for varying noise mode"""
        if self.noise_mode == 'fixed':
            self.noise_values = np.full(n_samples, self.noise_level)
            self.log_noise_values = np.log10(self.noise_values)
            return self.noise_values
        
        rng = np.random.RandomState(self.seed + 1000)
        
        if self.log_noise_as_param:
            log_min, log_max = self.log_noise_range
            
            if method == 'uniform' or method == 'log_uniform':
                log_noise_levels = rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_noise_levels = qmc.scale(samples, log_min, log_max).flatten()
            
            self.log_noise_values = log_noise_levels
            self.noise_values = 10**log_noise_levels
        else:
            min_noise, max_noise = self.noise_range
            
            if method == 'uniform':
                noise_levels = rng.uniform(min_noise, max_noise, n_samples)
            elif method == 'log_uniform':
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**qmc.scale(samples, log_min, log_max).flatten()
            
            self.noise_values = noise_levels
            self.log_noise_values = np.log10(noise_levels)
        
        return self.noise_values
    
    def load_fixed_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        self.fixed_halocat_file = halocat_file
        data = np.load(halocat_file, allow_pickle=True)
        
        # Extract cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }
        
        self.fixed_halocat = data
        
        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")
        
        return self.fixed_cosmo_params
    
    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """
        Main processing method that handles both modes
        
        For varying_cosmology mode:
            - Requires halocat_files (list of paths)
            - Each halocat provides its own cosmology
            
        For fixed_cosmology mode:
            - Requires n_samples (number of realizations)
            - Must call load_fixed_halocat() first
        """
        if self.mode == 'fixed_cosmology':
            if self.fixed_halocat is None:
                raise ValueError("For fixed_cosmology mode, call load_fixed_halocat() first")
            return self._process_fixed_cosmology_batch(n_samples, store_maps)
        else:
            if halocat_files is None:
                raise ValueError("For varying_cosmology mode, provide halocat_files")
            return self._process_varying_cosmology_batch(halocat_files, store_maps)
    
    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with fixed cosmology, varying only astrophysics"""
        if n_samples is None:
            raise ValueError("Specify n_samples for fixed_cosmology mode")
        
        # Generate astrophysical parameters if needed
        if self.astrophysical_params and self.latin_hypercube_samples is None:
            self.generate_latin_hypercube_samples(n_samples)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_samples, method='latin_hypercube')
        
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"\nProcessing {n_samples} realizations with fixed cosmology")
        print(f"Fixed cosmology: {self.fixed_cosmo_params}")
        
        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Build parameters
                params_internal = self.fixed_cosmo_params.copy()
                
                # Add astrophysical parameters
                if self.latin_hypercube_samples is not None:
                    for j, param in enumerate(self.astrophysical_params):
                        params_internal[param] = self.latin_hypercube_samples[i, j]
                
                # Process with LIMpy
                intensity_2d, _, _ = self.process_single_halocat(
                    self.fixed_halocat_file, 
                    index=i, 
                    astro_params=self.latin_hypercube_samples[i] if self.latin_hypercube_samples is not None else None
                )
                
                # Add noise
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters (only the ones we're varying)
                param_values = []
                for param_name in self.param_names_internal:
                    if param_name in self.cosmological_params:
                        # For fixed cosmology mode, still include cosmo params in output if requested
                        param_values.append(params_internal[param_name])
                    else:
                        param_values.append(params_internal[param_name])
                
                # Add noise parameter if varying
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology using different halocats"""
        n_files = len(halocat_files)
        
        # Generate astrophysical parameters if needed
        if self.astrophysical_params and self.latin_hypercube_samples is None:
            self.generate_latin_hypercube_samples(n_files)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_files, method='latin_hypercube')
        
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"\nProcessing {n_files} halocats with varying cosmology")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters
                param_values = [params[p] for p in self.param_names_internal]
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process single halocat with parameter combination"""
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None):
        """Save results with metadata"""
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # File naming
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Metadata
        metadata = {
            'seed': seed,
            'mode': self.mode,
            'noise_mode': self.noise_mode,
        }
        
        if self.noise_mode == 'fixed':
            metadata['noise_level'] = self.noise_level
        else:
            if self.log_noise_as_param:
                metadata['log_noise_range'] = self.log_noise_range
                metadata['log_noise_values'] = self.log_noise_values
            else:
                metadata['noise_range'] = self.noise_range
            metadata['noise_values'] = self.noise_values
        
        if self.mode == 'fixed_cosmology' and self.fixed_cosmo_params:
            metadata['fixed_cosmology'] = self.fixed_cosmo_params
        
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            **metadata
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            **metadata
        )
        
        # Save parameter configuration
        config = {
            'mode': self.mode,
            'param_names': self.param_names,
            'param_ranges': self.param_ranges,
            'seed': seed,
            'noise_mode': self.noise_mode,
            'noise_level': self.noise_level if self.noise_mode == 'fixed' else None,
            'noise_range': self.noise_range if self.noise_as_param else None,
            'log_noise_range': self.log_noise_range if self.log_noise_as_param else None,
            'boxsize': self.boxsize,
            'ngrid': self.ngrid,
            'redshift': self.redshift,
            'line_name': self.line_name,
            'model_name': self.model_name
        }
        
        if self.mode == 'fixed_cosmology' and self.fixed_cosmo_params:
            config['fixed_cosmology'] = self.fixed_cosmo_params
        
        config_file = output_path / 'parameter_config.json'
        with open(config_file, 'w') as f:
            json.dump(config, f, indent=2)
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Config: {config_file}")
        print(f"\nMode: {self.mode}")
        print(f"Seed: {seed}")
        print(f"Parameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            values = self.params_array[:, i]
            print(f"  {name}: [{values.min():.3f}, {values.max():.3f}]")


# ============================================================================
# EXAMPLE USAGE
# ============================================================================

def example_varying_cosmology():
    """
    EXAMPLE 1: Varying Cosmology Mode
    Uses different halo catalogs, each with its own cosmology
    """
    print("="*80)
    print("EXAMPLE 1: VARYING COSMOLOGY MODE")
    print("Uses different halo catalogs for each parameter combination")
    print("="*80)
    
    # Initialize generator with cosmo + astro parameters
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],
        mode='varying_cosmology',  # Explicitly set mode
        seed=42,
        noise_level=1e-5,
        noise_mode='fixed'
    )
    
    # Setup parameter ranges for astrophysical parameters
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Get halo catalogs
    input_dir = Path("/Users/anirbanroy/Desktop/halocats/")
    halocat_files = sorted(input_dir.glob("halocat_*.npz"))[:100]  # Use 100 files
    
    print(f"\nProcessing {len(halocat_files)} halo catalogs")
    print("Each catalog provides its own cosmology (sigma_8, omega_m)")
    print("Astrophysical parameters (a_off, b_off) are sampled via Latin hypercube")
    
    # Process
    n_processed = generator.process_batch(
        halocat_files=halocat_files,
        store_maps=True  # Store for visualization
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_varying_cosmo/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_fixed_cosmology():
    """
    EXAMPLE 2: Fixed Cosmology Mode
    Uses ONE halo catalog repeatedly with different astrophysical parameters
    """
    print("="*80)
    print("EXAMPLE 2: FIXED COSMOLOGY MODE")
    print("Uses single halo catalog with varying astrophysical parameters only")
    print("="*80)
    
    # Initialize generator for fixed cosmology mode
    generator = LIMpyGenerator(
        param_names=['a_off', 'b_off'],  # Only astrophysical parameters
        mode='fixed_cosmology',  # Explicitly set mode
        seed=42,
        noise_level=1e-5,
        noise_mode='fixed'
    )
    
    # Setup parameter ranges
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Load the single halo catalog to use
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    fixed_cosmo = generator.load_fixed_halocat(single_halocat)
    
    print("\nGenerating 100 realizations with fixed cosmology")
    print("Only astrophysical parameters (a_off, b_off) will vary")
    
    # Process with fixed cosmology
    n_processed = generator.process_batch(
        n_samples=100,  # Number of realizations
        store_maps=True
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_fixed_cosmo/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_fixed_cosmology_with_noise():
    """
    EXAMPLE 3: Fixed Cosmology with Noise as Parameter
    Uses ONE halo catalog with varying astrophysical parameters AND noise
    """
    print("="*80)
    print("EXAMPLE 3: FIXED COSMOLOGY WITH NOISE AS PARAMETER")
    print("Uses single halo catalog with varying astro params + noise level")
    print("="*80)
    
    # Initialize with noise as a parameter
    generator = LIMpyGenerator(
        param_names=['a_off', 'b_off', 'log_noise_level'],  # Include noise
        mode='fixed_cosmology',
        seed=42,
        log_noise_range=(0, 4),  # log10(noise) from 10^0 to 10^4
    )
    
    # Setup parameter ranges
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Load the single halo catalog
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    generator.load_fixed_halocat(single_halocat)
    
    print("\nGenerating 100 realizations with:")
    print("  - Fixed cosmology from halocat")
    print("  - Varying a_off and b_off")
    print("  - Varying log_noise_level from 0 to 4")
    
    # Process
    n_processed = generator.process_batch(
        n_samples=100,
        store_maps=False
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_fixed_cosmo_with_noise/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_hybrid_mode():
    """
    EXAMPLE 4: Hybrid - Include cosmo params in output even with fixed cosmology
    This is useful when you want the output to have cosmo params for compatibility
    """
    print("="*80)
    print("EXAMPLE 4: FIXED COSMOLOGY BUT INCLUDE COSMO PARAMS IN OUTPUT")
    print("Useful for maintaining consistent output format")
    print("="*80)
    
    # Initialize with all parameters, but use fixed_cosmology mode
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],  # Include cosmo params
        mode='fixed_cosmology',  # But use fixed mode
        seed=42,
        noise_level=1e-5,
        noise_mode='fixed'
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Load single halo catalog
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    fixed_cosmo = generator.load_fixed_halocat(single_halocat)
    
    print("\nNote: sigma_8 and omega_m will be fixed from the halocat")
    print("They will appear in the output but with constant values")
    
    # Process
    n_processed = generator.process_batch(
        n_samples=50,
        store_maps=False
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_hybrid/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    print("Output will have 4 parameters, but sigma_8 and omega_m are constant")
    
    return generator


if __name__ == "__main__":
    # Run different examples based on your needs
    
    # Choose which example to run:
    example_num = 1  # Change this to run different examples
    
    if example_num == 1:
        # Varying cosmology (original behavior)
        generator = example_varying_cosmology()
        
    elif example_num == 2:
        # Fixed cosmology (new feature)
        generator = example_fixed_cosmology()
        
    elif example_num == 3:
        # Fixed cosmology with noise as parameter
        generator = example_fixed_cosmology_with_noise()
        
    elif example_num == 4:
        # Hybrid mode for compatibility
        generator = example_hybrid_mode()
    
    # Optional: Visualize results if maps were stored
    if generator and len(generator.intensity_maps) > 0:
        print("\nVisualizing parameter effects...")
        if 'a_off' in generator.param_names:
            generator.visualize_parameter_dependence(
                param_to_vary='a_off',
                n_samples=min(5, len(generator.intensity_maps))
            )
```

```python
import numpy as np
from scipy.stats import qmc
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    
    Two operational modes:
    1. VARYING COSMOLOGY: Different halo catalogs for each parameter set (cosmo + astro params)
    2. FIXED COSMOLOGY: Single halo catalog with varying astrophysical parameters only
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, 
                 noise_level=1e1, noise_mode='fixed',
                 noise_range=(1e-5, 1e3),
                 log_noise_range=(0, 4),
                 mode='varying_cosmology'):
        """
        Initialize generator
        
        Args:
            mode: 'varying_cosmology' or 'fixed_cosmology'
            param_names: List of parameters to vary
                - For varying_cosmology: can include cosmo + astro params
                - For fixed_cosmology: should only include astro params (cosmo will be fixed from halocat)
        """
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        self.mode = mode
        
        # Noise configuration
        self.noise_mode = noise_mode
        self.noise_level = noise_level
        self.noise_range = noise_range
        self.log_noise_range = log_noise_range
        
        # Check if noise is in param_names
        self.noise_as_param = 'noise_level' in param_names
        self.log_noise_as_param = 'log_noise_level' in param_names
        
        if self.noise_as_param or self.log_noise_as_param:
            self.noise_mode = 'varying'
            param_names_filtered = [p for p in param_names if p not in ['noise_level', 'log_noise_level']]
        else:
            param_names_filtered = param_names
        
        # Parameter configuration
        self.param_names_internal = param_names_filtered
        self.param_names = param_names
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        self.astrophysical_params = [p for p in param_names_filtered if p not in self.cosmological_params]
        
        # Validate mode and parameters
        if mode == 'fixed_cosmology':
            cosmo_in_params = [p for p in param_names_filtered if p in self.cosmological_params]
            if cosmo_in_params:
                print(f"WARNING: In fixed_cosmology mode, cosmological parameters {cosmo_in_params} will be ignored")
                print("They will be taken from the fixed halo catalog instead")
        
        print(f"Mode: {mode}")
        print(f"Parameters to vary: {self.param_names}")
        print(f"  Cosmological: {[p for p in param_names_filtered if p in self.cosmological_params]}")
        print(f"  Astrophysical: {self.astrophysical_params}")
        print(f"Noise mode: {self.noise_mode}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.noise_values = None
        self.log_noise_values = None
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        self.fixed_halocat_file = None
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
        
        # Add noise range if treating as parameter
        if self.noise_as_param:
            self.param_ranges['noise_level'] = self.noise_range
        elif self.log_noise_as_param:
            self.param_ranges['log_noise_level'] = self.log_noise_range
    
    def generate_latin_hypercube_samples(self, n_samples):
        """Generate Latin hypercube samples for astrophysical parameters"""
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        samples_unit = sampler.random(n=n_samples)
        
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        for i, param in enumerate(self.astrophysical_params):
            print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
        
        return samples_scaled
    
    def generate_noise_levels(self, n_samples, method='uniform'):
        """Generate noise levels for varying noise mode"""
        if self.noise_mode == 'fixed':
            self.noise_values = np.full(n_samples, self.noise_level)
            self.log_noise_values = np.log10(self.noise_values)
            return self.noise_values
        
        rng = np.random.RandomState(self.seed + 1000)
        
        if self.log_noise_as_param:
            log_min, log_max = self.log_noise_range
            
            if method == 'uniform' or method == 'log_uniform':
                log_noise_levels = rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_noise_levels = qmc.scale(samples, log_min, log_max).flatten()
            
            self.log_noise_values = log_noise_levels
            self.noise_values = 10**log_noise_levels
        else:
            min_noise, max_noise = self.noise_range
            
            if method == 'uniform':
                noise_levels = rng.uniform(min_noise, max_noise, n_samples)
            elif method == 'log_uniform':
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**qmc.scale(samples, log_min, log_max).flatten()
            
            self.noise_values = noise_levels
            self.log_noise_values = np.log10(noise_levels)
        
        return self.noise_values
    
    def load_fixed_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        self.fixed_halocat_file = halocat_file
        data = np.load(halocat_file, allow_pickle=True)
        
        # Extract cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }
        
        self.fixed_halocat = data
        
        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")
        
        return self.fixed_cosmo_params
    
    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """
        Main processing method that handles both modes
        
        For varying_cosmology mode:
            - Requires halocat_files (list of paths)
            - Each halocat provides its own cosmology
            
        For fixed_cosmology mode:
            - Requires n_samples (number of realizations)
            - Must call load_fixed_halocat() first
        """
        if self.mode == 'fixed_cosmology':
            if self.fixed_halocat is None:
                raise ValueError("For fixed_cosmology mode, call load_fixed_halocat() first")
            return self._process_fixed_cosmology_batch(n_samples, store_maps)
        else:
            if halocat_files is None:
                raise ValueError("For varying_cosmology mode, provide halocat_files")
            return self._process_varying_cosmology_batch(halocat_files, store_maps)
    
    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with fixed cosmology, varying only astrophysics"""
        if n_samples is None:
            raise ValueError("Specify n_samples for fixed_cosmology mode")
        
        # Generate astrophysical parameters if needed
        if self.astrophysical_params and self.latin_hypercube_samples is None:
            self.generate_latin_hypercube_samples(n_samples)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_samples, method='latin_hypercube')
        
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"\nProcessing {n_samples} realizations with fixed cosmology")
        print(f"Fixed cosmology: {self.fixed_cosmo_params}")
        
        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Build parameters
                params_internal = self.fixed_cosmo_params.copy()
                
                # Add astrophysical parameters
                if self.latin_hypercube_samples is not None:
                    for j, param in enumerate(self.astrophysical_params):
                        params_internal[param] = self.latin_hypercube_samples[i, j]
                
                # Process with LIMpy
                intensity_2d, _, _ = self.process_single_halocat(
                    self.fixed_halocat_file, 
                    index=i, 
                    astro_params=self.latin_hypercube_samples[i] if self.latin_hypercube_samples is not None else None
                )
                
                # Add noise
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters (only the ones we're varying)
                param_values = []
                for param_name in self.param_names_internal:
                    if param_name in self.cosmological_params:
                        # For fixed cosmology mode, still include cosmo params in output if requested
                        param_values.append(params_internal[param_name])
                    else:
                        param_values.append(params_internal[param_name])
                
                # Add noise parameter if varying
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology using different halocats"""
        n_files = len(halocat_files)
        
        # Generate astrophysical parameters if needed
        if self.astrophysical_params and self.latin_hypercube_samples is None:
            self.generate_latin_hypercube_samples(n_files)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_files, method='latin_hypercube')
        
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"\nProcessing {n_files} halocats with varying cosmology")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy)
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters
                param_values = [params[p] for p in self.param_names_internal]
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        return len(all_params)
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process single halocat with parameter combination"""
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None):
        """Save results with metadata"""
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # File naming
        ps_file = output_path / f"ps_80_256_{seed}.npz"
        pdf_file = output_path / f"pdf_80_256_{seed}.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Metadata
        metadata = {
            'seed': seed,
            'mode': self.mode,
            'noise_mode': self.noise_mode,
        }
        
        if self.noise_mode == 'fixed':
            metadata['noise_level'] = self.noise_level
        else:
            if self.log_noise_as_param:
                metadata['log_noise_range'] = self.log_noise_range
                metadata['log_noise_values'] = self.log_noise_values
            else:
                metadata['noise_range'] = self.noise_range
            metadata['noise_values'] = self.noise_values
        
        if self.mode == 'fixed_cosmology' and self.fixed_cosmo_params:
            metadata['fixed_cosmology'] = self.fixed_cosmo_params
        
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            **metadata
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            **metadata
        )
        
        # Save parameter configuration
        config = {
            'mode': self.mode,
            'param_names': self.param_names,
            'param_ranges': self.param_ranges,
            'seed': seed,
            'noise_mode': self.noise_mode,
            'noise_level': self.noise_level if self.noise_mode == 'fixed' else None,
            'noise_range': self.noise_range if self.noise_as_param else None,
            'log_noise_range': self.log_noise_range if self.log_noise_as_param else None,
            'boxsize': self.boxsize,
            'ngrid': self.ngrid,
            'redshift': self.redshift,
            'line_name': self.line_name,
            'model_name': self.model_name
        }
        
        if self.mode == 'fixed_cosmology' and self.fixed_cosmo_params:
            config['fixed_cosmology'] = self.fixed_cosmo_params
        
        config_file = output_path / 'parameter_config.json'
        with open(config_file, 'w') as f:
            json.dump(config, f, indent=2)
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Config: {config_file}")
        print(f"\nMode: {self.mode}")
        print(f"Seed: {seed}")
        print(f"Parameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            values = self.params_array[:, i]
            print(f"  {name}: [{values.min():.3f}, {values.max():.3f}]")


# ============================================================================
# EXAMPLE USAGE
# ============================================================================

def example_varying_cosmology(n_maps=100, input_dir=None, output_dir=None, 
                              store_maps=False, seed=42, noise_level=1e-5):
    """
    EXAMPLE 1: Varying Cosmology Mode
    Uses different halo catalogs, each with its own cosmology
    
    Args:
        n_maps: Number of maps to create (will use this many halo catalogs)
        input_dir: Directory containing halo catalogs (default: /Users/anirbanroy/Desktop/halocats/)
        output_dir: Directory to save results (default: /Users/anirbanroy/Desktop/processed_varying_cosmo/)
        store_maps: Whether to store intensity maps in memory for visualization
        seed: Random seed for reproducibility
        noise_level: Fixed noise level to add to maps
    """
    print("="*80)
    print("EXAMPLE 1: VARYING COSMOLOGY MODE")
    print("Uses different halo catalogs for each parameter combination")
    print("="*80)
    
    # Set default directories
    if input_dir is None:
        input_dir = Path("/Users/anirbanroy/Desktop/halocats/")
    else:
        input_dir = Path(input_dir)
    
    if output_dir is None:
        output_dir = f"/Users/anirbanroy/Desktop/processed_varying_cosmo_{n_maps}maps/"
    
    # Initialize generator with cosmo + astro parameters
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],
        mode='varying_cosmology',  # Explicitly set mode
        seed=seed,
        noise_level=noise_level,
        noise_mode='fixed'
    )
    
    # Setup parameter ranges for astrophysical parameters
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Get halo catalogs
    all_halocat_files = sorted(input_dir.glob("halocat_*.npz"))
    
    # Check if we have enough files
    if len(all_halocat_files) < n_maps:
        print(f"WARNING: Only {len(all_halocat_files)} halo catalogs available, but {n_maps} requested")
        n_maps = len(all_halocat_files)
    
    halocat_files = all_halocat_files[:n_maps]  # Use requested number of files
    
    print(f"\nProcessing {len(halocat_files)} halo catalogs")
    print("Each catalog provides its own cosmology (sigma_8, omega_m)")
    print("Astrophysical parameters (a_off, b_off) are sampled via Latin hypercube")
    print(f"Store maps in memory: {store_maps}")
    print(f"Output directory: {output_dir}")
    
    # Process
    n_processed = generator.process_batch(
        halocat_files=halocat_files,
        store_maps=store_maps  # Store for visualization if requested
    )
    
    # Save results
    generator.save_results(output_dir, seed=seed)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_fixed_cosmology(n_maps=100, input_dir=None, output_dir=None, 
                              store_maps=False, seed=42, noise_level=1e-5):
    """
    EXAMPLE 2: Fixed Cosmology Mode
    Uses ONE halo catalog repeatedly with different astrophysical parameters
    """
    print("="*80)
    print("EXAMPLE 2: FIXED COSMOLOGY MODE")
    print("Uses single halo catalog with varying astrophysical parameters only")
    print("="*80)
    
    # Initialize generator for fixed cosmology mode
    generator = LIMpyGenerator(
        param_names=['a_off', 'b_off'],  # Only astrophysical parameters
        mode='fixed_cosmology',  # Explicitly set mode
        seed=42,
        noise_level=1e-5,
        noise_mode='fixed'
    )
    
    # Setup parameter ranges
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Load the single halo catalog to use
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    fixed_cosmo = generator.load_fixed_halocat(single_halocat)
    
    print("\nGenerating 100 realizations with fixed cosmology")
    print("Only astrophysical parameters (a_off, b_off) will vary")
    
    # Process with fixed cosmology
    n_processed = generator.process_batch(
        n_samples=n_maps,  # Number of realizations
        store_maps=True
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_fixed_cosmo/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_fixed_cosmology_with_noise():
    """
    EXAMPLE 3: Fixed Cosmology with Noise as Parameter
    Uses ONE halo catalog with varying astrophysical parameters AND noise
    """
    print("="*80)
    print("EXAMPLE 3: FIXED COSMOLOGY WITH NOISE AS PARAMETER")
    print("Uses single halo catalog with varying astro params + noise level")
    print("="*80)
    
    # Initialize with noise as a parameter
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off', 'log_noise_level'],  # Include noise
        mode='fixed_cosmology',
        seed=42,
        log_noise_range=(0, 4),  # log10(noise) from 10^0 to 10^4
    )
    
    # Setup parameter ranges
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Load the single halo catalog
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    generator.load_fixed_halocat(single_halocat)
    
    print("\nGenerating 100 realizations with:")
    print("  - Fixed cosmology from halocat")
    print("  - Varying a_off and b_off")
    print("  - Varying log_noise_level from 0 to 4")
    
    # Process
    n_processed = generator.process_batch(
        n_samples=3000,
        store_maps=True
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_fixed_cosmo_with_noise/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_hybrid_mode():
    """
    EXAMPLE 4: Hybrid - Include cosmo params in output even with fixed cosmology
    This is useful when you want the output to have cosmo params for compatibility
    """
    print("="*80)
    print("EXAMPLE 4: FIXED COSMOLOGY BUT INCLUDE COSMO PARAMS IN OUTPUT")
    print("Useful for maintaining consistent output format")
    print("="*80)
    
    # Initialize with all parameters, but use fixed_cosmology mode
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],  # Include cosmo params
        mode='fixed_cosmology',  # But use fixed mode
        seed=42,
        noise_level=1e-5,
        noise_mode='fixed'
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Load single halo catalog
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    fixed_cosmo = generator.load_fixed_halocat(single_halocat)
    
    print("\nNote: sigma_8 and omega_m will be fixed from the halocat")
    print("They will appear in the output but with constant values")
    
    # Process
    n_processed = generator.process_batch(
        n_samples=50,
        store_maps=False
    )
    
    # Save results
    output_dir = "/Users/anirbanroy/Desktop/processed_hybrid/"
    generator.save_results(output_dir, seed=42)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    print("Output will have 4 parameters, but sigma_8 and omega_m are constant")
    
    return generator


def example_varying_all_parameters(n_maps=3000, input_dir=None, output_dir=None, 
                                  store_maps=False, seed=42):
    """
    EXAMPLE 5: Varying Cosmology + Astrophysics + Noise
    Uses different halo catalogs (varying cosmology) with varying astro params and noise
    
    Args:
        n_maps: Number of maps to create
        input_dir: Directory containing halo catalogs
        output_dir: Directory to save results
        store_maps: Whether to store intensity maps in memory
        seed: Random seed for reproducibility
    """
    print("="*80)
    print("EXAMPLE 5: VARYING ALL PARAMETERS (COSMO + ASTRO + NOISE)")
    print("Uses different halo catalogs with varying everything")
    print("="*80)
    
    # Set default directories
    if input_dir is None:
        input_dir = Path("/Users/anirbanroy/Desktop/halocats/")
    else:
        input_dir = Path(input_dir)
    
    if output_dir is None:
        output_dir = f"/Users/anirbanroy/Desktop/processed_all_params_noise_{n_maps}maps/"
    
    # Initialize generator with ALL parameters including log_noise_level
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off', 'log_noise_level'],
        mode='varying_cosmology',  # Use varying cosmology mode
        seed=seed,
        log_noise_range=(-5, 3),  # log10(noise) from 10^-5 to 10^3 (wide range)
    )
    
    # Setup parameter ranges for astrophysical parameters
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    # Get halo catalogs
    all_halocat_files = sorted(input_dir.glob("halocat_*.npz"))
    
    if len(all_halocat_files) < n_maps:
        print(f"WARNING: Only {len(all_halocat_files)} halo catalogs available, but {n_maps} requested")
        n_maps = len(all_halocat_files)
    
    halocat_files = all_halocat_files[:n_maps]
    
    print(f"\nProcessing {len(halocat_files)} halo catalogs")
    print("Parameters varying:")
    print("  - sigma_8, omega_m: from halo catalogs")
    print("  - a_off: [4, 10]")
    print("  - b_off: [0, 2]")
    print(f"  - log_noise_level: [{generator.log_noise_range[0]}, {generator.log_noise_range[1]}]")
    print(f"    (noise level: [10^{generator.log_noise_range[0]:.1f}, 10^{generator.log_noise_range[1]:.1f}])")
    print(f"Store maps in memory: {store_maps}")
    print(f"Output directory: {output_dir}")
    
    # Process
    n_processed = generator.process_batch(
        halocat_files=halocat_files,
        store_maps=store_maps
    )
    
    # Save results
    generator.save_results(output_dir, seed=seed)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    print("\nParameter summary:")
    for i, name in enumerate(generator.param_names):
        values = generator.params_array[:, i]
        print(f"  {name}: [{values.min():.3f}, {values.max():.3f}] (mean={values.mean():.3f}, std={values.std():.3f})")
    
    # Show noise distribution
    if generator.log_noise_values is not None:
        print(f"\nNoise level distribution:")
        print(f"  Min noise: {10**generator.log_noise_values.min():.2e}")
        print(f"  Max noise: {10**generator.log_noise_values.max():.2e}")
        print(f"  Median noise: {10**np.median(generator.log_noise_values):.2e}")
    
    return generator


if __name__ == "__main__":
    # ========================================================================
    # QUICK EXAMPLES - Uncomment the one you want to run
    # ========================================================================
    
    # Example 1: Varying cosmology with 10 maps
    # generator = example_varying_cosmology(n_maps=10, store_maps=True)
    
    # Example 2: Varying cosmology with 3000 maps (full dataset)
    # generator = example_varying_cosmology(n_maps=3000, store_maps=False)
    
    # Example 3: Varying cosmology with custom settings
    
    
    generator = example_varying_all_parameters(
        n_maps=3000,
        input_dir="/Users/anirbanroy/Desktop/halocats/",
        output_dir="/Users/anirbanroy/Desktop/test_maps_all_params_varying_noise/",
        store_maps=True,  # Don't store 3000 maps in memory
        seed=42
    )

    
    
    # Example 4: Fixed cosmology mode
    # generator = example_fixed_cosmology()
    
    # Example 5: Fixed cosmology with noise as parameter
    # generator = example_fixed_cosmology_with_noise()
    
    # Example 6: Hybrid mode for compatibility
    # generator = example_hybrid_mode()
    
    # Optional: Visualize results if maps were stored
    if generator and len(generator.intensity_maps) > 0:
        print("\nVisualizing parameter effects...")
        if 'a_off' in generator.param_names:
            # Visualize how a_off affects the maps
            generator.visualize_parameter_dependence(
                param_to_vary='a_off',
                n_samples=min(5, len(generator.intensity_maps))
            )
```

```python
import numpy as np
from scipy.stats import qmc
import limpy.lines as ll
import limpy.powerspectra as lp
from pathlib import Path
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec

class LIMpyGenerator:
    """
    Generate Line Intensity Maps with flexible parameter combinations.
    
    Two operational modes:
    1. VARYING COSMOLOGY: Different halo catalogs for each parameter set (cosmo + astro params)
    2. FIXED COSMOLOGY: Single halo catalog with varying astrophysical parameters only
    """
    
    def __init__(self, boxsize=80.0, ngrid=256, redshift=3.6, 
                 line_name='CII158', model_name='Alma_scaling',
                 param_names=['sigma_8', 'omega_m'], seed=42, 
                 noise_level=1e1, noise_mode='fixed',
                 noise_range=(1e-5, 1e3),
                 log_noise_range=(0, 4),
                 mode='varying_cosmology'):
        """
        Initialize generator
        
        Args:
            mode: 'varying_cosmology' or 'fixed_cosmology'
            param_names: List of parameters to vary
                - For varying_cosmology: can include cosmo + astro params
                - For fixed_cosmology: should only include astro params (cosmo will be fixed from halocat)
        """
        self.boxsize = boxsize
        self.ngrid = ngrid
        self.redshift = redshift
        self.line_name = line_name
        self.model_name = model_name
        self.mmin = 1e9
        self.seed = seed
        self.mode = mode
        
        # Noise configuration
        self.noise_mode = noise_mode
        self.noise_level = noise_level
        self.noise_range = noise_range
        self.log_noise_range = log_noise_range
        
        # Check if noise is in param_names
        self.noise_as_param = 'noise_level' in param_names
        self.log_noise_as_param = 'log_noise_level' in param_names
        
        if self.noise_as_param or self.log_noise_as_param:
            self.noise_mode = 'varying'
            param_names_filtered = [p for p in param_names if p not in ['noise_level', 'log_noise_level']]
        else:
            param_names_filtered = param_names
        
        # Parameter configuration
        self.param_names_internal = param_names_filtered
        self.param_names = param_names
        self.cosmological_params = ['sigma_8', 'omega_m', 'h']
        self.astrophysical_params = [p for p in param_names_filtered if p not in self.cosmological_params]
        
        # Validate mode and parameters
        if mode == 'fixed_cosmology':
            cosmo_in_params = [p for p in param_names_filtered if p in self.cosmological_params]
            if cosmo_in_params:
                print(f"WARNING: In fixed_cosmology mode, cosmological parameters {cosmo_in_params} will be ignored")
                print("They will be taken from the fixed halo catalog instead")
        
        print(f"Mode: {mode}")
        print(f"Parameters to vary: {self.param_names}")
        print(f"  Cosmological: {[p for p in param_names_filtered if p in self.cosmological_params]}")
        print(f"  Astrophysical: {self.astrophysical_params}")
        print(f"Noise mode: {self.noise_mode}")
        
        # Storage
        self.params_array = None
        self.param_ranges = {}
        self.latin_hypercube_samples = None
        self.noise_values = None
        self.log_noise_values = None
        self.intensity_maps = []
        self.power_spectra = []
        self.pdfs = []
        self.k_values = None
        self.pdf_bins = None
        
        # For fixed cosmology mode
        self.fixed_halocat = None
        self.fixed_cosmo_params = None
        self.fixed_halocat_file = None
    
    def setup_parameter_ranges(self, param_ranges=None):
        """Setup parameter ranges for astrophysical parameters"""
        default_ranges = {
            'a_off': (4, 10.0),
            'b_off': (0.0, 2.0)
        }
        
        if param_ranges:
            self.param_ranges.update(param_ranges)
        else:
            for param in self.astrophysical_params:
                if param in default_ranges:
                    self.param_ranges[param] = default_ranges[param]
        
        # Add noise range if treating as parameter
        if self.noise_as_param:
            self.param_ranges['noise_level'] = self.noise_range
        elif self.log_noise_as_param:
            self.param_ranges['log_noise_level'] = self.log_noise_range
    
    def generate_latin_hypercube_samples(self, n_samples):
        """Generate Latin hypercube samples for astrophysical parameters"""
        n_astro_params = len(self.astrophysical_params)
        
        if n_astro_params == 0:
            print("No astrophysical parameters to sample")
            self.latin_hypercube_samples = None
            return None
        
        sampler = qmc.LatinHypercube(d=n_astro_params, seed=self.seed)
        samples_unit = sampler.random(n=n_samples)
        
        samples_scaled = np.zeros((n_samples, n_astro_params))
        for i, param in enumerate(self.astrophysical_params):
            low, high = self.param_ranges[param]
            samples_scaled[:, i] = qmc.scale(samples_unit[:, i:i+1], low, high).flatten()
        
        self.latin_hypercube_samples = samples_scaled
        
        print(f"Generated {n_samples} Latin hypercube samples for {self.astrophysical_params}")
        for i, param in enumerate(self.astrophysical_params):
            print(f"  {param}: [{samples_scaled[:, i].min():.3f}, {samples_scaled[:, i].max():.3f}]")
        
        return samples_scaled
    
    def generate_noise_levels(self, n_samples, method='uniform'):
        """Generate noise levels for varying noise mode"""
        if self.noise_mode == 'fixed':
            self.noise_values = np.full(n_samples, self.noise_level)
            self.log_noise_values = np.log10(self.noise_values)
            return self.noise_values
        
        rng = np.random.RandomState(self.seed + 1000)
        
        if self.log_noise_as_param:
            log_min, log_max = self.log_noise_range
            
            if method == 'uniform' or method == 'log_uniform':
                log_noise_levels = rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_noise_levels = qmc.scale(samples, log_min, log_max).flatten()
            
            self.log_noise_values = log_noise_levels
            self.noise_values = 10**log_noise_levels
        else:
            min_noise, max_noise = self.noise_range
            
            if method == 'uniform':
                noise_levels = rng.uniform(min_noise, max_noise, n_samples)
            elif method == 'log_uniform':
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**rng.uniform(log_min, log_max, n_samples)
            elif method == 'latin_hypercube':
                sampler = qmc.LatinHypercube(d=1, seed=self.seed + 1000)
                samples = sampler.random(n=n_samples)
                log_min = np.log10(min_noise)
                log_max = np.log10(max_noise)
                noise_levels = 10**qmc.scale(samples, log_min, log_max).flatten()
            
            self.noise_values = noise_levels
            self.log_noise_values = np.log10(noise_levels)
        
        return self.noise_values
    
    def load_fixed_halocat(self, halocat_file):
        """Load a single halo catalog for fixed cosmology mode"""
        self.fixed_halocat_file = halocat_file
        data = np.load(halocat_file, allow_pickle=True)
        
        # Extract cosmological parameters
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        self.fixed_cosmo_params = {
            'sigma_8': params.get('sigma_8', params.get('sigma8', 0.8)),
            'omega_m': params.get('omega_m', params.get('Omega_m', 0.3)),
            'h': params.get('h', 0.7)
        }
        
        self.fixed_halocat = data
        
        print(f"Fixed cosmology loaded from {Path(halocat_file).name}:")
        for key, val in self.fixed_cosmo_params.items():
            print(f"  {key} = {val:.3f}")
        
        return self.fixed_cosmo_params
    
    def process_batch(self, halocat_files=None, n_samples=None, store_maps=False):
        """
        Main processing method that handles both modes
        
        For varying_cosmology mode:
            - Requires halocat_files (list of paths)
            - Each halocat provides its own cosmology
            
        For fixed_cosmology mode:
            - Requires n_samples (number of realizations)
            - Must call load_fixed_halocat() first
        """
        if self.mode == 'fixed_cosmology':
            if self.fixed_halocat is None:
                raise ValueError("For fixed_cosmology mode, call load_fixed_halocat() first")
            return self._process_fixed_cosmology_batch(n_samples, store_maps)
        else:
            if halocat_files is None:
                raise ValueError("For varying_cosmology mode, provide halocat_files")
            return self._process_varying_cosmology_batch(halocat_files, store_maps)
    
    def _process_fixed_cosmology_batch(self, n_samples, store_maps):
        """Process with fixed cosmology, varying only astrophysics"""
        if n_samples is None:
            raise ValueError("Specify n_samples for fixed_cosmology mode")
        
        # Generate astrophysical parameters if needed
        if self.astrophysical_params and self.latin_hypercube_samples is None:
            self.generate_latin_hypercube_samples(n_samples)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_samples, method='latin_hypercube')
        
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"\nProcessing {n_samples} realizations with fixed cosmology")
        print(f"Fixed cosmology: {self.fixed_cosmo_params}")
        print(f"Store maps: {store_maps}")
        
        for i in tqdm(range(n_samples), desc="Processing"):
            try:
                # Build parameters
                params_internal = self.fixed_cosmo_params.copy()
                
                # Add astrophysical parameters
                if self.latin_hypercube_samples is not None:
                    for j, param in enumerate(self.astrophysical_params):
                        params_internal[param] = self.latin_hypercube_samples[i, j]
                
                # Process with LIMpy
                intensity_2d, _, _ = self.process_single_halocat(
                    self.fixed_halocat_file, 
                    index=i, 
                    astro_params=self.latin_hypercube_samples[i] if self.latin_hypercube_samples is not None else None
                )
                
                # Add noise
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                # FIXED: Store maps if requested
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy.copy())
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters (only the ones we're varying)
                param_values = []
                for param_name in self.param_names_internal:
                    if param_name in self.cosmological_params:
                        # For fixed cosmology mode, still include cosmo params in output if requested
                        param_values.append(params_internal[param_name])
                    else:
                        param_values.append(params_internal[param_name])
                
                # Add noise parameter if varying
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                import traceback
                traceback.print_exc()
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        print(f"\nStored {len(self.intensity_maps)} maps in memory")
        
        return len(all_params)
    
    def _process_varying_cosmology_batch(self, halocat_files, store_maps):
        """Process with varying cosmology using different halocats"""
        n_files = len(halocat_files)
        
        # Generate astrophysical parameters if needed
        if self.astrophysical_params and self.latin_hypercube_samples is None:
            self.generate_latin_hypercube_samples(n_files)
        
        # Generate noise levels
        noise_levels = self.generate_noise_levels(n_files, method='latin_hypercube')
        
        rng = np.random.RandomState(self.seed)
        
        all_params = []
        all_pk = []
        all_pdf = []
        
        print(f"\nProcessing {n_files} halocats with varying cosmology")
        print(f"Store maps: {store_maps}")
        
        for i, halocat_file in enumerate(tqdm(halocat_files, desc="Processing")):
            try:
                astro_params = None
                if self.latin_hypercube_samples is not None:
                    astro_params = self.latin_hypercube_samples[i]
                
                intensity_2d, _, params = self.process_single_halocat(
                    halocat_file, index=i, astro_params=astro_params
                )
                
                # Add noise
                current_noise_level = noise_levels[i]
                noise = rng.normal(0, current_noise_level, intensity_2d.shape)
                intensity_2d_noisy = intensity_2d + noise
                
                # FIXED: Store maps if requested
                if store_maps:
                    self.intensity_maps.append(intensity_2d_noisy.copy())
                
                # Compute observables
                k, pk = self.compute_power_spectrum_2d(intensity_2d_noisy)
                bins, pdf = self.compute_pdf(intensity_2d_noisy, n_bins=40)
                
                # Store parameters
                param_values = [params[p] for p in self.param_names_internal]
                if self.log_noise_as_param:
                    param_values.append(self.log_noise_values[i])
                elif self.noise_as_param:
                    param_values.append(current_noise_level)
                
                all_params.append(param_values)
                all_pk.append(pk)
                all_pdf.append(pdf)
                
                if self.k_values is None:
                    self.k_values = k
                if self.pdf_bins is None:
                    self.pdf_bins = bins
                    
            except Exception as e:
                print(f"Error processing {halocat_file.name}: {e}")
                import traceback
                traceback.print_exc()
                continue
        
        self.params_array = np.array(all_params)
        self.power_spectra = all_pk
        self.pdfs = all_pdf
        
        print(f"\nStored {len(self.intensity_maps)} maps in memory")
        
        return len(all_params)
    
    def process_single_halocat(self, halocat_file, index=0, astro_params=None):
        """Process single halocat with parameter combination"""
        data = np.load(halocat_file, allow_pickle=True)
        
        if 'params' in data:
            params = data['params']
            if hasattr(params, 'item'):
                params = params.item()
        else:
            params = {}
        
        # Build parameter dictionary
        params_internal = {}
        
        # Add cosmological parameters
        for cosmo_param in ['sigma_8', 'omega_m', 'h']:
            if cosmo_param == 'sigma_8':
                params_internal['sigma_8'] = params.get('sigma_8', params.get('sigma8', 0.8))
            elif cosmo_param == 'omega_m':
                params_internal['omega_m'] = params.get('omega_m', params.get('Omega_m', 0.3))
            elif cosmo_param == 'h':
                params_internal['h'] = params.get('h', 0.7)
        
        # Add astrophysical parameters
        if astro_params is not None:
            for i, param in enumerate(self.astrophysical_params):
                params_internal[param] = astro_params[i]
        
        # Fix boundary conditions
        halocat_file_to_use = self._fix_boundary_conditions(data, halocat_file)
        
        try:
            lim_sim = ll.lim_sims(
                halocat_file_to_use,
                self.redshift,
                model_name=self.model_name,
                line_name=self.line_name,
                halo_cutoff_mass=self.mmin,
                halocat_type="input_cat",
                parameters=params_internal,
                ngrid_x=self.ngrid,
                ngrid_y=self.ngrid,
                ngrid_z=self.ngrid,
                boxsize_x=self.boxsize,
                boxsize_y=self.boxsize,
                boxsize_z=self.boxsize,
                nu_obs=220,
                theta_fwhm=1,
                dnu_obs=2.2
            )
            
            intensity_3d = lim_sim.make_intensity_grid()
            intensity_2d = np.mean(intensity_3d, axis=2)
            
        finally:
            if halocat_file_to_use != str(halocat_file):
                import os
                try:
                    os.unlink(halocat_file_to_use)
                except:
                    pass
        
        return intensity_2d, intensity_3d, params_internal
    
    def _fix_boundary_conditions(self, data, halocat_file):
        """Fix boundary conditions in halo positions"""
        import tempfile
        
        if 'x' in data or 'X' in data or 'pos' in data:
            temp_file = tempfile.NamedTemporaryFile(suffix='.npz', delete=False)
            save_dict = {key: data[key] for key in data.keys()}
            
            if 'pos' in data:
                positions = data['pos'].copy()
                positions = positions % self.boxsize
                positions[positions >= self.boxsize] = self.boxsize - 1e-6
                save_dict['pos'] = positions
            else:
                for coord in ['x', 'X', 'y', 'Y', 'z', 'Z']:
                    if coord in data:
                        pos = data[coord].copy()
                        pos = pos % self.boxsize
                        pos[pos >= self.boxsize] = self.boxsize - 1e-6
                        save_dict[coord] = pos
            
            np.savez(temp_file.name, **save_dict)
            return temp_file.name
        return str(halocat_file)
    
    def compute_power_spectrum_2d(self, intensity_2d):
        """Compute 2D power spectrum"""
        k, pk = lp.get_pk2d(intensity_2d, self.boxsize, self.boxsize, 
                           self.ngrid, self.ngrid)
        return k, pk
    
    def compute_pdf(self, intensity_map, n_bins=40):
        """Compute PDF"""
        flat_intensity = intensity_map.flatten()
        bin_edges = np.linspace(flat_intensity.min(), flat_intensity.max(), n_bins + 1)
        pdf, _ = np.histogram(flat_intensity, bins=bin_edges, density=True)
        return bin_edges, pdf
    
    def save_results(self, output_dir, seed=None, save_intensity_maps=False):
        """
        Save results with metadata
        
        Args:
            output_dir: Directory to save results
            seed: Random seed for file naming
            save_intensity_maps: If True, save intensity maps to disk (can be large!)
        """
        if seed is None:
            seed = self.seed
        
        output_path = Path(output_dir)
        output_path.mkdir(parents=True, exist_ok=True)
        
        # Create structured array
        dtype_list = [(name, '<f8') for name in self.param_names]
        params_structured = np.zeros(len(self.params_array), dtype=dtype_list)
        
        for i, param_vals in enumerate(self.params_array):
            params_structured[i] = tuple(param_vals)
        
        # File naming
        ps_file = output_path / f"ps_80_256.npz"
        pdf_file = output_path / f"pdf_80_256.npz"
        
        all_k = np.array([self.k_values] * len(self.power_spectra))
        
        # Metadata
        metadata = {
            'seed': seed,
            'mode': self.mode,
            'noise_mode': self.noise_mode,
        }
        
        if self.noise_mode == 'fixed':
            metadata['noise_level'] = self.noise_level
        else:
            if self.log_noise_as_param:
                metadata['log_noise_range'] = self.log_noise_range
                metadata['log_noise_values'] = self.log_noise_values
            else:
                metadata['noise_range'] = self.noise_range
            metadata['noise_values'] = self.noise_values
        
        if self.mode == 'fixed_cosmology' and self.fixed_cosmo_params:
            metadata['fixed_cosmology'] = self.fixed_cosmo_params
        
        np.savez_compressed(
            ps_file,
            parameters=params_structured,
            k_2d=all_k,
            pk_2d=np.array(self.power_spectra),
            **metadata
        )
        
        np.savez_compressed(
            pdf_file,
            parameters=params_structured,
            lim_bin_edges_linear_2d=self.pdf_bins,
            lim_pdf_linear_2d=np.array(self.pdfs),
            **metadata
        )
        
        # Save intensity maps if requested and available
        if save_intensity_maps and len(self.intensity_maps) > 0:
            maps_file = output_path / f"intensity_maps_80_256_{seed}.npz"
            print(f"\nSaving {len(self.intensity_maps)} intensity maps to disk...")
            np.savez_compressed(
                maps_file,
                parameters=params_structured,
                intensity_maps=np.array(self.intensity_maps),
                **metadata
            )
            print(f"  Intensity maps: {maps_file}")
            print(f"  File size: ~{maps_file.stat().st_size / 1e9:.2f} GB")
        
        # Save parameter configuration
        config = {
            'mode': self.mode,
            'param_names': self.param_names,
            'param_ranges': self.param_ranges,
            'seed': seed,
            'noise_mode': self.noise_mode,
            'noise_level': self.noise_level if self.noise_mode == 'fixed' else None,
            'noise_range': self.noise_range if self.noise_as_param else None,
            'log_noise_range': self.log_noise_range if self.log_noise_as_param else None,
            'boxsize': self.boxsize,
            'ngrid': self.ngrid,
            'redshift': self.redshift,
            'line_name': self.line_name,
            'model_name': self.model_name,
            'n_maps_stored': len(self.intensity_maps)
        }
        
        if self.mode == 'fixed_cosmology' and self.fixed_cosmo_params:
            config['fixed_cosmology'] = self.fixed_cosmo_params
        
        config_file = output_path / 'parameter_config.json'
        with open(config_file, 'w') as f:
            json.dump(config, f, indent=2)
        
        print(f"\nResults saved:")
        print(f"  Power spectra: {ps_file}")
        print(f"  PDFs: {pdf_file}")
        print(f"  Config: {config_file}")
        print(f"\nMode: {self.mode}")
        print(f"Seed: {seed}")
        print(f"Maps stored in memory: {len(self.intensity_maps)}")
        print(f"\nParameter summary ({len(self.params_array)} samples):")
        for i, name in enumerate(self.param_names):
            values = self.params_array[:, i]
            print(f"  {name}: [{values.min():.3f}, {values.max():.3f}]")


# ============================================================================
# EXAMPLE USAGE
# ============================================================================

def example_varying_cosmology(n_maps=100, input_dir=None, output_dir=None, 
                              store_maps=False, seed=42, noise_level=1e-5):
    """
    EXAMPLE 1: Varying Cosmology Mode
    Uses different halo catalogs, each with its own cosmology
    """
    print("="*80)
    print("EXAMPLE 1: VARYING COSMOLOGY MODE")
    print("Uses different halo catalogs for each parameter combination")
    print("="*80)
    
    if input_dir is None:
        input_dir = Path("/Users/anirbanroy/Desktop/halocats/")
    else:
        input_dir = Path(input_dir)
    
    if output_dir is None:
        output_dir = f"/Users/anirbanroy/Desktop/processed_varying_cosmo_{n_maps}maps/"
    
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],
        mode='varying_cosmology',
        seed=seed,
        noise_level=noise_level,
        noise_mode='fixed'
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    all_halocat_files = sorted(input_dir.glob("halocat_*.npz"))
    
    if len(all_halocat_files) < n_maps:
        print(f"WARNING: Only {len(all_halocat_files)} halo catalogs available, but {n_maps} requested")
        n_maps = len(all_halocat_files)
    
    halocat_files = all_halocat_files[:n_maps]
    
    print(f"\nProcessing {len(halocat_files)} halo catalogs")
    print("Each catalog provides its own cosmology (sigma_8, omega_m)")
    print("Astrophysical parameters (a_off, b_off) are sampled via Latin hypercube")
    print(f"Store maps in memory: {store_maps}")
    print(f"Output directory: {output_dir}")
    
    n_processed = generator.process_batch(
        halocat_files=halocat_files,
        store_maps=store_maps
    )
    
    generator.save_results(output_dir, seed=seed, save_intensity_maps=store_maps)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_fixed_cosmology(n_maps=100, input_dir=None, output_dir=None, 
                              store_maps=False, seed=42, noise_level=1e-5):
    """
    EXAMPLE 2: Fixed Cosmology Mode
    Uses ONE halo catalog repeatedly with different astrophysical parameters
    """
    print("="*80)
    print("EXAMPLE 2: FIXED COSMOLOGY MODE")
    print("Uses single halo catalog with varying astrophysical parameters only")
    print("="*80)
    
    generator = LIMpyGenerator(
        param_names=['a_off', 'b_off'],
        mode='fixed_cosmology',
        seed=seed,
        noise_level=noise_level,
        noise_mode='fixed'
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    fixed_cosmo = generator.load_fixed_halocat(single_halocat)
    
    print(f"\nGenerating {n_maps} realizations with fixed cosmology")
    print("Only astrophysical parameters (a_off, b_off) will vary")
    
    n_processed = generator.process_batch(
        n_samples=n_maps,
        store_maps=store_maps
    )
    
    if output_dir is None:
        output_dir = "/Users/anirbanroy/Desktop/processed_fixed_cosmo/"
    generator.save_results(output_dir, seed=seed, save_intensity_maps=store_maps)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_fixed_cosmology_with_noise(n_maps=100, seed=42):
    """
    EXAMPLE 3: Fixed Cosmology with Noise as Parameter
    Uses ONE halo catalog with varying astrophysical parameters AND noise
    """
    print("="*80)
    print("EXAMPLE 3: FIXED COSMOLOGY WITH NOISE AS PARAMETER")
    print("Uses single halo catalog with varying astro params + noise level")
    print("="*80)
    
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off', 'log_noise_level'],
        mode='fixed_cosmology',
        seed=seed,
        log_noise_range=(0, 4),
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    generator.load_fixed_halocat(single_halocat)
    
    print(f"\nGenerating {n_maps} realizations with:")
    print("  - Fixed cosmology from halocat")
    print("  - Varying a_off and b_off")
    print("  - Varying log_noise_level from 0 to 4")
    
    n_processed = generator.process_batch(
        n_samples=n_maps,
        store_maps=True
    )
    
    output_dir = "/Users/anirbanroy/Desktop/processed_fixed_cosmo_with_noise/"
    generator.save_results(output_dir, seed=seed, save_intensity_maps=store_maps)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    
    return generator


def example_hybrid_mode(n_maps=50, seed=42):
    """
    EXAMPLE 4: Hybrid - Include cosmo params in output even with fixed cosmology
    This is useful when you want the output to have cosmo params for compatibility
    """
    print("="*80)
    print("EXAMPLE 4: FIXED COSMOLOGY BUT INCLUDE COSMO PARAMS IN OUTPUT")
    print("Useful for maintaining consistent output format")
    print("="*80)
    
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off'],
        mode='fixed_cosmology',
        seed=seed,
        noise_level=1e-5,
        noise_mode='fixed'
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    single_halocat = "/Users/anirbanroy/Desktop/halocats/halocat_1396.npz"
    fixed_cosmo = generator.load_fixed_halocat(single_halocat)
    
    print("\nNote: sigma_8 and omega_m will be fixed from the halocat")
    print("They will appear in the output but with constant values")
    
    n_processed = generator.process_batch(
        n_samples=n_maps,
        store_maps=False
    )
    
    output_dir = "/Users/anirbanroy/Desktop/processed_hybrid/"
    generator.save_results(output_dir, seed=seed, save_intensity_maps=store_maps)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    print("Output will have 4 parameters, but sigma_8 and omega_m are constant")
    
    return generator


def example_varying_all_parameters(n_maps=3000, input_dir=None, output_dir=None, 
                                  store_maps=False, seed=42):
    """
    EXAMPLE 5: Varying Cosmology + Astrophysics + Noise
    Uses different halo catalogs (varying cosmology) with varying astro params and noise
    """
    print("="*80)
    print("EXAMPLE 5: VARYING ALL PARAMETERS (COSMO + ASTRO + NOISE)")
    print("Uses different halo catalogs with varying everything")
    print("="*80)
    
    if input_dir is None:
        input_dir = Path("/Users/anirbanroy/Desktop/halocats/")
    else:
        input_dir = Path(input_dir)
    
    if output_dir is None:
        output_dir = f"/Users/anirbanroy/Desktop/processed_all_params_noise_{n_maps}maps/"
    
    generator = LIMpyGenerator(
        param_names=['sigma_8', 'omega_m', 'a_off', 'b_off', 'log_noise_level'],
        mode='varying_cosmology',
        seed=seed,
        log_noise_range=(-5, 3),
    )
    
    generator.setup_parameter_ranges({
        'a_off': (4, 10),
        'b_off': (0, 2)
    })
    
    all_halocat_files = sorted(input_dir.glob("halocat_*.npz"))
    
    if len(all_halocat_files) < n_maps:
        print(f"WARNING: Only {len(all_halocat_files)} halo catalogs available, but {n_maps} requested")
        n_maps = len(all_halocat_files)
    
    halocat_files = all_halocat_files[:n_maps]
    
    print(f"\nProcessing {len(halocat_files)} halo catalogs")
    print("Parameters varying:")
    print("  - sigma_8, omega_m: from halo catalogs")
    print("  - a_off: [4, 10]")
    print("  - b_off: [0, 2]")
    print(f"  - log_noise_level: [{generator.log_noise_range[0]}, {generator.log_noise_range[1]}]")
    print(f"    (noise level: [10^{generator.log_noise_range[0]:.1f}, 10^{generator.log_noise_range[1]:.1f}])")
    print(f"Store maps in memory: {store_maps}")
    print(f"Output directory: {output_dir}")
    
    n_processed = generator.process_batch(
        halocat_files=halocat_files,
        store_maps=store_maps
    )
    
    generator.save_results(output_dir, seed=seed, save_intensity_maps=store_maps)
    
    print(f"\nProcessed {n_processed} parameter combinations")
    print("\nParameter summary:")
    for i, name in enumerate(generator.param_names):
        values = generator.params_array[:, i]
        print(f"  {name}: [{values.min():.3f}, {values.max():.3f}] (mean={values.mean():.3f}, std={values.std():.3f})")
    
    if generator.log_noise_values is not None:
        print(f"\nNoise level distribution:")
        print(f"  Min noise: {10**generator.log_noise_values.min():.2e}")
        print(f"  Max noise: {10**generator.log_noise_values.max():.2e}")
        print(f"  Median noise: {10**np.median(generator.log_noise_values):.2e}")
    
    return generator


if __name__ == "__main__":
    # ========================================================================
    # QUICK EXAMPLES - Uncomment the one you want to run
    # ========================================================================
    
    # Example 1: Varying cosmology with 10 maps
    # generator = example_varying_cosmology(n_maps=10, store_maps=True)
    
    generator = example_varying_cosmology(
        n_maps=30,
        input_dir="/Users/anirbanroy/Desktop/halocats/",
        output_dir="/Users/anirbanroy/Desktop/test_maps_all_params/",
        store_maps=True,
        seed=42
    )
    
    
    # Example 2: Varying cosmology with 3000 maps (full dataset)
    # generator = example_varying_cosmology(n_maps=3000, store_maps=False)
    
    # Example 3: Varying all parameters including noise
    #generator = example_varying_all_parameters(
    #    n_maps=3000,
    #    input_dir="/Users/anirbanroy/Desktop/halocats/",
    #    output_dir="/Users/anirbanroy/Desktop/test_maps_all_params_varying_noise/",
    #    store_maps=True,
    #    seed=42
    #)
    
    # Example 4: Fixed cosmology mode
    # generator = example_fixed_cosmology(n_maps=100, store_maps=True)
    
    # Example 5: Fixed cosmology with noise as parameter
    # generator = example_fixed_cosmology_with_noise(n_maps=100)
    
    # Example 6: Hybrid mode for compatibility
    # generator = example_hybrid_mode(n_maps=50)
    
    # Optional: Check if maps were stored
    if generator and len(generator.intensity_maps) > 0:
        print(f"\n✓ Successfully stored {len(generator.intensity_maps)} maps in memory")
        print(f"  First map shape: {generator.intensity_maps[0].shape}")
        print(f"  Memory usage: ~{len(generator.intensity_maps) * generator.intensity_maps[0].nbytes / 1e9:.2f} GB")
```
