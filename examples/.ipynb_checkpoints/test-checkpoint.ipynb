```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any

from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)
from limference.convergence import ConvergenceAnalyzer

class SBIConvergenceTester:
    """
    General class for testing any SBI method with convergence analysis.
    Works with any set of parameters specified by the user.
    """
    
    def __init__(
        self,
        param_names: List[str],  # Specify which parameters to use
        param_labels: Optional[Dict[str, str]] = None,  # LaTeX labels for plotting
        inference_methods: List[str] = ["NPE"],
        analysis_methods: List[str] = ["pdf"],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        """
        Initialize tester with flexible parameter specification.
        
        Args:
            param_names: List of parameter names to constrain (e.g., ['omega_m', 'h'])
            param_labels: Optional dict mapping param names to LaTeX labels for plots
            inference_methods: SBI methods to use
            analysis_methods: Data analysis methods (pdf, power_spectrum, etc.)
            n_train: Number of training samples
            n_test: Number of test samples
            n_calibration: Number of calibration samples
            config_kwargs: Additional configuration options
        """
        self.param_names = param_names
        self.n_params = len(param_names)
        
        # Set up parameter labels for plotting
        if param_labels is None:
            self.param_labels = self._create_default_labels(param_names)
        else:
            self.param_labels = param_labels
            # Fill in any missing labels with defaults
            for name in param_names:
                if name not in self.param_labels:
                    self.param_labels[name] = self._create_default_labels([name])[name]
        
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
    
    def _create_default_labels(self, names: List[str]) -> Dict[str, str]:
        """Create default LaTeX labels for common parameter names."""
        default_labels = {
            'omega_m': r'$\Omega_m$',
            'Omega_m': r'$\Omega_m$',
            'h': r'$h$',
            'H0': r'$H_0$',
            'sigma8': r'$\sigma_8$',
            'sigma_8': r'$\sigma_8$',
            'ns': r'$n_s$',
            'n_s': r'$n_s$',
            'omega_b': r'$\Omega_b$',
            'Omega_b': r'$\Omega_b$',
            'omega_c': r'$\Omega_c$',
            'Omega_c': r'$\Omega_c$',
            'A_s': r'$A_s$',
            'tau': r'$\tau$',
            'w0': r'$w_0$',
            'wa': r'$w_a$',
            'mnu': r'$\sum m_\nu$',
            'omh2': r'$\Omega_m h^2$',
            'ombh2': r'$\Omega_b h^2$',
            'omch2': r'$\Omega_c h^2$'
        }
        
        labels = {}
        for name in names:
            if name in default_labels:
                labels[name] = default_labels[name]
            else:
                # Create a simple label for unknown parameters
                labels[name] = name.replace('_', ' ')
        
        return labels
    
    def load_data(self, sim_root: str, seed: int = 1, param_file_keys: Optional[Dict[str, str]] = None):
        """
        Load simulation data with flexible parameter extraction.
        
        Args:
            sim_root: Root directory for simulation data
            seed: Random seed for data files
            param_file_keys: Optional mapping from param_names to keys in the data file
                            e.g., {'omega_m': 'Omega_m', 'h': 'hubble'}
        """
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Extract specified parameters
        self.params_array = self._extract_parameters(self.params_sims, param_file_keys)
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        # Print summary
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters being constrained: {', '.join(self.param_names)}")
        print(f"Parameter ranges:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def _extract_parameters(self, params_sims: Any, param_file_keys: Optional[Dict[str, str]] = None) -> np.ndarray:
        """
        Extract specified parameters from simulation data.
        
        Args:
            params_sims: Parameter data from file
            param_file_keys: Mapping from desired param names to file keys
            
        Returns:
            Array of shape (n_sims, n_params) with extracted parameters
        """
        if param_file_keys is None:
            param_file_keys = {}
        
        params_list = []
        
        for sim_params in params_sims:
            param_values = []
            
            for param_name in self.param_names:
                # Check if there's a mapping for this parameter
                file_key = param_file_keys.get(param_name, param_name)
                
                # Try to extract the parameter with various possible keys
                value = None
                
                # Try direct access
                if file_key in sim_params:
                    value = sim_params[file_key]
                # Try common variations
                elif param_name in sim_params:
                    value = sim_params[param_name]
                # Try with underscores vs no underscores
                elif param_name.replace('_', '') in sim_params:
                    value = sim_params[param_name.replace('_', '')]
                elif param_name.replace('_', '8') in sim_params:  # for sigma_8 -> sigma8
                    value = sim_params[param_name.replace('_', '8')]
                # Try capitalized versions
                elif param_name.capitalize() in sim_params:
                    value = sim_params[param_name.capitalize()]
                elif 'Omega_' + param_name.split('_')[-1] in sim_params:  # omega_m -> Omega_m
                    value = sim_params['Omega_' + param_name.split('_')[-1]]
                else:
                    raise KeyError(f"Could not find parameter '{param_name}' (or '{file_key}') in simulation data. "
                                 f"Available keys: {list(sim_params.keys())}")
                
                param_values.append(value)
            
            params_list.append(param_values)
        
        return np.array(params_list)
    
    def setup_limference(self, output_root: str):
        """Initialize LIMference with configuration for specified parameters."""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Create configuration
        from limference import create_training_kwargs
        
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "device": "cuda" if torch.cuda.is_available() else "cpu"
        }
        default_config.update(self.config_kwargs)
        
        self.config = LIMConfig(**default_config)
        
        # Create parameter configuration
        param_ranges = {}
        for i, name in enumerate(self.param_names):
            param_ranges[name] = (
                float(self.params_array[:, i].min()),
                float(self.params_array[:, i].max())
            )
        
        self.param_config = ParameterConfig(
            param_names=self.param_names,
            param_ranges=param_ranges,
            param_labels=self.param_labels
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        print(f"Parameters to constrain: {', '.join(self.param_names)}")
    
    def train_models(self, test_obs_idx: int = 0, training_kwargs: Optional[Dict] = None):
        """Train all specified SBI models."""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        for i, name in enumerate(self.param_names):
            print(f"  {name} = {self.test_true_params[i]:.3f}")
        
        # Create true params dict
        true_params_dict = {
            name: float(self.test_true_params[i]) 
            for i, name in enumerate(self.param_names)
        }
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params=true_params_dict,
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            training_kwargs=training_kwargs,
            verbose=True
        )
        
        self.models_trained = True
        self.lim.results = self.lim_results  # Store for later access
    
    def analyze_sensitivity(self):
        """
        Analyze how sensitive the data (PDF/PS) is to each parameter.
        Works for any number of parameters.
        """
        if not self.data_loaded:
            print("Please load data first!")
            return
        
        from sklearn.decomposition import PCA
        from sklearn.preprocessing import StandardScaler
        
        # Determine what data to analyze
        if "pdf" in self.analysis_methods:
            data_to_analyze = self.pdf_sim_linear[:500]
            data_name = "PDF"
        elif "power_spectrum" in self.analysis_methods:
            data_to_analyze = self.pk_sim[:500]
            data_name = "Power Spectrum"
        else:
            print("No recognized analysis method for sensitivity analysis")
            return
        
        n_params = len(self.param_names)
        
        # Create appropriate subplot layout
        n_cols = min(3, n_params + 1)
        n_rows = (n_params + n_cols) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        axes = axes.flatten() if n_rows * n_cols > 1 else [axes]
        
        # Plot 1: PCA colored by each parameter
        ax_idx = 0
        
        # Standardize data for PCA
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data_to_analyze)
        
        # Apply PCA
        pca = PCA(n_components=2)
        data_pca = pca.fit_transform(data_scaled)
        
        for param_idx, param_name in enumerate(self.param_names):
            if ax_idx < len(axes):
                ax = axes[ax_idx]
                scatter = ax.scatter(data_pca[:, 0], data_pca[:, 1], 
                                   c=self.params_array[:500, param_idx], 
                                   cmap='viridis', alpha=0.6, s=20)
                plt.colorbar(scatter, ax=ax, label=self.param_labels[param_name])
                ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} var)')
                ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} var)')
                ax.set_title(f'PCA colored by {param_name}')
                ax.grid(True, alpha=0.3)
                ax_idx += 1
        
        # Plot 2: Correlation analysis
        if ax_idx < len(axes):
            ax = axes[ax_idx]
            
            all_correlations = {}
            for param_idx, param_name in enumerate(self.param_names):
                correlations = []
                for i in range(data_to_analyze.shape[1]):
                    if np.std(data_to_analyze[:, i]) > 1e-10:
                        corr = np.corrcoef(data_to_analyze[:, i], 
                                          self.params_array[:500, param_idx])[0, 1]
                    else:
                        corr = 0
                    correlations.append(corr)
                
                all_correlations[param_name] = correlations
                ax.plot(correlations, label=f'{data_name}-{param_name}', linewidth=2, alpha=0.7)
            
            ax.axhline(0, color='gray', linestyle='--', alpha=0.5)
            ax.axhline(0.1, color='gray', linestyle=':', alpha=0.3)
            ax.axhline(-0.1, color='gray', linestyle=':', alpha=0.3)
            ax.set_xlabel(f'{data_name} bin index')
            ax.set_ylabel('Correlation coefficient')
            ax.set_title(f'Parameter-{data_name} Correlations')
            ax.legend()
            ax.grid(True, alpha=0.3)
            ax.set_ylim(-0.5, 0.5)
            ax_idx += 1
        
        # Hide unused subplots
        for idx in range(ax_idx, len(axes)):
            axes[idx].set_visible(False)
        
        plt.suptitle(f'{data_name} Sensitivity to Parameters', fontsize=16)
        plt.tight_layout()
        plt.show()
        
        # Print analysis summary
        print("\n" + "="*60)
        print(f"{data_name.upper()} INFORMATION CONTENT ANALYSIS")
        print("="*60)
        
        print(f"\nParameter ranges in data:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
        
        print(f"\nCorrelation with {data_name} bins:")
        max_corrs = {}
        for param_name in self.param_names:
            max_corr = np.max(np.abs(all_correlations[param_name]))
            mean_corr = np.mean(np.abs(all_correlations[param_name]))
            max_corrs[param_name] = max_corr
            print(f"  {param_name}:")
            print(f"    Maximum |correlation|: {max_corr:.3f}")
            print(f"    Mean |correlation|:    {mean_corr:.3f}")
        
        # Find the most and least informative parameters
        sorted_params = sorted(max_corrs.items(), key=lambda x: x[1], reverse=True)
        
        print(f"\nParameter informativeness ranking:")
        for i, (param, corr) in enumerate(sorted_params, 1):
            print(f"  {i}. {param}: {corr:.3f}")
        
        # Check PCA correlations
        print(f"\nPCA correlations:")
        for param_idx, param_name in enumerate(self.param_names):
            pc1_corr = np.corrcoef(data_pca[:, 0], self.params_array[:500, param_idx])[0, 1]
            pc2_corr = np.corrcoef(data_pca[:, 1], self.params_array[:500, param_idx])[0, 1]
            print(f"  {param_name}:")
            print(f"    PC1: {pc1_corr:.3f}")
            print(f"    PC2: {pc2_corr:.3f}")
        
        # Diagnosis
        print("\n" + "="*60)
        print("DIAGNOSIS:")
        
        poorly_constrained = [p for p, c in max_corrs.items() if c < 0.1]
        if poorly_constrained:
            print(f"❌ {data_name} contains very little information about: {', '.join(poorly_constrained)}")
            print("   These parameters will be poorly constrained")
        
        well_constrained = [p for p, c in max_corrs.items() if c > 0.3]
        if well_constrained:
            print(f"✓ {data_name} contains good information about: {', '.join(well_constrained)}")
        
        print("\nRECOMMENDATIONS:")
        if poorly_constrained:
            print(f"1. Consider using a different summary statistic for {', '.join(poorly_constrained)}")
            print("2. Try combined analysis methods (PS + PDF + field-level)")
            print("3. Check if your simulations properly vary these parameters")
    
    def plot_parameter_recovery(self):
        """Plot parameter recovery for any number of parameters."""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_params = len(self.param_names)
        
        fig, axes = plt.subplots(n_methods, n_params, figsize=(5*n_params, 4*n_methods))
        
        if n_methods == 1 and n_params == 1:
            axes = np.array([[axes]])
        elif n_methods == 1:
            axes = axes.reshape(1, -1)
        elif n_params == 1:
            axes = axes.reshape(-1, 1)
        
        method_idx = 0
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                
                test_params_subset = self.test_params[50:150]
                temp = result.get('calibration_temp', 1.0)
                
                # Collect predictions
                predictions_mean = []
                predictions_16 = []
                predictions_84 = []
                true_vals = []
                n_eval = min(50, len(test_data))
                
                for i in range(n_eval):
                    x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                    samples = result['posterior'].sample((1000,), x=x_obs).numpy()
                    
                    # Apply calibration if available
                    if 'calibration_temp' in result:
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                    else:
                        samples_cal = samples
                    
                    predictions_mean.append(np.mean(samples_cal, axis=0))
                    predictions_16.append(np.percentile(samples_cal, 16, axis=0))
                    predictions_84.append(np.percentile(samples_cal, 84, axis=0))
                    true_vals.append(test_params_subset[i])
                
                predictions_mean = np.array(predictions_mean)
                predictions_16 = np.array(predictions_16)
                predictions_84 = np.array(predictions_84)
                true_vals = np.array(true_vals)
                
                # Plot for each parameter
                for param_idx, param_name in enumerate(self.param_names):
                    ax = axes[method_idx, param_idx]
                    
                    # Error bars
                    yerr_lower = predictions_mean[:, param_idx] - predictions_16[:, param_idx]
                    yerr_upper = predictions_84[:, param_idx] - predictions_mean[:, param_idx]
                    yerr = np.array([yerr_lower, yerr_upper])
                    
                    ax.errorbar(true_vals[:, param_idx], 
                              predictions_mean[:, param_idx],
                              yerr=yerr,
                              fmt='o', alpha=0.6, markersize=4,
                              elinewidth=0.5, capsize=2, capthick=0.5)
                    
                    # Perfect recovery line
                    lims = [true_vals[:, param_idx].min(), true_vals[:, param_idx].max()]
                    ax.plot(lims, lims, 'r--', alpha=0.7, linewidth=2)
                    
                    # Calculate metrics
                    from scipy.stats import pearsonr
                    r2 = pearsonr(true_vals[:, param_idx], 
                                predictions_mean[:, param_idx])[0]**2
                    rmse = np.sqrt(np.mean((true_vals[:, param_idx] - 
                                          predictions_mean[:, param_idx])**2))
                    
                    ax.set_xlabel(f'True {self.param_labels[param_name]}')
                    ax.set_ylabel(f'Predicted {self.param_labels[param_name]}')
                    ax.set_title(f'{method_key}\nR²={r2:.3f}, RMSE={rmse:.4f}')
                    ax.grid(True, alpha=0.3)
                
                method_idx += 1
        
        plt.suptitle('Parameter Recovery Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()

# Additional helper methods remain the same but update param references...
```

```python
# %% [markdown]
# # General SBI Method Convergence Testing with Calibration and Uncertainty Analysis
# Works with any SBI method: SNPE, NPE, NRE, SNLE, etc.

# %% Import libraries
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any

# Import LIMference
from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)

from limference.convergence import ConvergenceAnalyzer

# %% [markdown]
# ## Configuration Section - Change These Parameters

# %%
# ============ MAIN CONFIGURATION ============
# Change these parameters to test different methods

# SBI Method Selection
INFERENCE_METHODS = ["NPE"]  # Can be any combination of: SNPE, NPE, NRE, SNLE, SNL
ANALYSIS_METHODS = ["pdf", "power_spectrum"]  # Can include: power_spectrum, pdf, field_level

# Data Configuration
N_TRAIN = 2000  # Number of training samples
N_TEST = 500    # Number of test samples
N_CALIBRATION = 50  # Number of samples for temperature calibration



# In the configuration section, modify these parameters:
MAX_EPOCHS = 200  # Reduce from 300
STOP_AFTER_EPOCHS = 30  # Increase patience from 20
BATCH_SIZE = 100  # Increase from 50 for better gradient estimates
LEARNING_RATE = 1e-4  # Reduce from 5e-4
VALIDATION_FRACTION = 0.3  # Increase from 0.2
WEIGHT_DECAY = 1e-4  # Add weight decay for regularization

NOISE_REDUCTION_FACTOR = 50.0

# File paths
SEED = 2
SIM_ROOT = f"/Users/anirbanroy/Desktop/product_with_noise_{SEED}/"
OUTPUT_ROOT = "/Users/anirbanroy/Desktop/output_small_z_3.600_sorted"

# Device
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# %% [markdown]
# ## 1. Load and Prepare Data

# %%
class SBIConvergenceTester:
    """General class for testing any SBI method with convergence analysis"""
    
    def __init__(
        self,
        inference_methods: List[str],
        analysis_methods: List[str],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
        
    def load_data(self, sim_root: str, seed: int = 1):
        """Load simulation data"""
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Convert parameters
        self.params_array = np.array([[d['sigma8'], d['omega_m']] for d in self.params_sims])
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters range:")
        print(f"  sigma8: [{self.params_array[:, 0].min():.3f}, {self.params_array[:, 0].max():.3f}]")
        print(f"  omega_m: [{self.params_array[:, 1].min():.3f}, {self.params_array[:, 1].max():.3f}]")
        
    def setup_limference(self, output_root: str):
        """Initialize LIMference with configuration"""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Create configuration
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "max_num_epochs": MAX_EPOCHS,
            "stop_after_epochs": STOP_AFTER_EPOCHS,
            "training_batch_size": BATCH_SIZE,
            "learning_rate": LEARNING_RATE,
            "validation_fraction": VALIDATION_FRACTION,
            "noise_reduction_factor": NOISE_REDUCTION_FACTOR,
            "device": DEVICE
        }
        default_config.update(self.config_kwargs)
        
        self.config = LIMConfig(**default_config)
        
        self.param_config = ParameterConfig(
            param_names=["sigma8", "omega_m"],
            param_ranges={
                "sigma8": (self.params_array[:, 0].min(), self.params_array[:, 0].max()),
                "omega_m": (self.params_array[:, 1].min(), self.params_array[:, 1].max())
            }
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        
    def train_models(self, test_obs_idx: int = 0):
        """Train all specified SBI models"""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        print(f"  sigma8 = {self.test_true_params[0]:.3f}")
        print(f"  omega_m = {self.test_true_params[1]:.3f}")
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params={
                "sigma8": self.test_true_params[0], 
                "omega_m": self.test_true_params[1]
            },
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            verbose=True
        )
        
        self.models_trained = True
        
    def calibrate_posteriors(self, target_coverage: float = 0.68):
        """Calibrate all posteriors using temperature scaling"""
        if not self.models_trained:
            raise ValueError("Train models first using train_models()")
        
        print("\n" + "="*50)
        print("CALIBRATING POSTERIORS")
        print("="*50)
        
        def calibrate_temperature(posterior, cal_data, cal_params, target=0.68):
            """Find optimal temperature scaling"""
            def coverage_error(temp):
                coverage = 0
                n_cal = min(30, len(cal_data))
                
                for i in range(n_cal):
                    x_obs = torch.tensor(cal_data[i], dtype=torch.float32)
                    samples = posterior.sample((300,), x=x_obs).numpy()
                    
                    mean = np.mean(samples, axis=0)
                    scaled = mean + (samples - mean) * temp
                    
                    q16, q84 = np.percentile(scaled, [16, 84], axis=0)
                    if np.all((cal_params[i] >= q16) & (cal_params[i] <= q84)):
                        coverage += 1
                
                return abs(coverage/n_cal - target)
            
            result = minimize_scalar(coverage_error, bounds=(0.5, 3.0), method='bounded')
            return result.x
        
        # Calibrate each method
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Determine which test data to use
                if "power" in method_key:
                    cal_data = self.test_pk[:self.n_calibration]
                elif "pdf" in method_key:
                    cal_data = self.test_pdf[:self.n_calibration]
                else:
                    continue
                
                cal_params = self.test_params[:self.n_calibration]
                
                # Find optimal temperature
                temp = calibrate_temperature(
                    result['posterior'], cal_data, cal_params, target_coverage
                )
                
                print(f"{method_key}: temperature = {temp:.3f}")
                result['calibration_temp'] = temp
                
                # Interpretation
                if temp > 1.2:
                    print(f"  → Model was overconfident, widening by {(temp-1)*100:.1f}%")
                elif temp < 0.8:
                    print(f"  → Model was underconfident, narrowing by {(1-temp)*100:.1f}%")
                else:
                    print(f"  → Model was well-calibrated")
    
    def analyze_convergence(self):
        """Analyze training convergence for all methods"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        analyzer = ConvergenceAnalyzer(self.config, self.param_config)
        
        # Determine plot layout
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        if n_methods == 1:
            axes = [axes]
        elif n_rows == 1:
            axes = axes.reshape(1, -1)
        
        convergence_results = {}
        plot_idx = 0
        
        for method_key, result in self.lim.results.items():
            if "error" not in result and 'loss_history' in result:
                print(f"\nAnalyzing {method_key}...")
                
                # Extract metrics
                metrics = self._extract_metrics(result['loss_history'])
                convergence_results[method_key] = metrics
                
                # Plot
                if plot_idx < n_rows * n_cols:
                    row = plot_idx // n_cols
                    col = plot_idx % n_cols
                    ax = axes[row, col] if n_rows > 1 else axes[col]
                    
                    self._plot_convergence(ax, metrics, method_key)
                    plot_idx += 1
                
                # Print metrics
                self._print_convergence_metrics(metrics)
        
        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col] if n_rows > 1 else axes[col]
            ax.set_visible(False)
        
        plt.suptitle(f'Convergence Analysis: {", ".join(self.inference_methods)}', fontsize=14)
        plt.tight_layout()
        plt.show()
        
        self.convergence_results = convergence_results
        
        
    def analyze_pdf_sensitivity(self):
        """Check how sensitive the PDF is to parameter changes"""

        # First ensure we have data
        if not self.data_loaded:
            print("Please load data first!")
            return

        import matplotlib.pyplot as plt
        from sklearn.decomposition import PCA
        from sklearn.preprocessing import StandardScaler

        # Create figure with subplots
        fig, axes = plt.subplots(2, 2, figsize=(14, 12))

        # Plot 1: PDF variation with omega_m at fixed sigma8
        ax = axes[0, 0]

        # Find simulations with sigma8 in narrow range (around median)
        median_sigma8 = np.median(self.params_array[:, 0])
        tolerance = 0.02
        mask = np.abs(self.params_array[:500, 0] - median_sigma8) < tolerance

        if np.sum(mask) > 10:  # Need at least 10 samples
            subset_params = self.params_array[:500][mask]
            subset_pdf = self.pdf_sim_linear[:500][mask]

            # Sort by omega_m for clearer visualization
            sort_idx = np.argsort(subset_params[:, 1])
            subset_params = subset_params[sort_idx]
            subset_pdf = subset_pdf[sort_idx]

            # Plot several PDFs with different omega_m
            n_samples = min(5, len(subset_params))
            colors = plt.cm.viridis(np.linspace(0, 1, n_samples))

            for i in range(n_samples):
                ax.plot(subset_pdf[i], alpha=0.7, color=colors[i],
                       label=f'Ωₘ={subset_params[i, 1]:.3f}')

            ax.set_xlabel('PDF bin')
            ax.set_ylabel('PDF value')
            
            ax.set_yscale('log')
            ax.set_xscale('log')
            ax.set_title(f'PDFs at σ₈≈{median_sigma8:.3f}, varying Ωₘ')
            ax.legend(fontsize=8)
            ax.grid(True, alpha=0.3)

        # Plot 2: Mean PDF for different omega_m ranges
        ax = axes[0, 1]

        # Split data into low and high omega_m
        omega_m_percentiles = np.percentile(self.params_array[:, 1], [25, 75])
        omega_m_low = self.params_array[:500, 1] < omega_m_percentiles[0]
        omega_m_high = self.params_array[:500, 1] > omega_m_percentiles[1]

        mean_pdf_low = np.mean(self.pdf_sim_linear[:500][omega_m_low], axis=0)
        std_pdf_low = np.std(self.pdf_sim_linear[:500][omega_m_low], axis=0)
        mean_pdf_high = np.mean(self.pdf_sim_linear[:500][omega_m_high], axis=0)
        std_pdf_high = np.std(self.pdf_sim_linear[:500][omega_m_high], axis=0)

        bins = np.arange(len(mean_pdf_low))

        # Plot with error bands
        ax.plot(bins, mean_pdf_low, 'b-', label=f'Ωₘ < {omega_m_percentiles[0]:.3f} (n={omega_m_low.sum()})')
        ax.fill_between(bins, mean_pdf_low - std_pdf_low, mean_pdf_low + std_pdf_low, 
                        alpha=0.2, color='blue')

        ax.plot(bins, mean_pdf_high, 'r-', label=f'Ωₘ > {omega_m_percentiles[1]:.3f} (n={omega_m_high.sum()})')
        ax.fill_between(bins, mean_pdf_high - std_pdf_high, mean_pdf_high + std_pdf_high, 
                        alpha=0.2, color='red')

        # Calculate and display relative difference
        rel_diff = (mean_pdf_high - mean_pdf_low) / (mean_pdf_low + 1e-10)
        ax2 = ax.twinx()
        ax2.plot(bins, rel_diff * 100, 'g--', alpha=0.5, linewidth=1)
        ax2.set_ylabel('Relative diff (%)', color='g')
        ax2.tick_params(axis='y', labelcolor='g')

        ax.set_xlabel('PDF bin')
        ax.set_ylabel('Mean PDF')
        ax.set_title('Mean PDF: Low vs High Ωₘ')
        ax.legend()
        ax.grid(True, alpha=0.3)

        # Plot 3: PCA analysis
        ax = axes[1, 0]

        # Standardize PDFs for PCA
        scaler = StandardScaler()
        pdf_scaled = scaler.fit_transform(self.pdf_sim_linear[:500])

        # Apply PCA
        pca = PCA(n_components=2)
        pdf_pca = pca.fit_transform(pdf_scaled)

        # Create scatter plot colored by omega_m
        scatter = ax.scatter(pdf_pca[:, 0], pdf_pca[:, 1], 
                            c=self.params_array[:500, 1], 
                            cmap='viridis', alpha=0.6, s=20)
        plt.colorbar(scatter, ax=ax, label='Ωₘ')

        ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
        ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
        ax.set_title('PCA of PDFs colored by Ωₘ')
        ax.grid(True, alpha=0.3)

        # Plot 4: Correlation analysis
        ax = axes[1, 1]

        correlations_omega = []
        correlations_sigma = []

        # Calculate correlation for each PDF bin
        for i in range(self.pdf_sim_linear.shape[1]):
            # Skip if variance is too small (constant bin)
            if np.std(self.pdf_sim_linear[:500, i]) > 1e-10:
                corr_omega = np.corrcoef(self.pdf_sim_linear[:500, i], 
                                         self.params_array[:500, 1])[0, 1]
                corr_sigma = np.corrcoef(self.pdf_sim_linear[:500, i], 
                                         self.params_array[:500, 0])[0, 1]
            else:
                corr_omega = 0
                corr_sigma = 0

            correlations_omega.append(corr_omega)
            correlations_sigma.append(corr_sigma)

        x = range(len(correlations_omega))
        ax.plot(x, correlations_omega, 'b-', label='PDF-Ωₘ correlation', linewidth=2)
        ax.plot(x, correlations_sigma, 'r-', label='PDF-σ₈ correlation', linewidth=2)
        ax.axhline(0, color='gray', linestyle='--', alpha=0.5)
        ax.axhline(0.1, color='gray', linestyle=':', alpha=0.3)
        ax.axhline(-0.1, color='gray', linestyle=':', alpha=0.3)

        ax.set_xlabel('PDF bin index')
        ax.set_ylabel('Correlation coefficient')
        ax.set_title('Parameter-PDF Correlations by Bin')
        ax.legend()
        ax.grid(True, alpha=0.3)
        ax.set_ylim(-0.5, 0.5)

        plt.suptitle('PDF Sensitivity to Parameters Analysis', fontsize=16, y=1.02)
        plt.tight_layout()
        plt.show()

        # Print detailed statistics
        print("\n" + "="*60)
        print("PDF INFORMATION CONTENT ANALYSIS")
        print("="*60)

        print(f"\nParameter ranges in data:")
        print(f"  σ₈:  [{self.params_array[:, 0].min():.3f}, {self.params_array[:, 0].max():.3f}]")
        print(f"  Ωₘ:  [{self.params_array[:, 1].min():.3f}, {self.params_array[:, 1].max():.3f}]")

        print(f"\nCorrelation with PDF bins:")
        print(f"  Maximum |correlation| with Ωₘ:  {np.max(np.abs(correlations_omega)):.3f}")
        print(f"  Mean |correlation| with Ωₘ:     {np.mean(np.abs(correlations_omega)):.3f}")
        print(f"  Maximum |correlation| with σ₈:  {np.max(np.abs(correlations_sigma)):.3f}")
        print(f"  Mean |correlation| with σ₈:     {np.mean(np.abs(correlations_sigma)):.3f}")

        info_ratio = np.max(np.abs(correlations_omega)) / (np.max(np.abs(correlations_sigma)) + 1e-10)
        print(f"  Information ratio (Ωₘ/σ₈):      {info_ratio:.3f}")

        print(f"\nPCA variance explained:")
        print(f"  PC1: {pca.explained_variance_ratio_[0]:.1%}")
        print(f"  PC2: {pca.explained_variance_ratio_[1]:.1%}")
        print(f"  Total (first 2): {sum(pca.explained_variance_ratio_[:2]):.1%}")

        # Check if PCs correlate with parameters
        pc1_omega_corr = np.corrcoef(pdf_pca[:, 0], self.params_array[:500, 1])[0, 1]
        pc1_sigma_corr = np.corrcoef(pdf_pca[:, 0], self.params_array[:500, 0])[0, 1]
        pc2_omega_corr = np.corrcoef(pdf_pca[:, 1], self.params_array[:500, 1])[0, 1]
        pc2_sigma_corr = np.corrcoef(pdf_pca[:, 1], self.params_array[:500, 0])[0, 1]

        print(f"\nPrincipal component correlations:")
        print(f"  PC1 with Ωₘ: {pc1_omega_corr:.3f}")
        print(f"  PC1 with σ₈: {pc1_sigma_corr:.3f}")
        print(f"  PC2 with Ωₘ: {pc2_omega_corr:.3f}")
        print(f"  PC2 with σ₈: {pc2_sigma_corr:.3f}")

        # Diagnosis
        print("\n" + "="*60)
        print("DIAGNOSIS:")

        if info_ratio < 0.3:
            print("❌ PDF contains very little information about Ωₘ")
            print("   The PDF is ~{:.0f}x less sensitive to Ωₘ than to σ₈".format(1/info_ratio))
        elif info_ratio < 0.6:
            print("⚠️  PDF contains limited information about Ωₘ")
        else:
            print("✓  PDF contains reasonable information about Ωₘ")

        if np.max(np.abs(correlations_omega)) < 0.1:
            print("❌ No PDF bins show meaningful correlation with Ωₘ")
            print("   This suggests the PDF is fundamentally insensitive to Ωₘ")

        if abs(pc1_omega_corr) < 0.2 and abs(pc2_omega_corr) < 0.2:
            print("❌ Principal components don't capture Ωₘ variation")

        print("\nRECOMMENDATIONS:")
        if info_ratio < 0.3:
            print("1. Switch to power spectrum - it contains spatial information")
            print("2. Use combined analysis (PS + PDF) for complementary constraints")
            print("3. Consider alternative statistics (Minkowski functionals, peaks)")
            print("4. Check if your simulations properly vary Ωₘ effects")

    def test_recovery(self):
        """Test posterior recovery on test set"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        print("\n" + "="*50)
        print("TESTING POSTERIOR RECOVERY")
        print("="*50)
        
        recovery_results = {}
        
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                print(f"\n{method_key}:")
                
                # Get appropriate test data
                if "power" in method_key:
                    test_data = self.test_pk[self.n_calibration:]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[self.n_calibration:]
                else:
                    continue
                
                test_params_subset = self.test_params[self.n_calibration:]
                
                # Test recovery
                metrics = self._test_single_method_recovery(
                    result, test_data, test_params_subset
                )
                recovery_results[method_key] = metrics
                
                # Print results
                self._print_recovery_metrics(metrics)
        
        self.recovery_results = recovery_results
        
    def plot_results(self):
        """Create comprehensive visualization of results"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        # 1. Coverage calibration curves
        self._plot_coverage_curves()
        
        # 2. Parameter recovery scatter plots
        self._plot_parameter_recovery()
        
        # 3. Uncertainty analysis
        self._plot_uncertainty_analysis()
        
    def generate_report(self):
        """Generate final summary report"""
        print("\n" + "="*70)
        print(f"SBI METHODS CONVERGENCE TEST SUMMARY")
        print(f"Methods tested: {', '.join(self.inference_methods)}")
        print(f"Analysis types: {', '.join(self.analysis_methods)}")
        print("="*70)
        
        for method_key in self.lim.results:
            if "error" not in self.lim.results[method_key]:
                print(f"\n{method_key}:")
                print("-"*40)
                
                # Convergence info
                if hasattr(self, 'convergence_results') and method_key in self.convergence_results:
                    conv = self.convergence_results[method_key]
                    print(f"Training:")
                    print(f"  Epochs: {conv.get('n_epochs', 'N/A')}")
                    print(f"  Converged: {conv.get('converged', 'Unknown')}")
                    if conv.get('overfitting'):
                        print(f"  ⚠️ Overfitting detected")
                
                # Calibration info
                result = self.lim.results[method_key]
                if 'calibration_temp' in result:
                    print(f"Calibration:")
                    print(f"  Temperature: {result['calibration_temp']:.3f}")
                
                # Recovery performance
                if hasattr(self, 'recovery_results') and method_key in self.recovery_results:
                    rec = self.recovery_results[method_key]
                    print(f"Performance:")
                    print(f"  RMSE: {rec['mean_rmse']:.4f}")
                    print(f"  68% coverage: {rec['coverage_68_rate']:.1%}")
                    print(f"  95% coverage: {rec['coverage_95_rate']:.1%}")
                    
                    if rec['well_calibrated_68'] and rec['well_calibrated_95']:
                        print("  ✅ Well calibrated")
                    else:
                        print("  ⚠️ Needs adjustment")
    
    # Helper methods
    def _extract_metrics(self, loss_history):
        """Extract metrics from loss history"""
        metrics = {}
        
        # Handle different possible formats
        train_losses = loss_history.get('training_loss', 
                                       loss_history.get('train_log_probs', []))
        val_losses = loss_history.get('validation_loss',
                                     loss_history.get('validation_log_probs', []))
        
        if isinstance(train_losses, np.ndarray):
            train_losses = train_losses.tolist()
        if isinstance(val_losses, np.ndarray):
            val_losses = val_losses.tolist()
            
        metrics['train_losses'] = train_losses
        metrics['val_losses'] = val_losses
        
        # Get epochs
        epochs = loss_history.get('epochs_trained', len(train_losses))
        if isinstance(epochs, (list, np.ndarray)):
            epochs = int(epochs[0] if len(epochs) > 0 else len(train_losses))
        metrics['n_epochs'] = epochs
        
        # Check convergence
        if len(train_losses) > 10:
            window = len(train_losses) // 10
            recent = train_losses[-window:]
            metrics['converged'] = np.std(recent) < 0.01 * abs(np.mean(recent))
        
        # Check overfitting
        if train_losses and val_losses:
            gap = val_losses[-1] - train_losses[-1]
            metrics['overfitting'] = gap > 0.5
            
        return metrics
    
    def _plot_convergence(self, ax, metrics, method_key):
        """Plot convergence curves"""
        if metrics['train_losses']:
            ax.plot(range(1, len(metrics['train_losses'])+1), 
                   metrics['train_losses'], 'b-', label='Training', linewidth=2)
        
        if metrics['val_losses']:
            ax.plot(range(1, len(metrics['val_losses'])+1), 
                   metrics['val_losses'], 'r-', label='Validation', linewidth=2)
        
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(method_key.replace('_', ' '))
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    def _print_convergence_metrics(self, metrics):
        """Print convergence metrics"""
        print(f"  Epochs: {metrics.get('n_epochs', 'N/A')}")
        print(f"  Converged: {metrics.get('converged', 'Unknown')}")
        if metrics.get('overfitting'):
            print(f"  ⚠️ Potential overfitting")
    
    def _test_single_method_recovery(self, result, test_data, test_params):
        """Test recovery for a single method"""
        n_test_samples = min(50, len(test_data))
        coverage_68 = 0
        coverage_95 = 0
        errors = []
        
        temp = result.get('calibration_temp', 1.0)
        
        for i in range(n_test_samples):
            x_obs = torch.tensor(test_data[i], dtype=torch.float32)
            samples = result['posterior'].sample((1000,), x=x_obs).numpy()
            
            # Apply temperature scaling
            mean = np.mean(samples, axis=0)
            samples = mean + (samples - mean) * temp
            
            # Calculate statistics
            post_mean = np.mean(samples, axis=0)
            errors.append(post_mean - test_params[i])
            
            # Check coverage
            q16, q84 = np.percentile(samples, [16, 84], axis=0)
            q2_5, q97_5 = np.percentile(samples, [2.5, 97.5], axis=0)
            
            if np.all((test_params[i] >= q16) & (test_params[i] <= q84)):
                coverage_68 += 1
            if np.all((test_params[i] >= q2_5) & (test_params[i] <= q97_5)):
                coverage_95 += 1
        
        errors = np.array(errors)
        
        return {
            'coverage_68_rate': coverage_68 / n_test_samples,
            'coverage_95_rate': coverage_95 / n_test_samples,
            'mean_rmse': np.sqrt(np.mean(errors**2)),
            'mean_bias': np.mean(np.abs(errors)),
            'well_calibrated_68': abs(coverage_68/n_test_samples - 0.68) < 0.1,
            'well_calibrated_95': abs(coverage_95/n_test_samples - 0.95) < 0.1,
            'calibration_temp': temp
        }
    
    def _print_recovery_metrics(self, metrics):
        """Print recovery metrics"""
        print(f"  Temperature: {metrics['calibration_temp']:.3f}")
        print(f"  68% coverage: {metrics['coverage_68_rate']:.1%}")
        print(f"  95% coverage: {metrics['coverage_95_rate']:.1%}")
        print(f"  RMSE: {metrics['mean_rmse']:.4f}")
        
    def _plot_coverage_curves(self):
        """Plot coverage calibration curves"""
        confidence_levels = np.array([50, 60, 70, 80, 90, 95, 99])
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        if n_methods == 1:
            axes = [axes]
        elif n_rows == 1:
            axes = axes.reshape(1, -1)
        
        plot_idx = 0
        
        for method_key, result in self.lim.results.items():
            if "error" not in result and plot_idx < n_rows * n_cols:
                row = plot_idx // n_cols
                col = plot_idx % n_cols
                ax = axes[row, col] if n_rows > 1 else axes[col]
                
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                    
                test_params_subset = self.test_params[50:150]
                
                # Calculate coverage for different confidence levels
                coverage_rates_orig = []
                coverage_rates_cal = []
                
                temp = result.get('calibration_temp', 1.0)
                
                for conf in confidence_levels:
                    coverage_orig = 0
                    coverage_cal = 0
                    n_samples = min(30, len(test_data))
                    
                    for i in range(n_samples):
                        x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                        samples = result['posterior'].sample((500,), x=x_obs).numpy()
                        
                        lower = (100-conf)/2
                        upper = 100-lower
                        
                        # Original
                        ci_low_orig = np.percentile(samples, lower, axis=0)
                        ci_high_orig = np.percentile(samples, upper, axis=0)
                        
                        # Calibrated
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                        ci_low_cal = np.percentile(samples_cal, lower, axis=0)
                        ci_high_cal = np.percentile(samples_cal, upper, axis=0)
                        
                        if np.all((test_params_subset[i] >= ci_low_orig) & 
                                 (test_params_subset[i] <= ci_high_orig)):
                            coverage_orig += 1
                        if np.all((test_params_subset[i] >= ci_low_cal) & 
                                 (test_params_subset[i] <= ci_high_cal)):
                            coverage_cal += 1
                    
                    coverage_rates_orig.append(100 * coverage_orig/n_samples)
                    coverage_rates_cal.append(100 * coverage_cal/n_samples)
                
                # Plot
                ax.plot(confidence_levels, coverage_rates_orig, 'o-', 
                       label='Original', linewidth=2, color='red')
                ax.plot(confidence_levels, coverage_rates_cal, 's-', 
                       label='Calibrated', linewidth=2, color='green')
                ax.plot([0, 100], [0, 100], 'k--', alpha=0.5, label='Perfect')
                
                ax.set_xlabel('Nominal Coverage (%)')
                ax.set_ylabel('Empirical Coverage (%)')
                ax.set_title(f'{method_key} (T={temp:.2f})')
                ax.legend(fontsize=8)
                ax.grid(True, alpha=0.3)
                ax.set_xlim(45, 100)
                ax.set_ylim(45, 100)
                
                plot_idx += 1
        
        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col] if n_rows > 1 else axes[col]
            ax.set_visible(False)
        
        plt.suptitle('Coverage Calibration: Before vs After', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def _plot_parameter_recovery(self):
        """Plot parameter recovery with error bars"""
        param_names = ['σ₈', 'Ωₘ']
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        
        fig, axes = plt.subplots(n_methods, 2, figsize=(10, 4*n_methods))
        if n_methods == 1:
            axes = axes.reshape(1, -1)
        
        for idx, (method_key, result) in enumerate(self.lim.results.items()):
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                    
                test_params_subset = self.test_params[50:150]
                temp = result.get('calibration_temp', 1.0)
                
                # Collect predictions
                predictions_mean = []
                predictions_16 = []
                predictions_84 = []
                true_vals = []
                n_eval = min(50, len(test_data))
                
                for i in range(n_eval):
                    x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                    samples = result['posterior'].sample((1000,), x=x_obs).numpy()
                    
                    # Apply calibration
                    mean = np.mean(samples, axis=0)
                    samples_cal = mean + (samples - mean) * temp
                    
                    predictions_mean.append(np.mean(samples_cal, axis=0))
                    predictions_16.append(np.percentile(samples_cal, 16, axis=0))
                    predictions_84.append(np.percentile(samples_cal, 84, axis=0))
                    true_vals.append(test_params_subset[i])
                
                predictions_mean = np.array(predictions_mean)
                predictions_16 = np.array(predictions_16)
                predictions_84 = np.array(predictions_84)
                true_vals = np.array(true_vals)
                
                for param_idx in range(2):
                    ax = axes[idx, param_idx]
                    
                    # Error bars
                    yerr_lower = predictions_mean[:, param_idx] - predictions_16[:, param_idx]
                    yerr_upper = predictions_84[:, param_idx] - predictions_mean[:, param_idx]
                    yerr = np.array([yerr_lower, yerr_upper])
                    
                    ax.errorbar(true_vals[:, param_idx], 
                               predictions_mean[:, param_idx],
                               yerr=yerr,
                               fmt='o', alpha=0.6, markersize=4,
                               elinewidth=0.5, capsize=2, capthick=0.5,
                               label='Posterior mean ± 68% CI')
                    
                    # Perfect recovery line
                    lims = [true_vals[:, param_idx].min(), true_vals[:, param_idx].max()]
                    ax.plot(lims, lims, 'r--', alpha=0.7, linewidth=2, 
                           label='Perfect recovery')
                    
                    # Calculate metrics
                    from scipy.stats import pearsonr
                    r2 = pearsonr(true_vals[:, param_idx], 
                                 predictions_mean[:, param_idx])[0]**2
                    rmse = np.sqrt(np.mean((true_vals[:, param_idx] - 
                                           predictions_mean[:, param_idx])**2))
                    
                    ax.set_xlabel(f'True {param_names[param_idx]}')
                    ax.set_ylabel(f'Predicted {param_names[param_idx]}')
                    ax.set_title(f'{method_key}\nR²={r2:.3f}, RMSE={rmse:.4f}')
                    ax.grid(True, alpha=0.3)
                    if param_idx == 0:
                        ax.legend(loc='upper left', fontsize=8)
        
        plt.suptitle('Parameter Recovery with Uncertainty', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def _plot_uncertainty_analysis(self):
        """Plot uncertainty analysis"""
        param_names = ['σ₈', 'Ωₘ']
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        
        fig, axes = plt.subplots(2, n_methods, figsize=(5*n_methods, 8))
        if n_methods == 1:
            axes = axes.reshape(-1, 1)
        
        for idx, (method_key, result) in enumerate(self.lim.results.items()):
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:100]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:100]
                else:
                    continue
                    
                test_params_subset = self.test_params[50:100]
                temp = result.get('calibration_temp', 1.0)
                
                for param_idx in range(2):
                    ax = axes[param_idx, idx]
                    
                    # Collect errors and uncertainties
                    all_errors = []
                    all_uncertainties = []
                    
                    for i in range(len(test_data)):
                        x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                        samples = result['posterior'].sample((500,), x=x_obs).numpy()
                        
                        # Apply calibration
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                        
                        pred_mean = np.mean(samples_cal[:, param_idx])
                        pred_std = np.std(samples_cal[:, param_idx])
                        true_val = test_params_subset[i, param_idx]
                        
                        error = pred_mean - true_val
                        all_errors.append(error)
                        all_uncertainties.append(pred_std)
                    
                    all_errors = np.array(all_errors)
                    all_uncertainties = np.array(all_uncertainties)
                    
                    # Plot
                    scatter = ax.scatter(all_uncertainties, all_errors, 
                                       c=np.abs(all_errors), cmap='viridis',
                                       alpha=0.6, s=30)
                    
                    ax.axhline(0, color='red', linestyle='--', alpha=0.5)
                    
                    x_range = np.linspace(all_uncertainties.min(), 
                                        all_uncertainties.max(), 100)
                    ax.fill_between(x_range, -x_range, x_range, 
                                   alpha=0.2, color='gray', label='±1σ')
                    ax.fill_between(x_range, -2*x_range, 2*x_range, 
                                   alpha=0.1, color='gray', label='±2σ')
                    
                    # Calculate z-scores
                    z_scores = all_errors / (all_uncertainties + 1e-10)
                    z_rms = np.sqrt(np.mean(z_scores**2))
                    
                    ax.set_xlabel(f'Posterior Std for {param_names[param_idx]}')
                    ax.set_ylabel(f'Error in {param_names[param_idx]}')
                    ax.set_title(f'{method_key}\nZ-RMS={z_rms:.2f}')
                    ax.grid(True, alpha=0.3)
                    
                    if param_idx == 0 and idx == 0:
                        ax.legend(loc='upper left', fontsize=8)
        
        plt.suptitle('Error vs Uncertainty Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()

# %% [markdown]
# ## Run the Complete Test Pipeline

# %%
# Initialize the tester with your chosen methods
tester = SBIConvergenceTester(
    inference_methods=INFERENCE_METHODS,
    analysis_methods=ANALYSIS_METHODS,
    n_train=N_TRAIN,
    n_test=N_TEST,
    n_calibration=N_CALIBRATION
)

# Load data
tester.load_data(SIM_ROOT, SEED)

# Setup LIMference
tester.setup_limference(OUTPUT_ROOT)

# Train models
tester.train_models(test_obs_idx=0)

# Calibrate posteriors
tester.calibrate_posteriors(target_coverage=0.68)

# Analyze convergence
tester.analyze_convergence()

# Test recovery
tester.test_recovery()

# Create visualizations
tester.plot_results()

# Generate final report
tester.generate_report()
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)
from limference.convergence import ConvergenceAnalyzer

class SBIConvergenceTester:
    """
    General class for testing any SBI method with convergence analysis.
    Works with any set of parameters specified by the user.
    """
    
    def __init__(
        self,
        param_names: List[str],  # Specify which parameters to use
        param_labels: Optional[Dict[str, str]] = None,  # LaTeX labels for plotting
        inference_methods: List[str] = ["NPE"],
        analysis_methods: List[str] = ["pdf"],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        """
        Initialize tester with flexible parameter specification.
        
        Args:
            param_names: List of parameter names to constrain (e.g., ['omega_m', 'h'])
            param_labels: Optional dict mapping param names to LaTeX labels for plots
            inference_methods: SBI methods to use
            analysis_methods: Data analysis methods (pdf, power_spectrum, etc.)
            n_train: Number of training samples
            n_test: Number of test samples
            n_calibration: Number of calibration samples
            config_kwargs: Additional configuration options
        """
        self.param_names = param_names
        self.n_params = len(param_names)
        
        # Set up parameter labels for plotting
        if param_labels is None:
            self.param_labels = self._create_default_labels(param_names)
        else:
            self.param_labels = param_labels
            # Fill in any missing labels with defaults
            for name in param_names:
                if name not in self.param_labels:
                    self.param_labels[name] = self._create_default_labels([name])[name]
        
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
    
    def _create_default_labels(self, names: List[str]) -> Dict[str, str]:
        """Create default LaTeX labels for common parameter names."""
        default_labels = {
            'omega_m': r'$\Omega_m$',
            'Omega_m': r'$\Omega_m$',
            'h': r'$h$',
            'H0': r'$H_0$',
            'sigma8': r'$\sigma_8$',
            'sigma_8': r'$\sigma_8$',
            'ns': r'$n_s$',
            'n_s': r'$n_s$',
            'omega_b': r'$\Omega_b$',
            'Omega_b': r'$\Omega_b$',
            'omega_c': r'$\Omega_c$',
            'Omega_c': r'$\Omega_c$',
            'A_s': r'$A_s$',
            'tau': r'$\tau$',
            'w0': r'$w_0$',
            'wa': r'$w_a$',
            'mnu': r'$\sum m_\nu$',
            'omh2': r'$\Omega_m h^2$',
            'ombh2': r'$\Omega_b h^2$',
            'omch2': r'$\Omega_c h^2$',
            'a_off': r'$A_{\mathrm{off}}$',
            'b_off': r'$B_{\mathrm{off}}$',
            'A_off': r'$A_{\mathrm{off}}$',
            'B_off': r'$B_{\mathrm{off}}$'
        }
        
        labels = {}
        for name in names:
            if name in default_labels:
                labels[name] = default_labels[name]
            else:
                # Create a simple label for unknown parameters
                labels[name] = name.replace('_', ' ')
        
        return labels
    
    def load_data(self, sim_root: str, seed: int = 1, param_file_keys: Optional[Dict[str, str]] = None):
        """
        Load simulation data with flexible parameter extraction.
        
        Args:
            sim_root: Root directory for simulation data
            seed: Random seed for data files
            param_file_keys: Optional mapping from param_names to keys in the data file
                            e.g., {'omega_m': 'Omega_m', 'h': 'hubble'}
        """
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Extract specified parameters
        self.params_array = self._extract_parameters(self.params_sims, param_file_keys)
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        # Print summary
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters being constrained: {', '.join(self.param_names)}")
        print(f"Parameter ranges:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def _extract_parameters(self, params_sims: Any, param_file_keys: Optional[Dict[str, str]] = None) -> np.ndarray:
        """
        Extract specified parameters from simulation data.
        
        Args:
            params_sims: Parameter data from file
            param_file_keys: Mapping from desired param names to file keys
            
        Returns:
            Array of shape (n_sims, n_params) with extracted parameters
        """
        if param_file_keys is None:
            param_file_keys = {}
        
        params_list = []
        
        for sim_params in params_sims:
            param_values = []
            
            for param_name in self.param_names:
                # Check if there's a mapping for this parameter
                file_key = param_file_keys.get(param_name, param_name)
                
                # Try to extract the parameter with various possible keys
                value = None
                
                # Try direct access
                if file_key in sim_params:
                    value = sim_params[file_key]
                # Try common variations
                elif param_name in sim_params:
                    value = sim_params[param_name]
                # Try with underscores vs no underscores
                elif param_name.replace('_', '') in sim_params:
                    value = sim_params[param_name.replace('_', '')]
                elif param_name.replace('_', '8') in sim_params:  # for sigma_8 -> sigma8
                    value = sim_params[param_name.replace('_', '8')]
                # Try capitalized versions
                elif param_name.capitalize() in sim_params:
                    value = sim_params[param_name.capitalize()]
                elif 'Omega_' + param_name.split('_')[-1] in sim_params:  # omega_m -> Omega_m
                    value = sim_params['Omega_' + param_name.split('_')[-1]]
                else:
                    raise KeyError(f"Could not find parameter '{param_name}' (or '{file_key}') in simulation data. "
                                 f"Available keys: {list(sim_params.keys())}")
                
                param_values.append(value)
            
            params_list.append(param_values)
        
        return np.array(params_list)
    
    def setup_limference(self, output_root: str, **extra_config):
        """Initialize LIMference with configuration for specified parameters."""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Create configuration
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "device": "cuda" if torch.cuda.is_available() else "cpu"
        }
        default_config.update(self.config_kwargs)
        default_config.update(extra_config)  # Allow runtime config updates
        
        self.config = LIMConfig(**default_config)
        
        # Create parameter configuration
        param_ranges = {}
        for i, name in enumerate(self.param_names):
            param_ranges[name] = (
                float(self.params_array[:, i].min()),
                float(self.params_array[:, i].max())
            )
        
        self.param_config = ParameterConfig(
            param_names=self.param_names,
            param_ranges=param_ranges,
            param_labels=self.param_labels
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        print(f"Parameters to constrain: {', '.join(self.param_names)}")
    
    def train_models(self, test_obs_idx: int = 0, training_kwargs: Optional[Dict] = None):
        """Train all specified SBI models."""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        for i, name in enumerate(self.param_names):
            print(f"  {name} = {self.test_true_params[i]:.3f}")
        
        # Create true params dict
        true_params_dict = {
            name: float(self.test_true_params[i]) 
            for i, name in enumerate(self.param_names)
        }
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params=true_params_dict,
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            training_kwargs=training_kwargs,
            verbose=True
        )
        
        self.models_trained = True
        self.lim.results = self.lim_results  # Store for later access
    
    def calibrate_posteriors(self, target_coverage: float = 0.68):
        """Calibrate all posteriors using temperature scaling"""
        if not self.models_trained:
            raise ValueError("Train models first using train_models()")
        
        print("\n" + "="*50)
        print("CALIBRATING POSTERIORS")
        print("="*50)
        
        def calibrate_temperature(posterior, cal_data, cal_params, target=0.68):
            """Find optimal temperature scaling"""
            def coverage_error(temp):
                coverage = 0
                n_cal = min(30, len(cal_data))
                
                for i in range(n_cal):
                    x_obs = torch.tensor(cal_data[i], dtype=torch.float32)
                    samples = posterior.sample((300,), x=x_obs).numpy()
                    
                    mean = np.mean(samples, axis=0)
                    scaled = mean + (samples - mean) * temp
                    
                    q16, q84 = np.percentile(scaled, [16, 84], axis=0)
                    if np.all((cal_params[i] >= q16) & (cal_params[i] <= q84)):
                        coverage += 1
                
                return abs(coverage/n_cal - target)
            
            result = minimize_scalar(coverage_error, bounds=(0.5, 3.0), method='bounded')
            return result.x
        
        # Calibrate each method
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Determine which test data to use
                if "power" in method_key:
                    cal_data = self.test_pk[:self.n_calibration]
                elif "pdf" in method_key:
                    cal_data = self.test_pdf[:self.n_calibration]
                else:
                    continue
                
                cal_params = self.test_params[:self.n_calibration]
                
                # Find optimal temperature
                temp = calibrate_temperature(
                    result['posterior'], cal_data, cal_params, target_coverage
                )
                
                print(f"{method_key}: temperature = {temp:.3f}")
                result['calibration_temp'] = temp
                
                # Interpretation
                if temp > 1.2:
                    print(f"  → Model was overconfident, widening by {(temp-1)*100:.1f}%")
                elif temp < 0.8:
                    print(f"  → Model was underconfident, narrowing by {(1-temp)*100:.1f}%")
                else:
                    print(f"  → Model was well-calibrated")
    
    def analyze_convergence(self):
        """Analyze training convergence for all methods"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        analyzer = ConvergenceAnalyzer(self.config, self.param_config)
        
        # Determine plot layout
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        if n_methods == 1:
            axes = [axes]
        elif n_rows == 1:
            axes = axes.reshape(1, -1)
        
        convergence_results = {}
        plot_idx = 0
        
        for method_key, result in self.lim.results.items():
            if "error" not in result and 'loss_history' in result:
                print(f"\nAnalyzing {method_key}...")
                
                # Extract metrics
                metrics = self._extract_metrics(result['loss_history'])
                convergence_results[method_key] = metrics
                
                # Plot
                if plot_idx < n_rows * n_cols:
                    row = plot_idx // n_cols
                    col = plot_idx % n_cols
                    ax = axes[row, col] if n_rows > 1 else axes[col]
                    
                    self._plot_convergence(ax, metrics, method_key)
                    plot_idx += 1
                
                # Print metrics
                self._print_convergence_metrics(metrics)
        
        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col] if n_rows > 1 else axes[col]
            ax.set_visible(False)
        
        plt.suptitle(f'Convergence Analysis: {", ".join(self.inference_methods)}', fontsize=14)
        plt.tight_layout()
        plt.show()
        
        self.convergence_results = convergence_results
    
    def analyze_sensitivity(self):
        """
        Analyze how sensitive the data (PDF/PS) is to each parameter.
        Works for any number of parameters.
        """
        if not self.data_loaded:
            print("Please load data first!")
            return
        
        # Determine what data to analyze
        if "pdf" in self.analysis_methods:
            data_to_analyze = self.pdf_sim_linear[:500]
            data_name = "PDF"
        elif "power_spectrum" in self.analysis_methods:
            data_to_analyze = self.pk_sim[:500]
            data_name = "Power Spectrum"
        else:
            print("No recognized analysis method for sensitivity analysis")
            return
        
        n_params = len(self.param_names)
        
        # Create appropriate subplot layout
        n_cols = min(3, n_params + 1)
        n_rows = (n_params + n_cols) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        axes = axes.flatten() if n_rows * n_cols > 1 else [axes]
        
        # Plot 1: PCA colored by each parameter
        ax_idx = 0
        
        # Standardize data for PCA
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data_to_analyze)
        
        # Apply PCA
        pca = PCA(n_components=2)
        data_pca = pca.fit_transform(data_scaled)
        
        for param_idx, param_name in enumerate(self.param_names):
            if ax_idx < len(axes):
                ax = axes[ax_idx]
                scatter = ax.scatter(data_pca[:, 0], data_pca[:, 1], 
                                   c=self.params_array[:500, param_idx], 
                                   cmap='viridis', alpha=0.6, s=20)
                plt.colorbar(scatter, ax=ax, label=self.param_labels[param_name])
                ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} var)')
                ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} var)')
                ax.set_title(f'PCA colored by {param_name}')
                ax.grid(True, alpha=0.3)
                ax_idx += 1
        
        # Plot 2: Correlation analysis
        if ax_idx < len(axes):
            ax = axes[ax_idx]
            
            all_correlations = {}
            for param_idx, param_name in enumerate(self.param_names):
                correlations = []
                for i in range(data_to_analyze.shape[1]):
                    if np.std(data_to_analyze[:, i]) > 1e-10:
                        corr = np.corrcoef(data_to_analyze[:, i], 
                                          self.params_array[:500, param_idx])[0, 1]
                    else:
                        corr = 0
                    correlations.append(corr)
                
                all_correlations[param_name] = correlations
                ax.plot(correlations, label=f'{data_name}-{param_name}', linewidth=2, alpha=0.7)
            
            ax.axhline(0, color='gray', linestyle='--', alpha=0.5)
            ax.axhline(0.1, color='gray', linestyle=':', alpha=0.3)
            ax.axhline(-0.1, color='gray', linestyle=':', alpha=0.3)
            ax.set_xlabel(f'{data_name} bin index')
            ax.set_ylabel('Correlation coefficient')
            ax.set_title(f'Parameter-{data_name} Correlations')
            ax.legend()
            ax.grid(True, alpha=0.3)
            ax.set_ylim(-0.5, 0.5)
            ax_idx += 1
        
        # Hide unused subplots
        for idx in range(ax_idx, len(axes)):
            axes[idx].set_visible(False)
        
        plt.suptitle(f'{data_name} Sensitivity to Parameters', fontsize=16)
        plt.tight_layout()
        plt.show()
        
        # Print analysis summary
        print("\n" + "="*60)
        print(f"{data_name.upper()} INFORMATION CONTENT ANALYSIS")
        print("="*60)
        
        print(f"\nParameter ranges in data:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
        
        print(f"\nCorrelation with {data_name} bins:")
        max_corrs = {}
        for param_name in self.param_names:
            max_corr = np.max(np.abs(all_correlations[param_name]))
            mean_corr = np.mean(np.abs(all_correlations[param_name]))
            max_corrs[param_name] = max_corr
            print(f"  {param_name}:")
            print(f"    Maximum |correlation|: {max_corr:.3f}")
            print(f"    Mean |correlation|:    {mean_corr:.3f}")
        
        # Find the most and least informative parameters
        sorted_params = sorted(max_corrs.items(), key=lambda x: x[1], reverse=True)
        
        print(f"\nParameter informativeness ranking:")
        for i, (param, corr) in enumerate(sorted_params, 1):
            print(f"  {i}. {param}: {corr:.3f}")
        
        # Check PCA correlations
        print(f"\nPCA correlations:")
        for param_idx, param_name in enumerate(self.param_names):
            pc1_corr = np.corrcoef(data_pca[:, 0], self.params_array[:500, param_idx])[0, 1]
            pc2_corr = np.corrcoef(data_pca[:, 1], self.params_array[:500, param_idx])[0, 1]
            print(f"  {param_name}:")
            print(f"    PC1: {pc1_corr:.3f}")
            print(f"    PC2: {pc2_corr:.3f}")
        
        # Diagnosis
        print("\n" + "="*60)
        print("DIAGNOSIS:")
        
        poorly_constrained = [p for p, c in max_corrs.items() if c < 0.1]
        if poorly_constrained:
            print(f"❌ {data_name} contains very little information about: {', '.join(poorly_constrained)}")
            print("   These parameters will be poorly constrained")
        
        well_constrained = [p for p, c in max_corrs.items() if c > 0.3]
        if well_constrained:
            print(f"✓ {data_name} contains good information about: {', '.join(well_constrained)}")
        
        print("\nRECOMMENDATIONS:")
        if poorly_constrained:
            print(f"1. Consider using a different summary statistic for {', '.join(poorly_constrained)}")
            print("2. Try combined analysis methods (PS + PDF + field-level)")
            print("3. Check if your simulations properly vary these parameters")
    
    def test_recovery(self):
        """Test posterior recovery on test set"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        print("\n" + "="*50)
        print("TESTING POSTERIOR RECOVERY")
        print("="*50)
        
        recovery_results = {}
        
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                print(f"\n{method_key}:")
                
                # Get appropriate test data
                if "power" in method_key:
                    test_data = self.test_pk[self.n_calibration:]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[self.n_calibration:]
                else:
                    continue
                
                test_params_subset = self.test_params[self.n_calibration:]
                
                # Test recovery
                metrics = self._test_single_method_recovery(
                    result, test_data, test_params_subset
                )
                recovery_results[method_key] = metrics
                
                # Print results
                self._print_recovery_metrics(metrics)
        
        self.recovery_results = recovery_results
    
    def plot_parameter_recovery(self):
        """Plot parameter recovery for any number of parameters."""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_params = len(self.param_names)
        
        fig, axes = plt.subplots(n_methods, n_params, figsize=(5*n_params, 4*n_methods))
        
        if n_methods == 1 and n_params == 1:
            axes = np.array([[axes]])
        elif n_methods == 1:
            axes = axes.reshape(1, -1)
        elif n_params == 1:
            axes = axes.reshape(-1, 1)
        
        method_idx = 0
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                
                test_params_subset = self.test_params[50:150]
                temp = result.get('calibration_temp', 1.0)
                
                # Collect predictions
                predictions_mean = []
                predictions_16 = []
                predictions_84 = []
                true_vals = []
                n_eval = min(50, len(test_data))
                
                for i in range(n_eval):
                    x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                    samples = result['posterior'].sample((1000,), x=x_obs).numpy()
                    
                    # Apply calibration if available
                    if 'calibration_temp' in result:
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                    else:
                        samples_cal = samples
                    
                    predictions_mean.append(np.mean(samples_cal, axis=0))
                    predictions_16.append(np.percentile(samples_cal, 16, axis=0))
                    predictions_84.append(np.percentile(samples_cal, 84, axis=0))
                    true_vals.append(test_params_subset[i])
                
                predictions_mean = np.array(predictions_mean)
                predictions_16 = np.array(predictions_16)
                predictions_84 = np.array(predictions_84)
                true_vals = np.array(true_vals)
                
                # Plot for each parameter
                for param_idx, param_name in enumerate(self.param_names):
                    ax = axes[method_idx, param_idx] if n_methods > 1 or n_params > 1 else axes
                    
                    # Error bars
                    yerr_lower = predictions_mean[:, param_idx] - predictions_16[:, param_idx]
                    yerr_upper = predictions_84[:, param_idx] - predictions_mean[:, param_idx]
                    yerr = np.array([yerr_lower, yerr_upper])
                    
                    ax.errorbar(true_vals[:, param_idx], 
                              predictions_mean[:, param_idx],
                              yerr=yerr,
                              fmt='o', alpha=0.6, markersize=4,
                              elinewidth=0.5, capsize=2, capthick=0.5)
                    
                    # Perfect recovery line
                    lims = [true_vals[:, param_idx].min(), true_vals[:, param_idx].max()]
                    ax.plot(lims, lims, 'r--', alpha=0.7, linewidth=2)
                    
                    # Calculate metrics
                    from scipy.stats import pearsonr
                    r2 = pearsonr(true_vals[:, param_idx], 
                                predictions_mean[:, param_idx])[0]**2
                    rmse = np.sqrt(np.mean((true_vals[:, param_idx] - 
                                          predictions_mean[:, param_idx])**2))
                    
                    ax.set_xlabel(f'True {self.param_labels[param_name]}')
                    ax.set_ylabel(f'Predicted {self.param_labels[param_name]}')
                    ax.set_title(f'{method_key}\nR²={r2:.3f}, RMSE={rmse:.4f}')
                    ax.grid(True, alpha=0.3)
                
                method_idx += 1
        
        plt.suptitle('Parameter Recovery Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def generate_report(self):
        """Generate final summary report"""
        print("\n" + "="*70)
        print(f"SBI METHODS CONVERGENCE TEST SUMMARY")
        print(f"Methods tested: {', '.join(self.inference_methods)}")
        print(f"Analysis types: {', '.join(self.analysis_methods)}")
        print(f"Parameters: {', '.join(self.param_names)}")
        print("="*70)
        
        for method_key in self.lim.results:
            if "error" not in self.lim.results[method_key]:
                print(f"\n{method_key}:")
                print("-"*40)
                
                # Convergence info
                if hasattr(self, 'convergence_results') and method_key in self.convergence_results:
                    conv = self.convergence_results[method_key]
                    print(f"Training:")
                    print(f"  Epochs: {conv.get('n_epochs', 'N/A')}")
                    print(f"  Converged: {conv.get('converged', 'Unknown')}")
                    if conv.get('overfitting'):
                        print(f"  ⚠️ Overfitting detected")
                
                # Calibration info
                result = self.lim.results[method_key]
                if 'calibration_temp' in result:
                    print(f"Calibration:")
                    print(f"  Temperature: {result['calibration_temp']:.3f}")
                
                # Recovery performance
                if hasattr(self, 'recovery_results') and method_key in self.recovery_results:
                    rec = self.recovery_results[method_key]
                    print(f"Performance:")
                    print(f"  RMSE: {rec['mean_rmse']:.4f}")
                    print(f"  68% coverage: {rec['coverage_68_rate']:.1%}")
                    print(f"  95% coverage: {rec['coverage_95_rate']:.1%}")
                    
                    if rec['well_calibrated_68'] and rec['well_calibrated_95']:
                        print("  ✅ Well calibrated")
                    else:
                        print("  ⚠️ Needs adjustment")
    
    # Helper methods
    def _extract_metrics(self, loss_history):
        """Extract metrics from loss history"""
        metrics = {}
        
        # Handle different possible formats
        train_losses = loss_history.get('training_loss', 
                                       loss_history.get('train_log_probs', []))
        val_losses = loss_history.get('validation_loss',
                                     loss_history.get('validation_log_probs', []))
        
        if isinstance(train_losses, np.ndarray):
            train_losses = train_losses.tolist()
        if isinstance(val_losses, np.ndarray):
            val_losses = val_losses.tolist()
            
        metrics['train_losses'] = train_losses
        metrics['val_losses'] = val_losses
        
        # Get epochs
        epochs = loss_history.get('epochs_trained', len(train_losses))
        if isinstance(epochs, (list, np.ndarray)):
            epochs = int(epochs[0] if len(epochs) > 0 else len(train_losses))
        metrics['n_epochs'] = epochs
        
        # Check convergence
        if len(train_losses) > 10:
            window = len(train_losses) // 10
            recent = train_losses[-window:]
            metrics['converged'] = np.std(recent) < 0.01 * abs(np.mean(recent))
        else:
            metrics['converged'] = False
        
        # Check overfitting
        if train_losses and val_losses:
            gap = val_losses[-1] - train_losses[-1]
            metrics['overfitting'] = gap > 0.5
        else:
            metrics['overfitting'] = False
            
        return metrics
    
    def _plot_convergence(self, ax, metrics, method_key):
        """Plot convergence curves"""
        if metrics['train_losses']:
            ax.plot(range(1, len(metrics['train_losses'])+1), 
                   metrics['train_losses'], 'b-', label='Training', linewidth=2)
        
        if metrics['val_losses']:
            ax.plot(range(1, len(metrics['val_losses'])+1), 
                   metrics['val_losses'], 'r-', label='Validation', linewidth=2)
        
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(method_key.replace('_', ' '))
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    def _print_convergence_metrics(self, metrics):
        """Print convergence metrics"""
        print(f"  Epochs: {metrics.get('n_epochs', 'N/A')}")
        print(f"  Converged: {metrics.get('converged', 'Unknown')}")
        if metrics.get('overfitting'):
            print(f"  ⚠️ Potential overfitting")
    
    def _test_single_method_recovery(self, result, test_data, test_params):
        """Test recovery for a single method"""
        n_test_samples = min(50, len(test_data))
        coverage_68 = 0
        coverage_95 = 0
        errors = []
        
        temp = result.get('calibration_temp', 1.0)
        
        for i in range(n_test_samples):
            x_obs = torch.tensor(test_data[i], dtype=torch.float32)
            samples = result['posterior'].sample((1000,), x=x_obs).numpy()
            
            # Apply temperature scaling
            mean = np.mean(samples, axis=0)
            samples = mean + (samples - mean) * temp
            
            # Calculate statistics
            post_mean = np.mean(samples, axis=0)
            errors.append(post_mean - test_params[i])
            
            # Check coverage
            q16, q84 = np.percentile(samples, [16, 84], axis=0)
            q2_5, q97_5 = np.percentile(samples, [2.5, 97.5], axis=0)
            
            if np.all((test_params[i] >= q16) & (test_params[i] <= q84)):
                coverage_68 += 1
            if np.all((test_params[i] >= q2_5) & (test_params[i] <= q97_5)):
                coverage_95 += 1
        
        errors = np.array(errors)
        
        return {
            'coverage_68_rate': coverage_68 / n_test_samples,
            'coverage_95_rate': coverage_95 / n_test_samples,
            'mean_rmse': np.sqrt(np.mean(errors**2)),
            'mean_bias': np.mean(np.abs(errors)),
            'well_calibrated_68': abs(coverage_68/n_test_samples - 0.68) < 0.1,
            'well_calibrated_95': abs(coverage_95/n_test_samples - 0.95) < 0.1,
            'calibration_temp': temp
        }
    
    def _print_recovery_metrics(self, metrics):
        """Print recovery metrics"""
        print(f"  Temperature: {metrics['calibration_temp']:.3f}")
        print(f"  68% coverage: {metrics['coverage_68_rate']:.1%}")
        print(f"  95% coverage: {metrics['coverage_95_rate']:.1%}")
        print(f"  RMSE: {metrics['mean_rmse']:.4f}")
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any

from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)
from limference.convergence import ConvergenceAnalyzer

class SBIConvergenceTester:
    """
    General class for testing any SBI method with convergence analysis.
    Works with any set of parameters specified by the user.
    """
    
    def __init__(
        self,
        param_names: List[str],  # Specify which parameters to use
        param_labels: Optional[Dict[str, str]] = None,  # LaTeX labels for plotting
        inference_methods: List[str] = ["NPE"],
        analysis_methods: List[str] = ["pdf"],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        """
        Initialize tester with flexible parameter specification.
        
        Args:
            param_names: List of parameter names to constrain (e.g., ['omega_m', 'h'])
            param_labels: Optional dict mapping param names to LaTeX labels for plots
            inference_methods: SBI methods to use
            analysis_methods: Data analysis methods (pdf, power_spectrum, etc.)
            n_train: Number of training samples
            n_test: Number of test samples
            n_calibration: Number of calibration samples
            config_kwargs: Additional configuration options
        """
        self.param_names = param_names
        self.n_params = len(param_names)
        
        # Set up parameter labels for plotting
        if param_labels is None:
            self.param_labels = self._create_default_labels(param_names)
        else:
            self.param_labels = param_labels
            # Fill in any missing labels with defaults
            for name in param_names:
                if name not in self.param_labels:
                    self.param_labels[name] = self._create_default_labels([name])[name]
        
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
    
    def _create_default_labels(self, names: List[str]) -> Dict[str, str]:
        """Create default LaTeX labels for common parameter names."""
        default_labels = {
            'omega_m': r'$\Omega_m$',
            'Omega_m': r'$\Omega_m$',
            'h': r'$h$',
            'H0': r'$H_0$',
            'sigma8': r'$\sigma_8$',
            'sigma_8': r'$\sigma_8$',
            'ns': r'$n_s$',
            'n_s': r'$n_s$',
            'omega_b': r'$\Omega_b$',
            'Omega_b': r'$\Omega_b$',
            'omega_c': r'$\Omega_c$',
            'Omega_c': r'$\Omega_c$',
            'A_s': r'$A_s$',
            'tau': r'$\tau$',
            'w0': r'$w_0$',
            'wa': r'$w_a$',
            'mnu': r'$\sum m_\nu$',
            'omh2': r'$\Omega_m h^2$',
            'ombh2': r'$\Omega_b h^2$',
            'omch2': r'$\Omega_c h^2$'
        }
        
        labels = {}
        for name in names:
            if name in default_labels:
                labels[name] = default_labels[name]
            else:
                # Create a simple label for unknown parameters
                labels[name] = name.replace('_', ' ')
        
        return labels
    
    def load_data(self, sim_root: str, seed: int = 1, param_file_keys: Optional[Dict[str, str]] = None):
        """
        Load simulation data with flexible parameter extraction.
        
        Args:
            sim_root: Root directory for simulation data
            seed: Random seed for data files
            param_file_keys: Optional mapping from param_names to keys in the data file
                            e.g., {'omega_m': 'Omega_m', 'h': 'hubble'}
        """
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Extract specified parameters
        self.params_array = self._extract_parameters(self.params_sims, param_file_keys)
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        # Print summary
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters being constrained: {', '.join(self.param_names)}")
        print(f"Parameter ranges:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def _extract_parameters(self, params_sims: Any, param_file_keys: Optional[Dict[str, str]] = None) -> np.ndarray:
        """
        Extract specified parameters from simulation data.
        
        Args:
            params_sims: Parameter data from file
            param_file_keys: Mapping from desired param names to file keys
            
        Returns:
            Array of shape (n_sims, n_params) with extracted parameters
        """
        if param_file_keys is None:
            param_file_keys = {}
        
        params_list = []
        
        for sim_params in params_sims:
            param_values = []
            
            for param_name in self.param_names:
                # Check if there's a mapping for this parameter
                file_key = param_file_keys.get(param_name, param_name)
                
                # Try to extract the parameter with various possible keys
                value = None
                
                # Try direct access
                if file_key in sim_params:
                    value = sim_params[file_key]
                # Try common variations
                elif param_name in sim_params:
                    value = sim_params[param_name]
                # Try with underscores vs no underscores
                elif param_name.replace('_', '') in sim_params:
                    value = sim_params[param_name.replace('_', '')]
                elif param_name.replace('_', '8') in sim_params:  # for sigma_8 -> sigma8
                    value = sim_params[param_name.replace('_', '8')]
                # Try capitalized versions
                elif param_name.capitalize() in sim_params:
                    value = sim_params[param_name.capitalize()]
                elif 'Omega_' + param_name.split('_')[-1] in sim_params:  # omega_m -> Omega_m
                    value = sim_params['Omega_' + param_name.split('_')[-1]]
                else:
                    raise KeyError(f"Could not find parameter '{param_name}' (or '{file_key}') in simulation data. "
                                 f"Available keys: {list(sim_params.keys())}")
                
                param_values.append(value)
            
            params_list.append(param_values)
        
        return np.array(params_list)
    
    def setup_limference(self, output_root: str):
        """Initialize LIMference with configuration for specified parameters."""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Create configuration
        from limference import create_training_kwargs
        
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "device": "cuda" if torch.cuda.is_available() else "cpu"
        }
        default_config.update(self.config_kwargs)
        
        self.config = LIMConfig(**default_config)
        
        # Create parameter configuration
        param_ranges = {}
        for i, name in enumerate(self.param_names):
            param_ranges[name] = (
                float(self.params_array[:, i].min()),
                float(self.params_array[:, i].max())
            )
        
        self.param_config = ParameterConfig(
            param_names=self.param_names,
            param_ranges=param_ranges,
            param_labels=self.param_labels
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        print(f"Parameters to constrain: {', '.join(self.param_names)}")
    
    def train_models(self, test_obs_idx: int = 0, training_kwargs: Optional[Dict] = None):
        """Train all specified SBI models."""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        for i, name in enumerate(self.param_names):
            print(f"  {name} = {self.test_true_params[i]:.3f}")
        
        # Create true params dict
        true_params_dict = {
            name: float(self.test_true_params[i]) 
            for i, name in enumerate(self.param_names)
        }
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params=true_params_dict,
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            training_kwargs=training_kwargs,
            verbose=True
        )
        
        self.models_trained = True
        self.lim.results = self.lim_results  # Store for later access
    
    def analyze_sensitivity(self):
        """
        Analyze how sensitive the data (PDF/PS) is to each parameter.
        Works for any number of parameters.
        """
        if not self.data_loaded:
            print("Please load data first!")
            return
        
        from sklearn.decomposition import PCA
        from sklearn.preprocessing import StandardScaler
        
        # Determine what data to analyze
        if "pdf" in self.analysis_methods:
            data_to_analyze = self.pdf_sim_linear[:500]
            data_name = "PDF"
        elif "power_spectrum" in self.analysis_methods:
            data_to_analyze = self.pk_sim[:500]
            data_name = "Power Spectrum"
        else:
            print("No recognized analysis method for sensitivity analysis")
            return
        
        n_params = len(self.param_names)
        
        # Create appropriate subplot layout
        n_cols = min(3, n_params + 1)
        n_rows = (n_params + n_cols) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        axes = axes.flatten() if n_rows * n_cols > 1 else [axes]
        
        # Plot 1: PCA colored by each parameter
        ax_idx = 0
        
        # Standardize data for PCA
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data_to_analyze)
        
        # Apply PCA
        pca = PCA(n_components=2)
        data_pca = pca.fit_transform(data_scaled)
        
        for param_idx, param_name in enumerate(self.param_names):
            if ax_idx < len(axes):
                ax = axes[ax_idx]
                scatter = ax.scatter(data_pca[:, 0], data_pca[:, 1], 
                                   c=self.params_array[:500, param_idx], 
                                   cmap='viridis', alpha=0.6, s=20)
                plt.colorbar(scatter, ax=ax, label=self.param_labels[param_name])
                ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} var)')
                ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} var)')
                ax.set_title(f'PCA colored by {param_name}')
                ax.grid(True, alpha=0.3)
                ax_idx += 1
        
        # Plot 2: Correlation analysis
        if ax_idx < len(axes):
            ax = axes[ax_idx]
            
            all_correlations = {}
            for param_idx, param_name in enumerate(self.param_names):
                correlations = []
                for i in range(data_to_analyze.shape[1]):
                    if np.std(data_to_analyze[:, i]) > 1e-10:
                        corr = np.corrcoef(data_to_analyze[:, i], 
                                          self.params_array[:500, param_idx])[0, 1]
                    else:
                        corr = 0
                    correlations.append(corr)
                
                all_correlations[param_name] = correlations
                ax.plot(correlations, label=f'{data_name}-{param_name}', linewidth=2, alpha=0.7)
            
            ax.axhline(0, color='gray', linestyle='--', alpha=0.5)
            ax.axhline(0.1, color='gray', linestyle=':', alpha=0.3)
            ax.axhline(-0.1, color='gray', linestyle=':', alpha=0.3)
            ax.set_xlabel(f'{data_name} bin index')
            ax.set_ylabel('Correlation coefficient')
            ax.set_title(f'Parameter-{data_name} Correlations')
            ax.legend()
            ax.grid(True, alpha=0.3)
            ax.set_ylim(-0.5, 0.5)
            ax_idx += 1
        
        # Hide unused subplots
        for idx in range(ax_idx, len(axes)):
            axes[idx].set_visible(False)
        
        plt.suptitle(f'{data_name} Sensitivity to Parameters', fontsize=16)
        plt.tight_layout()
        plt.show()
        
        # Print analysis summary
        print("\n" + "="*60)
        print(f"{data_name.upper()} INFORMATION CONTENT ANALYSIS")
        print("="*60)
        
        print(f"\nParameter ranges in data:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
        
        print(f"\nCorrelation with {data_name} bins:")
        max_corrs = {}
        for param_name in self.param_names:
            max_corr = np.max(np.abs(all_correlations[param_name]))
            mean_corr = np.mean(np.abs(all_correlations[param_name]))
            max_corrs[param_name] = max_corr
            print(f"  {param_name}:")
            print(f"    Maximum |correlation|: {max_corr:.3f}")
            print(f"    Mean |correlation|:    {mean_corr:.3f}")
        
        # Find the most and least informative parameters
        sorted_params = sorted(max_corrs.items(), key=lambda x: x[1], reverse=True)
        
        print(f"\nParameter informativeness ranking:")
        for i, (param, corr) in enumerate(sorted_params, 1):
            print(f"  {i}. {param}: {corr:.3f}")
        
        # Check PCA correlations
        print(f"\nPCA correlations:")
        for param_idx, param_name in enumerate(self.param_names):
            pc1_corr = np.corrcoef(data_pca[:, 0], self.params_array[:500, param_idx])[0, 1]
            pc2_corr = np.corrcoef(data_pca[:, 1], self.params_array[:500, param_idx])[0, 1]
            print(f"  {param_name}:")
            print(f"    PC1: {pc1_corr:.3f}")
            print(f"    PC2: {pc2_corr:.3f}")
        
        # Diagnosis
        print("\n" + "="*60)
        print("DIAGNOSIS:")
        
        poorly_constrained = [p for p, c in max_corrs.items() if c < 0.1]
        if poorly_constrained:
            print(f"❌ {data_name} contains very little information about: {', '.join(poorly_constrained)}")
            print("   These parameters will be poorly constrained")
        
        well_constrained = [p for p, c in max_corrs.items() if c > 0.3]
        if well_constrained:
            print(f"✓ {data_name} contains good information about: {', '.join(well_constrained)}")
        
        print("\nRECOMMENDATIONS:")
        if poorly_constrained:
            print(f"1. Consider using a different summary statistic for {', '.join(poorly_constrained)}")
            print("2. Try combined analysis methods (PS + PDF + field-level)")
            print("3. Check if your simulations properly vary these parameters")
    
    def plot_parameter_recovery(self):
        """Plot parameter recovery for any number of parameters."""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_params = len(self.param_names)
        
        fig, axes = plt.subplots(n_methods, n_params, figsize=(5*n_params, 4*n_methods))
        
        if n_methods == 1 and n_params == 1:
            axes = np.array([[axes]])
        elif n_methods == 1:
            axes = axes.reshape(1, -1)
        elif n_params == 1:
            axes = axes.reshape(-1, 1)
        
        method_idx = 0
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                
                test_params_subset = self.test_params[50:150]
                temp = result.get('calibration_temp', 1.0)
                
                # Collect predictions
                predictions_mean = []
                predictions_16 = []
                predictions_84 = []
                true_vals = []
                n_eval = min(50, len(test_data))
                
                for i in range(n_eval):
                    x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                    samples = result['posterior'].sample((1000,), x=x_obs).numpy()
                    
                    # Apply calibration if available
                    if 'calibration_temp' in result:
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                    else:
                        samples_cal = samples
                    
                    predictions_mean.append(np.mean(samples_cal, axis=0))
                    predictions_16.append(np.percentile(samples_cal, 16, axis=0))
                    predictions_84.append(np.percentile(samples_cal, 84, axis=0))
                    true_vals.append(test_params_subset[i])
                
                predictions_mean = np.array(predictions_mean)
                predictions_16 = np.array(predictions_16)
                predictions_84 = np.array(predictions_84)
                true_vals = np.array(true_vals)
                
                # Plot for each parameter
                for param_idx, param_name in enumerate(self.param_names):
                    ax = axes[method_idx, param_idx]
                    
                    # Error bars
                    yerr_lower = predictions_mean[:, param_idx] - predictions_16[:, param_idx]
                    yerr_upper = predictions_84[:, param_idx] - predictions_mean[:, param_idx]
                    yerr = np.array([yerr_lower, yerr_upper])
                    
                    ax.errorbar(true_vals[:, param_idx], 
                              predictions_mean[:, param_idx],
                              yerr=yerr,
                              fmt='o', alpha=0.6, markersize=4,
                              elinewidth=0.5, capsize=2, capthick=0.5)
                    
                    # Perfect recovery line
                    lims = [true_vals[:, param_idx].min(), true_vals[:, param_idx].max()]
                    ax.plot(lims, lims, 'r--', alpha=0.7, linewidth=2)
                    
                    # Calculate metrics
                    from scipy.stats import pearsonr
                    r2 = pearsonr(true_vals[:, param_idx], 
                                predictions_mean[:, param_idx])[0]**2
                    rmse = np.sqrt(np.mean((true_vals[:, param_idx] - 
                                          predictions_mean[:, param_idx])**2))
                    
                    ax.set_xlabel(f'True {self.param_labels[param_name]}')
                    ax.set_ylabel(f'Predicted {self.param_labels[param_name]}')
                    ax.set_title(f'{method_key}\nR²={r2:.3f}, RMSE={rmse:.4f}')
                    ax.grid(True, alpha=0.3)
                
                method_idx += 1
        
        plt.suptitle('Parameter Recovery Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()

# Additional helper methods remain the same but update param references...
```

```python
# %% [markdown]
# # General SBI Method Convergence Testing with Calibration and Uncertainty Analysis
# Works with any SBI method: SNPE, NPE, NRE, SNLE, etc.

# %% Import libraries
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any

# Import LIMference
from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)

from limference.convergence import ConvergenceAnalyzer

# %% [markdown]
# ## Configuration Section - Change These Parameters

# %%
# ============ MAIN CONFIGURATION ============
# Change these parameters to test different methods

# SBI Method Selection
INFERENCE_METHODS = [ "NPE"]  # Can be any combination of: SNPE, NPE, NRE, SNLE, SNL
ANALYSIS_METHODS = ["power_spectrum", "pdf"]  # Can include: power_spectrum, pdf, field_level

# Data Configuration
N_TRAIN = 2000  # Number of training samples
N_TEST = 500    # Number of test samples
N_CALIBRATION = 50  # Number of samples for temperature calibration

# Training Configuration
MAX_EPOCHS = 200
STOP_AFTER_EPOCHS = 20
BATCH_SIZE = 50
LEARNING_RATE = 5e-4
VALIDATION_FRACTION = 0.2
NOISE_REDUCTION_FACTOR = 50.0

# File paths
SEED = 1
SIM_ROOT = f"/Users/anirbanroy/Desktop/product_with_noise_{SEED}/"
OUTPUT_ROOT = "/Users/anirbanroy/Desktop/output_small_z_3.600_sorted"

# Device
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# %% [markdown]
# ## 1. Load and Prepare Data

# %%
class SBIConvergenceTester:
    """General class for testing any SBI method with convergence analysis"""
    
    def __init__(
        self,
        inference_methods: List[str],
        analysis_methods: List[str],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
        
    def load_data(self, sim_root: str, seed: int = 1):
        """Load simulation data"""
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Convert parameters
        self.params_array = np.array([[d['sigma8'], d['omega_m']] for d in self.params_sims])
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters range:")
        print(f"  sigma8: [{self.params_array[:, 0].min():.3f}, {self.params_array[:, 0].max():.3f}]")
        print(f"  omega_m: [{self.params_array[:, 1].min():.3f}, {self.params_array[:, 1].max():.3f}]")
        
    def setup_limference(self, output_root: str):
        """Initialize LIMference with configuration"""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Create configuration
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "max_num_epochs": MAX_EPOCHS,
            "stop_after_epochs": STOP_AFTER_EPOCHS,
            "training_batch_size": BATCH_SIZE,
            "learning_rate": LEARNING_RATE,
            "validation_fraction": VALIDATION_FRACTION,
            "noise_reduction_factor": NOISE_REDUCTION_FACTOR,
            "device": DEVICE
        }
        default_config.update(self.config_kwargs)
        
        self.config = LIMConfig(**default_config)
        
        self.param_config = ParameterConfig(
            param_names=["sigma8", "omega_m"],
            param_ranges={
                "sigma8": (self.params_array[:, 0].min(), self.params_array[:, 0].max()),
                "omega_m": (self.params_array[:, 1].min(), self.params_array[:, 1].max())
            }
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        
    def train_models(self, test_obs_idx: int = 0):
        """Train all specified SBI models"""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        print(f"  sigma8 = {self.test_true_params[0]:.3f}")
        print(f"  omega_m = {self.test_true_params[1]:.3f}")
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params={
                "sigma8": self.test_true_params[0], 
                "omega_m": self.test_true_params[1]
            },
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            verbose=True
        )
        
        self.models_trained = True
        
    def calibrate_posteriors(self, target_coverage: float = 0.68):
        """Calibrate all posteriors using temperature scaling"""
        if not self.models_trained:
            raise ValueError("Train models first using train_models()")
        
        print("\n" + "="*50)
        print("CALIBRATING POSTERIORS")
        print("="*50)
        
        def calibrate_temperature(posterior, cal_data, cal_params, target=0.68):
            """Find optimal temperature scaling"""
            def coverage_error(temp):
                coverage = 0
                n_cal = min(30, len(cal_data))
                
                for i in range(n_cal):
                    x_obs = torch.tensor(cal_data[i], dtype=torch.float32)
                    samples = posterior.sample((300,), x=x_obs).numpy()
                    
                    mean = np.mean(samples, axis=0)
                    scaled = mean + (samples - mean) * temp
                    
                    q16, q84 = np.percentile(scaled, [16, 84], axis=0)
                    if np.all((cal_params[i] >= q16) & (cal_params[i] <= q84)):
                        coverage += 1
                
                return abs(coverage/n_cal - target)
            
            result = minimize_scalar(coverage_error, bounds=(0.5, 3.0), method='bounded')
            return result.x
        
        # Calibrate each method
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Determine which test data to use
                if "power" in method_key:
                    cal_data = self.test_pk[:self.n_calibration]
                elif "pdf" in method_key:
                    cal_data = self.test_pdf[:self.n_calibration]
                else:
                    continue
                
                cal_params = self.test_params[:self.n_calibration]
                
                # Find optimal temperature
                temp = calibrate_temperature(
                    result['posterior'], cal_data, cal_params, target_coverage
                )
                
                print(f"{method_key}: temperature = {temp:.3f}")
                result['calibration_temp'] = temp
                
                # Interpretation
                if temp > 1.2:
                    print(f"  → Model was overconfident, widening by {(temp-1)*100:.1f}%")
                elif temp < 0.8:
                    print(f"  → Model was underconfident, narrowing by {(1-temp)*100:.1f}%")
                else:
                    print(f"  → Model was well-calibrated")
    
    def analyze_convergence(self):
        """Analyze training convergence for all methods"""
        if not self.models_trained:
            raise ValueError("Train models first")

        analyzer = ConvergenceAnalyzer(self.config, self.param_config)

        # Determine plot layout
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols

        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))

        # Handle different subplot configurations
        if n_rows == 1 and n_cols == 1:
            # Single plot case
            axes = np.array([[axes]])
        elif n_rows == 1:
            # Single row case
            axes = axes.reshape(1, -1)
        elif n_cols == 1:
            # Single column case
            axes = axes.reshape(-1, 1)
        # Otherwise axes is already a 2D array

        convergence_results = {}
        plot_idx = 0

        for method_key, result in self.lim.results.items():
            if "error" not in result and 'loss_history' in result:
                print(f"\nAnalyzing {method_key}...")

                # Extract metrics
                metrics = self._extract_metrics(result['loss_history'])
                convergence_results[method_key] = metrics

                # Plot
                if plot_idx < n_rows * n_cols:
                    row = plot_idx // n_cols
                    col = plot_idx % n_cols
                    ax = axes[row, col]

                    self._plot_convergence(ax, metrics, method_key)
                    plot_idx += 1

                # Print metrics
                self._print_convergence_metrics(metrics)

        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]
            ax.set_visible(False)

        plt.suptitle(f'Convergence Analysis: {", ".join(self.inference_methods)}', fontsize=14)
        plt.tight_layout()
        plt.show()

        self.convergence_results = convergence_results
        
    def test_recovery(self):
        """Test posterior recovery on test set"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        print("\n" + "="*50)
        print("TESTING POSTERIOR RECOVERY")
        print("="*50)
        
        recovery_results = {}
        
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                print(f"\n{method_key}:")
                
                # Get appropriate test data
                if "power" in method_key:
                    test_data = self.test_pk[self.n_calibration:]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[self.n_calibration:]
                else:
                    continue
                
                test_params_subset = self.test_params[self.n_calibration:]
                
                # Test recovery
                metrics = self._test_single_method_recovery(
                    result, test_data, test_params_subset
                )
                recovery_results[method_key] = metrics
                
                # Print results
                self._print_recovery_metrics(metrics)
        
        self.recovery_results = recovery_results
        
    def plot_results(self):
        """Create comprehensive visualization of results"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        # 1. Coverage calibration curves
        self._plot_coverage_curves()
        
        # 2. Parameter recovery scatter plots
        self._plot_parameter_recovery()
        
        # 3. Uncertainty analysis
        self._plot_uncertainty_analysis()
        
    def generate_report(self):
        """Generate final summary report"""
        print("\n" + "="*70)
        print(f"SBI METHODS CONVERGENCE TEST SUMMARY")
        print(f"Methods tested: {', '.join(self.inference_methods)}")
        print(f"Analysis types: {', '.join(self.analysis_methods)}")
        print("="*70)
        
        for method_key in self.lim.results:
            if "error" not in self.lim.results[method_key]:
                print(f"\n{method_key}:")
                print("-"*40)
                
                # Convergence info
                if hasattr(self, 'convergence_results') and method_key in self.convergence_results:
                    conv = self.convergence_results[method_key]
                    print(f"Training:")
                    print(f"  Epochs: {conv.get('n_epochs', 'N/A')}")
                    print(f"  Converged: {conv.get('converged', 'Unknown')}")
                    if conv.get('overfitting'):
                        print(f"  ⚠️ Overfitting detected")
                
                # Calibration info
                result = self.lim.results[method_key]
                if 'calibration_temp' in result:
                    print(f"Calibration:")
                    print(f"  Temperature: {result['calibration_temp']:.3f}")
                
                # Recovery performance
                if hasattr(self, 'recovery_results') and method_key in self.recovery_results:
                    rec = self.recovery_results[method_key]
                    print(f"Performance:")
                    print(f"  RMSE: {rec['mean_rmse']:.4f}")
                    print(f"  68% coverage: {rec['coverage_68_rate']:.1%}")
                    print(f"  95% coverage: {rec['coverage_95_rate']:.1%}")
                    
                    if rec['well_calibrated_68'] and rec['well_calibrated_95']:
                        print("  ✅ Well calibrated")
                    else:
                        print("  ⚠️ Needs adjustment")
    
    # Helper methods
    def _extract_metrics(self, loss_history):
        """Extract metrics from loss history"""
        metrics = {}
        
        # Handle different possible formats
        train_losses = loss_history.get('training_loss', 
                                       loss_history.get('train_log_probs', []))
        val_losses = loss_history.get('validation_loss',
                                     loss_history.get('validation_log_probs', []))
        
        if isinstance(train_losses, np.ndarray):
            train_losses = train_losses.tolist()
        if isinstance(val_losses, np.ndarray):
            val_losses = val_losses.tolist()
            
        metrics['train_losses'] = train_losses
        metrics['val_losses'] = val_losses
        
        # Get epochs
        epochs = loss_history.get('epochs_trained', len(train_losses))
        if isinstance(epochs, (list, np.ndarray)):
            epochs = int(epochs[0] if len(epochs) > 0 else len(train_losses))
        metrics['n_epochs'] = epochs
        
        # Check convergence
        if len(train_losses) > 10:
            window = len(train_losses) // 10
            recent = train_losses[-window:]
            metrics['converged'] = np.std(recent) < 0.01 * abs(np.mean(recent))
        
        # Check overfitting
        if train_losses and val_losses:
            gap = val_losses[-1] - train_losses[-1]
            metrics['overfitting'] = gap > 0.5
            
        return metrics
    
    def _plot_convergence(self, ax, metrics, method_key):
        """Plot convergence curves"""
        if metrics['train_losses']:
            ax.plot(range(1, len(metrics['train_losses'])+1), 
                   metrics['train_losses'], 'b-', label='Training', linewidth=2)
        
        if metrics['val_losses']:
            ax.plot(range(1, len(metrics['val_losses'])+1), 
                   metrics['val_losses'], 'r-', label='Validation', linewidth=2)
        
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(method_key.replace('_', ' '))
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    def _print_convergence_metrics(self, metrics):
        """Print convergence metrics"""
        print(f"  Epochs: {metrics.get('n_epochs', 'N/A')}")
        print(f"  Converged: {metrics.get('converged', 'Unknown')}")
        if metrics.get('overfitting'):
            print(f"  ⚠️ Potential overfitting")
    
    def _test_single_method_recovery(self, result, test_data, test_params):
        """Test recovery for a single method"""
        n_test_samples = min(50, len(test_data))
        coverage_68 = 0
        coverage_95 = 0
        errors = []
        
        temp = result.get('calibration_temp', 1.0)
        
        for i in range(n_test_samples):
            x_obs = torch.tensor(test_data[i], dtype=torch.float32)
            samples = result['posterior'].sample((1000,), x=x_obs).numpy()
            
            # Apply temperature scaling
            mean = np.mean(samples, axis=0)
            samples = mean + (samples - mean) * temp
            
            # Calculate statistics
            post_mean = np.mean(samples, axis=0)
            errors.append(post_mean - test_params[i])
            
            # Check coverage
            q16, q84 = np.percentile(samples, [16, 84], axis=0)
            q2_5, q97_5 = np.percentile(samples, [2.5, 97.5], axis=0)
            
            if np.all((test_params[i] >= q16) & (test_params[i] <= q84)):
                coverage_68 += 1
            if np.all((test_params[i] >= q2_5) & (test_params[i] <= q97_5)):
                coverage_95 += 1
        
        errors = np.array(errors)
        
        return {
            'coverage_68_rate': coverage_68 / n_test_samples,
            'coverage_95_rate': coverage_95 / n_test_samples,
            'mean_rmse': np.sqrt(np.mean(errors**2)),
            'mean_bias': np.mean(np.abs(errors)),
            'well_calibrated_68': abs(coverage_68/n_test_samples - 0.68) < 0.1,
            'well_calibrated_95': abs(coverage_95/n_test_samples - 0.95) < 0.1,
            'calibration_temp': temp
        }
    
    def _print_recovery_metrics(self, metrics):
        """Print recovery metrics"""
        print(f"  Temperature: {metrics['calibration_temp']:.3f}")
        print(f"  68% coverage: {metrics['coverage_68_rate']:.1%}")
        print(f"  95% coverage: {metrics['coverage_95_rate']:.1%}")
        print(f"  RMSE: {metrics['mean_rmse']:.4f}")
       
        
        
    def _plot_coverage_curves(self):
        """Plot coverage calibration curves"""
        confidence_levels = np.array([50, 60, 70, 80, 90, 95, 99])

        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols

        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))

        # Handle different subplot configurations
        if n_rows == 1 and n_cols == 1:
            # Single plot case
            axes = np.array([[axes]])
        elif n_rows == 1:
            # Single row case
            axes = axes.reshape(1, -1)
        elif n_cols == 1:
            # Single column case
            axes = axes.reshape(-1, 1)
        # Otherwise axes is already a 2D array

        plot_idx = 0

        for method_key, result in self.lim.results.items():
            if "error" not in result and plot_idx < n_rows * n_cols:
                row = plot_idx // n_cols
                col = plot_idx % n_cols
                ax = axes[row, col]

                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue

                test_params_subset = self.test_params[50:150]

                # Calculate coverage for different confidence levels
                coverage_rates_orig = []
                coverage_rates_cal = []

                temp = result.get('calibration_temp', 1.0)

                for conf in confidence_levels:
                    coverage_orig = 0
                    coverage_cal = 0
                    n_samples = min(30, len(test_data))

                    for i in range(n_samples):
                        x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                        samples = result['posterior'].sample((500,), x=x_obs).numpy()

                        lower = (100-conf)/2
                        upper = 100-lower

                        # Original
                        ci_low_orig = np.percentile(samples, lower, axis=0)
                        ci_high_orig = np.percentile(samples, upper, axis=0)

                        # Calibrated
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                        ci_low_cal = np.percentile(samples_cal, lower, axis=0)
                        ci_high_cal = np.percentile(samples_cal, upper, axis=0)

                        if np.all((test_params_subset[i] >= ci_low_orig) & 
                                 (test_params_subset[i] <= ci_high_orig)):
                            coverage_orig += 1
                        if np.all((test_params_subset[i] >= ci_low_cal) & 
                                 (test_params_subset[i] <= ci_high_cal)):
                            coverage_cal += 1

                    coverage_rates_orig.append(100 * coverage_orig/n_samples)
                    coverage_rates_cal.append(100 * coverage_cal/n_samples)

                # Plot
                ax.plot(confidence_levels, coverage_rates_orig, 'o-', 
                       label='Original', linewidth=2, color='red')
                ax.plot(confidence_levels, coverage_rates_cal, 's-', 
                       label='Calibrated', linewidth=2, color='green')
                ax.plot([0, 100], [0, 100], 'k--', alpha=0.5, label='Perfect')

                ax.set_xlabel('Nominal Coverage (%)')
                ax.set_ylabel('Empirical Coverage (%)')
                ax.set_title(f'{method_key} (T={temp:.2f})')
                ax.legend(fontsize=8)
                ax.grid(True, alpha=0.3)
                ax.set_xlim(45, 100)
                ax.set_ylim(45, 100)

                plot_idx += 1

        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]
            ax.set_visible(False)

        plt.suptitle('Coverage Calibration: Before vs After', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def _plot_parameter_recovery(self):
        """Plot parameter recovery with error bars"""
        param_names = ['σ₈', 'Ωₘ']
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        
        fig, axes = plt.subplots(n_methods, 2, figsize=(10, 4*n_methods))
        if n_methods == 1:
            axes = axes.reshape(1, -1)
        
        for idx, (method_key, result) in enumerate(self.lim.results.items()):
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                    
                test_params_subset = self.test_params[50:150]
                temp = result.get('calibration_temp', 1.0)
                
                # Collect predictions
                predictions_mean = []
                predictions_16 = []
                predictions_84 = []
                true_vals = []
                n_eval = min(50, len(test_data))
                
                for i in range(n_eval):
                    x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                    samples = result['posterior'].sample((1000,), x=x_obs).numpy()
                    
                    # Apply calibration
                    mean = np.mean(samples, axis=0)
                    samples_cal = mean + (samples - mean) * temp
                    
                    predictions_mean.append(np.mean(samples_cal, axis=0))
                    predictions_16.append(np.percentile(samples_cal, 16, axis=0))
                    predictions_84.append(np.percentile(samples_cal, 84, axis=0))
                    true_vals.append(test_params_subset[i])
                
                predictions_mean = np.array(predictions_mean)
                predictions_16 = np.array(predictions_16)
                predictions_84 = np.array(predictions_84)
                true_vals = np.array(true_vals)
                
                for param_idx in range(2):
                    ax = axes[idx, param_idx]
                    
                    # Error bars
                    yerr_lower = predictions_mean[:, param_idx] - predictions_16[:, param_idx]
                    yerr_upper = predictions_84[:, param_idx] - predictions_mean[:, param_idx]
                    yerr = np.array([yerr_lower, yerr_upper])
                    
                    ax.errorbar(true_vals[:, param_idx], 
                               predictions_mean[:, param_idx],
                               yerr=yerr,
                               fmt='o', alpha=0.6, markersize=4,
                               elinewidth=0.5, capsize=2, capthick=0.5,
                               label='Posterior mean ± 68% CI')
                    
                    # Perfect recovery line
                    lims = [true_vals[:, param_idx].min(), true_vals[:, param_idx].max()]
                    ax.plot(lims, lims, 'r--', alpha=0.7, linewidth=2, 
                           label='Perfect recovery')
                    
                    # Calculate metrics
                    from scipy.stats import pearsonr
                    r2 = pearsonr(true_vals[:, param_idx], 
                                 predictions_mean[:, param_idx])[0]**2
                    rmse = np.sqrt(np.mean((true_vals[:, param_idx] - 
                                           predictions_mean[:, param_idx])**2))
                    
                    ax.set_xlabel(f'True {param_names[param_idx]}')
                    ax.set_ylabel(f'Predicted {param_names[param_idx]}')
                    ax.set_title(f'{method_key}\nR²={r2:.3f}, RMSE={rmse:.4f}')
                    ax.grid(True, alpha=0.3)
                    if param_idx == 0:
                        ax.legend(loc='upper left', fontsize=8)
        
        plt.suptitle('Parameter Recovery with Uncertainty', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def _plot_uncertainty_analysis(self):
        """Plot uncertainty analysis"""
        param_names = ['σ₈', 'Ωₘ']
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        
        fig, axes = plt.subplots(2, n_methods, figsize=(5*n_methods, 8))
        if n_methods == 1:
            axes = axes.reshape(-1, 1)
        
        for idx, (method_key, result) in enumerate(self.lim.results.items()):
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:100]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:100]
                else:
                    continue
                    
                test_params_subset = self.test_params[50:100]
                temp = result.get('calibration_temp', 1.0)
                
                for param_idx in range(2):
                    ax = axes[param_idx, idx]
                    
                    # Collect errors and uncertainties
                    all_errors = []
                    all_uncertainties = []
                    
                    for i in range(len(test_data)):
                        x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                        samples = result['posterior'].sample((500,), x=x_obs).numpy()
                        
                        # Apply calibration
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                        
                        pred_mean = np.mean(samples_cal[:, param_idx])
                        pred_std = np.std(samples_cal[:, param_idx])
                        true_val = test_params_subset[i, param_idx]
                        
                        error = pred_mean - true_val
                        all_errors.append(error)
                        all_uncertainties.append(pred_std)
                    
                    all_errors = np.array(all_errors)
                    all_uncertainties = np.array(all_uncertainties)
                    
                    # Plot
                    scatter = ax.scatter(all_uncertainties, all_errors, 
                                       c=np.abs(all_errors), cmap='viridis',
                                       alpha=0.6, s=30)
                    
                    ax.axhline(0, color='red', linestyle='--', alpha=0.5)
                    
                    x_range = np.linspace(all_uncertainties.min(), 
                                        all_uncertainties.max(), 100)
                    ax.fill_between(x_range, -x_range, x_range, 
                                   alpha=0.2, color='gray', label='±1σ')
                    ax.fill_between(x_range, -2*x_range, 2*x_range, 
                                   alpha=0.1, color='gray', label='±2σ')
                    
                    # Calculate z-scores
                    z_scores = all_errors / (all_uncertainties + 1e-10)
                    z_rms = np.sqrt(np.mean(z_scores**2))
                    
                    ax.set_xlabel(f'Posterior Std for {param_names[param_idx]}')
                    ax.set_ylabel(f'Error in {param_names[param_idx]}')
                    ax.set_title(f'{method_key}\nZ-RMS={z_rms:.2f}')
                    ax.grid(True, alpha=0.3)
                    
                    if param_idx == 0 and idx == 0:
                        ax.legend(loc='upper left', fontsize=8)
        
        plt.suptitle('Error vs Uncertainty Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()

# %% [markdown]
# ## Run the Complete Test Pipeline

# %%
# Initialize the tester with your chosen methods
tester = SBIConvergenceTester(
    inference_methods=INFERENCE_METHODS,
    analysis_methods=ANALYSIS_METHODS,
    n_train=N_TRAIN,
    n_test=N_TEST,
    n_calibration=N_CALIBRATION
)

# Load data
tester.load_data(SIM_ROOT, SEED)

# Setup LIMference
tester.setup_limference(OUTPUT_ROOT)

# Train models
tester.train_models(test_obs_idx=0)

# Calibrate posteriors
tester.calibrate_posteriors(target_coverage=0.68)

# Analyze convergence
tester.analyze_convergence()

# Test recovery
tester.test_recovery()

# Create visualizations
tester.plot_results()

# Generate final report
tester.generate_report()
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any

from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)
from limference.convergence import ConvergenceAnalyzer

class SBIConvergenceTester:
    """
    General class for testing any SBI method with convergence analysis.
    Works with any set of parameters specified by the user.
    """
    
    def __init__(
        self,
        param_names: List[str],  # Specify which parameters to use
        param_labels: Optional[Dict[str, str]] = None,  # LaTeX labels for plotting
        inference_methods: List[str] = ["NPE"],
        analysis_methods: List[str] = ["pdf"],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        """
        Initialize tester with flexible parameter specification.
        
        Args:
            param_names: List of parameter names to constrain (e.g., ['omega_m', 'h'])
            param_labels: Optional dict mapping param names to LaTeX labels for plots
            inference_methods: SBI methods to use
            analysis_methods: Data analysis methods (pdf, power_spectrum, etc.)
            n_train: Number of training samples
            n_test: Number of test samples
            n_calibration: Number of calibration samples
            config_kwargs: Additional configuration options
        """
        self.param_names = param_names
        self.n_params = len(param_names)
        
        # Set up parameter labels for plotting
        if param_labels is None:
            self.param_labels = self._create_default_labels(param_names)
        else:
            self.param_labels = param_labels
            # Fill in any missing labels with defaults
            for name in param_names:
                if name not in self.param_labels:
                    self.param_labels[name] = self._create_default_labels([name])[name]
        
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
    
    def _create_default_labels(self, names: List[str]) -> Dict[str, str]:
        """Create default LaTeX labels for common parameter names."""
        default_labels = {
            'omega_m': r'$\Omega_m$',
            'Omega_m': r'$\Omega_m$',
            'h': r'$h$',
            'H0': r'$H_0$',
            'sigma8': r'$\sigma_8$',
            'sigma_8': r'$\sigma_8$',
            'ns': r'$n_s$',
            'n_s': r'$n_s$',
            'omega_b': r'$\Omega_b$',
            'Omega_b': r'$\Omega_b$',
            'omega_c': r'$\Omega_c$',
            'Omega_c': r'$\Omega_c$',
            'A_s': r'$A_s$',
            'tau': r'$\tau$',
            'w0': r'$w_0$',
            'wa': r'$w_a$',
            'mnu': r'$\sum m_\nu$',
            'omh2': r'$\Omega_m h^2$',
            'ombh2': r'$\Omega_b h^2$',
            'omch2': r'$\Omega_c h^2$'
        }
        
        labels = {}
        for name in names:
            if name in default_labels:
                labels[name] = default_labels[name]
            else:
                # Create a simple label for unknown parameters
                labels[name] = name.replace('_', ' ')
        
        return labels
    
    def load_data(self, sim_root: str, seed: int = 1, param_file_keys: Optional[Dict[str, str]] = None):
        """
        Load simulation data with flexible parameter extraction.
        
        Args:
            sim_root: Root directory for simulation data
            seed: Random seed for data files
            param_file_keys: Optional mapping from param_names to keys in the data file
                            e.g., {'omega_m': 'Omega_m', 'h': 'hubble'}
        """
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Extract specified parameters
        self.params_array = self._extract_parameters(self.params_sims, param_file_keys)
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        # Print summary
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters being constrained: {', '.join(self.param_names)}")
        print(f"Parameter ranges:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def _extract_parameters(self, params_sims: Any, param_file_keys: Optional[Dict[str, str]] = None) -> np.ndarray:
        """
        Extract specified parameters from simulation data.
        
        Args:
            params_sims: Parameter data from file
            param_file_keys: Mapping from desired param names to file keys
            
        Returns:
            Array of shape (n_sims, n_params) with extracted parameters
        """
        if param_file_keys is None:
            param_file_keys = {}
        
        params_list = []
        
        for sim_params in params_sims:
            param_values = []
            
            for param_name in self.param_names:
                # Check if there's a mapping for this parameter
                file_key = param_file_keys.get(param_name, param_name)
                
                # Try to extract the parameter with various possible keys
                value = None
                
                # Try direct access
                if file_key in sim_params:
                    value = sim_params[file_key]
                # Try common variations
                elif param_name in sim_params:
                    value = sim_params[param_name]
                # Try with underscores vs no underscores
                elif param_name.replace('_', '') in sim_params:
                    value = sim_params[param_name.replace('_', '')]
                elif param_name.replace('_', '8') in sim_params:  # for sigma_8 -> sigma8
                    value = sim_params[param_name.replace('_', '8')]
                # Try capitalized versions
                elif param_name.capitalize() in sim_params:
                    value = sim_params[param_name.capitalize()]
                elif 'Omega_' + param_name.split('_')[-1] in sim_params:  # omega_m -> Omega_m
                    value = sim_params['Omega_' + param_name.split('_')[-1]]
                else:
                    raise KeyError(f"Could not find parameter '{param_name}' (or '{file_key}') in simulation data. "
                                 f"Available keys: {list(sim_params.keys())}")
                
                param_values.append(value)
            
            params_list.append(param_values)
        
        return np.array(params_list)
    
    def setup_limference(self, output_root: str, training_kwargs: Optional[Dict] = None):
        """Initialize LIMference with configuration for specified parameters."""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Default training configuration
        default_training = {
            "max_num_epochs": 200,
            "stop_after_epochs": 20,
            "training_batch_size": 50,
            "learning_rate": 5e-4,
            "validation_fraction": 0.2,
            "noise_reduction_factor": 50.0
        }
        
        if training_kwargs:
            default_training.update(training_kwargs)
        
        # Create configuration
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "device": "cuda" if torch.cuda.is_available() else "cpu",
            **default_training
        }
        default_config.update(self.config_kwargs)
        
        self.config = LIMConfig(**default_config)
        
        # Create parameter configuration
        param_ranges = {}
        for i, name in enumerate(self.param_names):
            param_ranges[name] = (
                float(self.params_array[:, i].min()),
                float(self.params_array[:, i].max())
            )
        
        self.param_config = ParameterConfig(
            param_names=self.param_names,
            param_ranges=param_ranges,
            param_labels=self.param_labels
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        print(f"Parameters to constrain: {', '.join(self.param_names)}")
    
    def train_models(self, test_obs_idx: int = 0):
        """Train all specified SBI models."""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        for i, name in enumerate(self.param_names):
            print(f"  {name} = {self.test_true_params[i]:.3f}")
        
        # Create true params dict
        true_params_dict = {
            name: float(self.test_true_params[i]) 
            for i, name in enumerate(self.param_names)
        }
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params=true_params_dict,
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            verbose=True
        )
        
        self.models_trained = True
        self.lim.results = self.lim_results  # Store for later access
    
    def calibrate_posteriors(self, target_coverage: float = 0.68):
        """Calibrate all posteriors using temperature scaling."""
        if not self.models_trained:
            raise ValueError("Train models first using train_models()")
        
        print("\n" + "="*50)
        print("CALIBRATING POSTERIORS")
        print("="*50)
        
        def calibrate_temperature(posterior, cal_data, cal_params, target=0.68):
            """Find optimal temperature scaling"""
            def coverage_error(temp):
                coverage = 0
                n_cal = min(30, len(cal_data))
                
                for i in range(n_cal):
                    x_obs = torch.tensor(cal_data[i], dtype=torch.float32)
                    samples = posterior.sample((300,), x=x_obs).numpy()
                    
                    mean = np.mean(samples, axis=0)
                    scaled = mean + (samples - mean) * temp
                    
                    q16, q84 = np.percentile(scaled, [16, 84], axis=0)
                    if np.all((cal_params[i] >= q16) & (cal_params[i] <= q84)):
                        coverage += 1
                
                return abs(coverage/n_cal - target)
            
            result = minimize_scalar(coverage_error, bounds=(0.5, 3.0), method='bounded')
            return result.x
        
        # Calibrate each method
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Determine which test data to use
                if "power" in method_key:
                    cal_data = self.test_pk[:self.n_calibration]
                elif "pdf" in method_key:
                    cal_data = self.test_pdf[:self.n_calibration]
                else:
                    continue
                
                cal_params = self.test_params[:self.n_calibration]
                
                # Find optimal temperature
                temp = calibrate_temperature(
                    result['posterior'], cal_data, cal_params, target_coverage
                )
                
                print(f"{method_key}: temperature = {temp:.3f}")
                result['calibration_temp'] = temp
                
                # Interpretation
                if temp > 1.2:
                    print(f"  → Model was overconfident, widening by {(temp-1)*100:.1f}%")
                elif temp < 0.8:
                    print(f"  → Model was underconfident, narrowing by {(1-temp)*100:.1f}%")
                else:
                    print(f"  → Model was well-calibrated")
    
    def analyze_convergence(self):
        """Analyze training convergence for all methods"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        # Determine plot layout
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        
        # Handle different subplot configurations
        if n_rows == 1 and n_cols == 1:
            axes = np.array([[axes]])
        elif n_rows == 1:
            axes = axes.reshape(1, -1)
        elif n_cols == 1:
            axes = axes.reshape(-1, 1)
        
        convergence_results = {}
        plot_idx = 0
        
        for method_key, result in self.lim.results.items():
            if "error" not in result and 'loss_history' in result:
                print(f"\nAnalyzing {method_key}...")
                
                # Extract metrics
                metrics = self._extract_metrics(result['loss_history'])
                convergence_results[method_key] = metrics
                
                # Plot
                if plot_idx < n_rows * n_cols:
                    row = plot_idx // n_cols
                    col = plot_idx % n_cols
                    ax = axes[row, col]
                    
                    self._plot_convergence(ax, metrics, method_key)
                    plot_idx += 1
                
                # Print metrics
                self._print_convergence_metrics(metrics)
        
        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]
            ax.set_visible(False)
        
        plt.suptitle(f'Convergence Analysis: {", ".join(self.inference_methods)}', fontsize=14)
        plt.tight_layout()
        plt.show()
        
        self.convergence_results = convergence_results
        
        
    def test_recovery(self):
        """Test posterior recovery on test set"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        print("\n" + "="*50)
        print("TESTING POSTERIOR RECOVERY")
        print("="*50)
        
        recovery_results = {}
        
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                print(f"\n{method_key}:")
                
                # Get appropriate test data
                if "power" in method_key:
                    test_data = self.test_pk[self.n_calibration:]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[self.n_calibration:]
                else:
                    continue
                
                test_params_subset = self.test_params[self.n_calibration:]
                
                # Test recovery
                metrics = self._test_single_method_recovery(
                    result, test_data, test_params_subset
                )
                recovery_results[method_key] = metrics
                
                # Print results
                self._print_recovery_metrics(metrics)
        
        self.recovery_results = recovery_results
    
    def _extract_metrics(self, loss_history):
        """Extract metrics from loss history"""
        metrics = {}
        
        # Handle different possible formats
        train_losses = loss_history.get('training_loss', 
                                       loss_history.get('train_log_probs', []))
        val_losses = loss_history.get('validation_loss',
                                     loss_history.get('validation_log_probs', []))
        
        if isinstance(train_losses, np.ndarray):
            train_losses = train_losses.tolist()
        if isinstance(val_losses, np.ndarray):
            val_losses = val_losses.tolist()
            
        metrics['train_losses'] = train_losses
        metrics['val_losses'] = val_losses
        
        # Get epochs
        epochs = loss_history.get('epochs_trained', len(train_losses))
        if isinstance(epochs, (list, np.ndarray)):
            epochs = int(epochs[0] if len(epochs) > 0 else len(train_losses))
        metrics['n_epochs'] = epochs
        
        # Check convergence
        if len(train_losses) > 10:
            window = len(train_losses) // 10
            recent = train_losses[-window:]
            metrics['converged'] = np.std(recent) < 0.01 * abs(np.mean(recent))
        else:
            metrics['converged'] = False
        
        # Check overfitting
        if train_losses and val_losses:
            gap = val_losses[-1] - train_losses[-1]
            metrics['overfitting'] = gap > 0.5
        else:
            metrics['overfitting'] = False
            
        return metrics
    
    def _plot_convergence(self, ax, metrics, method_key):
        """Plot convergence curves"""
        if metrics['train_losses']:
            ax.plot(range(1, len(metrics['train_losses'])+1), 
                   metrics['train_losses'], 'b-', label='Training', linewidth=2)
        
        if metrics['val_losses']:
            ax.plot(range(1, len(metrics['val_losses'])+1), 
                   metrics['val_losses'], 'r-', label='Validation', linewidth=2)
        
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(method_key.replace('_', ' '))
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    def _print_convergence_metrics(self, metrics):
        """Print convergence metrics"""
        print(f"  Epochs: {metrics.get('n_epochs', 'N/A')}")
        print(f"  Converged: {metrics.get('converged', 'Unknown')}")
        if metrics.get('overfitting'):
            print(f"  ⚠️ Potential overfitting")
    
    def generate_report(self):
        """Generate comprehensive summary report"""
        print("\n" + "="*70)
        print(f"SBI CONVERGENCE TEST SUMMARY")
        print(f"Methods tested: {', '.join(self.inference_methods)}")
        print(f"Analysis types: {', '.join(self.analysis_methods)}")
        print(f"Parameters constrained: {', '.join(self.param_names)}")
        print("="*70)
        
        for method_key in self.lim.results:
            if "error" not in self.lim.results[method_key]:
                print(f"\n{method_key}:")
                print("-"*40)
                
                # Convergence info
                if hasattr(self, 'convergence_results') and method_key in self.convergence_results:
                    conv = self.convergence_results[method_key]
                    print(f"Training:")
                    print(f"  Epochs: {conv.get('n_epochs', 'N/A')}")
                    print(f"  Converged: {conv.get('converged', 'Unknown')}")
                    if conv.get('overfitting'):
                        print(f"  ⚠️ Overfitting detected")
                
                # Calibration info
                result = self.lim.results[method_key]
                if 'calibration_temp' in result:
                    print(f"Calibration:")
                    print(f"  Temperature: {result['calibration_temp']:.3f}")


# ==============================================================================
# EXAMPLE USAGE
# ==============================================================================

if __name__ == "__main__":
    # Configuration
    SEED = 1
    SIM_ROOT = f"/Users/anirbanroy/Desktop/product_with_noise_{SEED}/"
    OUTPUT_ROOT = "/Users/anirbanroy/Desktop/output_test"
    
    # Example 1: Two parameters (sigma8 and omega_m)
    print("="*80)
    print("EXAMPLE 1: Testing with sigma8 and omega_m")
    print("="*80)
    
    tester = SBIConvergenceTester(
        param_names=["sigma8", "omega_m"],  # Specify which parameters to constrain
        inference_methods=["NPE"],
        analysis_methods=["pdf"],
        n_train=2000,
        n_test=500,
        n_calibration=50
    )
    
    # Load data
    tester.load_data(SIM_ROOT, SEED)
    
    # Setup LIMference
    tester.setup_limference(OUTPUT_ROOT)
    
    # Train models
    tester.train_models(test_obs_idx=0)
    
    # Calibrate posteriors
    tester.calibrate_posteriors(target_coverage=0.68)
    
    # Analyze convergence
    tester.analyze_convergence()
    
    # Test recovery
    tester.test_recovery()

    # Create visualizations
    tester.plot_results()
    
    # Generate report
    tester.generate_report()
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
from scipy import stats
from scipy.optimize import minimize_scalar
from typing import Dict, List, Optional, Tuple, Any

from limference import (
    LIMference, 
    LIMConfig,
    ParameterConfig,
    calculate_statistics,
    create_corner_plot
)
from limference.convergence import ConvergenceAnalyzer

class SBIConvergenceTester:
    """
    General class for testing any SBI method with convergence analysis.
    Works with any set of parameters specified by the user.
    """
    
    def __init__(
        self,
        param_names: List[str],  # Specify which parameters to use
        param_labels: Optional[Dict[str, str]] = None,  # LaTeX labels for plotting
        inference_methods: List[str] = ["NPE"],
        analysis_methods: List[str] = ["pdf"],
        n_train: int = 2000,
        n_test: int = 500,
        n_calibration: int = 50,
        config_kwargs: Optional[Dict] = None
    ):
        """
        Initialize tester with flexible parameter specification.
        
        Args:
            param_names: List of parameter names to constrain (e.g., ['omega_m', 'h'])
            param_labels: Optional dict mapping param names to LaTeX labels for plots
            inference_methods: SBI methods to use
            analysis_methods: Data analysis methods (pdf, power_spectrum, etc.)
            n_train: Number of training samples
            n_test: Number of test samples
            n_calibration: Number of calibration samples
            config_kwargs: Additional configuration options
        """
        self.param_names = param_names
        self.n_params = len(param_names)
        
        # Set up parameter labels for plotting
        if param_labels is None:
            self.param_labels = self._create_default_labels(param_names)
        else:
            self.param_labels = param_labels
            # Fill in any missing labels with defaults
            for name in param_names:
                if name not in self.param_labels:
                    self.param_labels[name] = self._create_default_labels([name])[name]
        
        self.inference_methods = inference_methods
        self.analysis_methods = analysis_methods
        self.n_train = n_train
        self.n_test = n_test
        self.n_calibration = n_calibration
        self.config_kwargs = config_kwargs or {}
        
        self.data_loaded = False
        self.models_trained = False
        self.results = {}
    
    def _create_default_labels(self, names: List[str]) -> Dict[str, str]:
        """Create default LaTeX labels for common parameter names."""
        default_labels = {
            'omega_m': r'$\Omega_m$',
            'Omega_m': r'$\Omega_m$',
            'h': r'$h$',
            'H0': r'$H_0$',
            'sigma8': r'$\sigma_8$',
            'sigma_8': r'$\sigma_8$',
            'ns': r'$n_s$',
            'n_s': r'$n_s$',
            'omega_b': r'$\Omega_b$',
            'Omega_b': r'$\Omega_b$',
            'omega_c': r'$\Omega_c$',
            'Omega_c': r'$\Omega_c$',
            'A_s': r'$A_s$',
            'tau': r'$\tau$',
            'w0': r'$w_0$',
            'wa': r'$w_a$',
            'mnu': r'$\sum m_\nu$',
            'omh2': r'$\Omega_m h^2$',
            'ombh2': r'$\Omega_b h^2$',
            'omch2': r'$\Omega_c h^2$'
        }
        
        labels = {}
        for name in names:
            if name in default_labels:
                labels[name] = default_labels[name]
            else:
                # Create a simple label for unknown parameters
                labels[name] = name.replace('_', ' ')
        
        return labels
    
    def load_data(self, sim_root: str, seed: int = 1, param_file_keys: Optional[Dict[str, str]] = None):
        """
        Load simulation data with flexible parameter extraction.
        
        Args:
            sim_root: Root directory for simulation data
            seed: Random seed for data files
            param_file_keys: Optional mapping from param_names to keys in the data file
                            e.g., {'omega_m': 'Omega_m', 'h': 'hubble'}
        """
        print(f"Loading data from {sim_root}...")
        
        # Load files
        ps_file = np.load(sim_root + f"ps_80_256_{seed}.npz", allow_pickle=True)
        pdf_file = np.load(sim_root + f"pdf_80_256_{seed}.npz", allow_pickle=True)
        
        # Extract data
        self.params_sims = ps_file["parameters"]
        self.k_sim = ps_file["k_2d"]
        self.pk_sim = ps_file["pk_2d"]
        self.bin_sim_linear = pdf_file["lim_bin_edges_linear_2d"]
        self.pdf_sim_linear = pdf_file["lim_pdf_linear_2d"]
        
        # Extract specified parameters
        self.params_array = self._extract_parameters(self.params_sims, param_file_keys)
        
        # Split data
        self.train_params = self.params_array[:self.n_train]
        self.train_pk = self.pk_sim[:self.n_train]
        self.train_pdf = self.pdf_sim_linear[:self.n_train]
        
        self.test_params = self.params_array[self.n_train:self.n_train+self.n_test]
        self.test_pk = self.pk_sim[self.n_train:self.n_train+self.n_test]
        self.test_pdf = self.pdf_sim_linear[self.n_train:self.n_train+self.n_test]
        
        self.data_loaded = True
        
        # Print summary
        print(f"Data loaded: {len(self.params_sims)} total simulations")
        print(f"Training set: {self.n_train} samples")
        print(f"Test set: {self.n_test} samples")
        print(f"Parameters being constrained: {', '.join(self.param_names)}")
        print(f"Parameter ranges:")
        for i, name in enumerate(self.param_names):
            print(f"  {name}: [{self.params_array[:, i].min():.3f}, {self.params_array[:, i].max():.3f}]")
    
    def _extract_parameters(self, params_sims: Any, param_file_keys: Optional[Dict[str, str]] = None) -> np.ndarray:
        """
        Extract specified parameters from simulation data.
        
        Args:
            params_sims: Parameter data from file
            param_file_keys: Mapping from desired param names to file keys
            
        Returns:
            Array of shape (n_sims, n_params) with extracted parameters
        """
        if param_file_keys is None:
            param_file_keys = {}
        
        params_list = []
        
        for sim_params in params_sims:
            param_values = []
            
            for param_name in self.param_names:
                # Check if there's a mapping for this parameter
                file_key = param_file_keys.get(param_name, param_name)
                
                # Try to extract the parameter with various possible keys
                value = None
                
                # Try direct access
                if file_key in sim_params:
                    value = sim_params[file_key]
                # Try common variations
                elif param_name in sim_params:
                    value = sim_params[param_name]
                # Try with underscores vs no underscores
                elif param_name.replace('_', '') in sim_params:
                    value = sim_params[param_name.replace('_', '')]
                elif param_name.replace('_', '8') in sim_params:  # for sigma_8 -> sigma8
                    value = sim_params[param_name.replace('_', '8')]
                # Try capitalized versions
                elif param_name.capitalize() in sim_params:
                    value = sim_params[param_name.capitalize()]
                elif 'Omega_' + param_name.split('_')[-1] in sim_params:  # omega_m -> Omega_m
                    value = sim_params['Omega_' + param_name.split('_')[-1]]
                else:
                    raise KeyError(f"Could not find parameter '{param_name}' (or '{file_key}') in simulation data. "
                                 f"Available keys: {list(sim_params.keys())}")
                
                param_values.append(value)
            
            params_list.append(param_values)
        
        return np.array(params_list)
    
    def setup_limference(self, output_root: str, training_kwargs: Optional[Dict] = None):
        """Initialize LIMference with configuration for specified parameters."""
        if not self.data_loaded:
            raise ValueError("Load data first using load_data()")
        
        # Default training configuration
        default_training = {
            "max_num_epochs": 100,
            "stop_after_epochs": 10,
            "training_batch_size": 100,
            "learning_rate": 1e-4,
            "validation_fraction": 0.3,
            "noise_reduction_factor": 50.0,
             "weight_decay": 1e-4
        }
        
        if training_kwargs:
            default_training.update(training_kwargs)
        
        # Create configuration
        default_config = {
            "root": output_root,
            "n_signal_sims": self.n_train,
            "num_noise_simulations": 1,
            "inference_methods": self.inference_methods,
            "analysis_methods": self.analysis_methods,
            "device": "cuda" if torch.cuda.is_available() else "cpu",
            **default_training
        }
        default_config.update(self.config_kwargs)
        
        self.config = LIMConfig(**default_config)
        
        # Create parameter configuration
        param_ranges = {}
        for i, name in enumerate(self.param_names):
            param_ranges[name] = (
                float(self.params_array[:, i].min()),
                float(self.params_array[:, i].max())
            )
        
        self.param_config = ParameterConfig(
            param_names=self.param_names,
            param_ranges=param_ranges,
            param_labels=self.param_labels
        )
        
        # Initialize LIMference
        self.lim = LIMference(self.config, param_config=self.param_config)
        
        # Load training data
        self.lim.parameters = self.train_params
        self.lim.simulations = {
            "power_spectrum": {"k": self.k_sim, "pk": self.train_pk},
            "pdf": {"bins_linear": self.bin_sim_linear, "pdf_linear": self.train_pdf}
        }
        
        print(f"LIMference initialized for methods: {self.inference_methods}")
        print(f"Parameters to constrain: {', '.join(self.param_names)}")
    
    def train_models(self, test_obs_idx: int = 0):
        """Train all specified SBI models."""
        if not hasattr(self, 'lim'):
            raise ValueError("Setup LIMference first using setup_limference()")
        
        # Get test observation
        self.test_true_params = self.test_params[test_obs_idx]
        
        print(f"\nTest observation true parameters:")
        for i, name in enumerate(self.param_names):
            print(f"  {name} = {self.test_true_params[i]:.3f}")
        
        # Create true params dict
        true_params_dict = {
            name: float(self.test_true_params[i]) 
            for i, name in enumerate(self.param_names)
        }
        
        # Run inference for all methods
        self.lim_results = self.lim.run_inference_comparison(
            observation_index=None,
            true_params=true_params_dict,
            analysis_methods=self.analysis_methods,
            inference_methods=self.inference_methods,
            n_posterior_samples=10000,
            verbose=True
        )
        
        self.models_trained = True
        self.lim.results = self.lim_results  # Store for later access
    
    def calibrate_posteriors(self, target_coverage: float = 0.68):
        """Calibrate all posteriors using temperature scaling."""
        if not self.models_trained:
            raise ValueError("Train models first using train_models()")
        
        print("\n" + "="*50)
        print("CALIBRATING POSTERIORS")
        print("="*50)
        
        def calibrate_temperature(posterior, cal_data, cal_params, target=0.68):
            """Find optimal temperature scaling"""
            def coverage_error(temp):
                coverage = 0
                n_cal = min(30, len(cal_data))
                
                for i in range(n_cal):
                    x_obs = torch.tensor(cal_data[i], dtype=torch.float32)
                    samples = posterior.sample((300,), x=x_obs).numpy()
                    
                    mean = np.mean(samples, axis=0)
                    scaled = mean + (samples - mean) * temp
                    
                    q16, q84 = np.percentile(scaled, [16, 84], axis=0)
                    if np.all((cal_params[i] >= q16) & (cal_params[i] <= q84)):
                        coverage += 1
                
                return abs(coverage/n_cal - target)
            
            result = minimize_scalar(coverage_error, bounds=(0.5, 3.0), method='bounded')
            return result.x
        
        # Calibrate each method
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Determine which test data to use
                if "power" in method_key:
                    cal_data = self.test_pk[:self.n_calibration]
                elif "pdf" in method_key:
                    cal_data = self.test_pdf[:self.n_calibration]
                else:
                    continue
                
                cal_params = self.test_params[:self.n_calibration]
                
                # Find optimal temperature
                temp = calibrate_temperature(
                    result['posterior'], cal_data, cal_params, target_coverage
                )
                
                print(f"{method_key}: temperature = {temp:.3f}")
                result['calibration_temp'] = temp
                
                # Interpretation
                if temp > 1.2:
                    print(f"  → Model was overconfident, widening by {(temp-1)*100:.1f}%")
                elif temp < 0.8:
                    print(f"  → Model was underconfident, narrowing by {(1-temp)*100:.1f}%")
                else:
                    print(f"  → Model was well-calibrated")
    
    def analyze_convergence(self):
        """Analyze training convergence for all methods"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        # Determine plot layout
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        
        # Handle different subplot configurations
        if n_rows == 1 and n_cols == 1:
            axes = np.array([[axes]])
        elif n_rows == 1:
            axes = axes.reshape(1, -1)
        elif n_cols == 1:
            axes = axes.reshape(-1, 1)
        
        convergence_results = {}
        plot_idx = 0
        
        for method_key, result in self.lim.results.items():
            if "error" not in result and 'loss_history' in result:
                print(f"\nAnalyzing {method_key}...")
                
                # Extract metrics
                metrics = self._extract_metrics(result['loss_history'])
                convergence_results[method_key] = metrics
                
                # Plot
                if plot_idx < n_rows * n_cols:
                    row = plot_idx // n_cols
                    col = plot_idx % n_cols
                    ax = axes[row, col]
                    
                    self._plot_convergence(ax, metrics, method_key)
                    plot_idx += 1
                
                # Print metrics
                self._print_convergence_metrics(metrics)
        
        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]
            ax.set_visible(False)
        
        plt.suptitle(f'Convergence Analysis: {", ".join(self.inference_methods)}', fontsize=14)
        plt.tight_layout()
        plt.show()
        
        self.convergence_results = convergence_results
    
    def test_recovery(self):
        """Test posterior recovery on test set"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        print("\n" + "="*50)
        print("TESTING POSTERIOR RECOVERY")
        print("="*50)
        
        recovery_results = {}
        
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                print(f"\n{method_key}:")
                
                # Get appropriate test data
                if "power" in method_key:
                    test_data = self.test_pk[self.n_calibration:]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[self.n_calibration:]
                else:
                    continue
                
                test_params_subset = self.test_params[self.n_calibration:]
                
                # Test recovery
                metrics = self._test_single_method_recovery(
                    result, test_data, test_params_subset
                )
                recovery_results[method_key] = metrics
                
                # Print results
                self._print_recovery_metrics(metrics)
        
        self.recovery_results = recovery_results
    
    def plot_results(self):
        """Create comprehensive visualization of results"""
        if not self.models_trained:
            raise ValueError("Train models first")
        
        # 1. Coverage calibration curves
        self._plot_coverage_curves()
        
        # 2. Parameter recovery scatter plots
        self._plot_parameter_recovery()
        
        # 3. Uncertainty analysis
        self._plot_uncertainty_analysis()
    
    def generate_report(self):
        """Generate comprehensive summary report"""
        print("\n" + "="*70)
        print(f"SBI CONVERGENCE TEST SUMMARY")
        print(f"Methods tested: {', '.join(self.inference_methods)}")
        print(f"Analysis types: {', '.join(self.analysis_methods)}")
        print(f"Parameters constrained: {', '.join(self.param_names)}")
        print("="*70)
        
        for method_key in self.lim.results:
            if "error" not in self.lim.results[method_key]:
                print(f"\n{method_key}:")
                print("-"*40)
                
                # Convergence info
                if hasattr(self, 'convergence_results') and method_key in self.convergence_results:
                    conv = self.convergence_results[method_key]
                    print(f"Training:")
                    print(f"  Epochs: {conv.get('n_epochs', 'N/A')}")
                    print(f"  Converged: {conv.get('converged', 'Unknown')}")
                    if conv.get('overfitting'):
                        print(f"  ⚠️ Overfitting detected")
                
                # Calibration info
                result = self.lim.results[method_key]
                if 'calibration_temp' in result:
                    print(f"Calibration:")
                    print(f"  Temperature: {result['calibration_temp']:.3f}")
                
                # Recovery performance
                if hasattr(self, 'recovery_results') and method_key in self.recovery_results:
                    rec = self.recovery_results[method_key]
                    print(f"Performance:")
                    print(f"  RMSE: {rec['mean_rmse']:.4f}")
                    print(f"  68% coverage: {rec['coverage_68_rate']:.1%}")
                    print(f"  95% coverage: {rec['coverage_95_rate']:.1%}")
                    
                    if rec['well_calibrated_68'] and rec['well_calibrated_95']:
                        print("  ✅ Well calibrated")
                    else:
                        print("  ⚠️ Needs adjustment")
    
    # Helper methods
    def _extract_metrics(self, loss_history):
        """Extract metrics from loss history"""
        metrics = {}
        
        # Handle different possible formats
        train_losses = loss_history.get('training_loss', 
                                       loss_history.get('train_log_probs', []))
        val_losses = loss_history.get('validation_loss',
                                     loss_history.get('validation_log_probs', []))
        
        if isinstance(train_losses, np.ndarray):
            train_losses = train_losses.tolist()
        if isinstance(val_losses, np.ndarray):
            val_losses = val_losses.tolist()
            
        metrics['train_losses'] = train_losses
        metrics['val_losses'] = val_losses
        
        # Get epochs
        epochs = loss_history.get('epochs_trained', len(train_losses))
        if isinstance(epochs, (list, np.ndarray)):
            epochs = int(epochs[0] if len(epochs) > 0 else len(train_losses))
        metrics['n_epochs'] = epochs
        
        # Check convergence
        if len(train_losses) > 10:
            window = len(train_losses) // 10
            recent = train_losses[-window:]
            metrics['converged'] = np.std(recent) < 0.01 * abs(np.mean(recent))
        else:
            metrics['converged'] = False
        
        # Check overfitting
        if train_losses and val_losses:
            gap = val_losses[-1] - train_losses[-1]
            metrics['overfitting'] = gap > 0.5
        else:
            metrics['overfitting'] = False
            
        return metrics
    
    def _plot_convergence(self, ax, metrics, method_key):
        """Plot convergence curves"""
        if metrics['train_losses']:
            ax.plot(range(1, len(metrics['train_losses'])+1), 
                   metrics['train_losses'], 'b-', label='Training', linewidth=2)
        
        if metrics['val_losses']:
            ax.plot(range(1, len(metrics['val_losses'])+1), 
                   metrics['val_losses'], 'r-', label='Validation', linewidth=2)
        
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')
        ax.set_title(method_key.replace('_', ' '))
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    def _print_convergence_metrics(self, metrics):
        """Print convergence metrics"""
        print(f"  Epochs: {metrics.get('n_epochs', 'N/A')}")
        print(f"  Converged: {metrics.get('converged', 'Unknown')}")
        if metrics.get('overfitting'):
            print(f"  ⚠️ Potential overfitting")
    
    def _test_single_method_recovery(self, result, test_data, test_params):
        """Test recovery for a single method"""
        n_test_samples = min(50, len(test_data))
        coverage_68 = 0
        coverage_95 = 0
        errors = []
        
        temp = result.get('calibration_temp', 1.0)
        
        for i in range(n_test_samples):
            x_obs = torch.tensor(test_data[i], dtype=torch.float32)
            samples = result['posterior'].sample((1000,), x=x_obs).numpy()
            
            # Apply temperature scaling
            mean = np.mean(samples, axis=0)
            samples = mean + (samples - mean) * temp
            
            # Calculate statistics
            post_mean = np.mean(samples, axis=0)
            errors.append(post_mean - test_params[i])
            
            # Check coverage
            q16, q84 = np.percentile(samples, [16, 84], axis=0)
            q2_5, q97_5 = np.percentile(samples, [2.5, 97.5], axis=0)
            
            if np.all((test_params[i] >= q16) & (test_params[i] <= q84)):
                coverage_68 += 1
            if np.all((test_params[i] >= q2_5) & (test_params[i] <= q97_5)):
                coverage_95 += 1
        
        errors = np.array(errors)
        
        return {
            'coverage_68_rate': coverage_68 / n_test_samples,
            'coverage_95_rate': coverage_95 / n_test_samples,
            'mean_rmse': np.sqrt(np.mean(errors**2)),
            'mean_bias': np.mean(np.abs(errors)),
            'well_calibrated_68': abs(coverage_68/n_test_samples - 0.68) < 0.1,
            'well_calibrated_95': abs(coverage_95/n_test_samples - 0.95) < 0.1,
            'calibration_temp': temp
        }
    
    def _print_recovery_metrics(self, metrics):
        """Print recovery metrics"""
        print(f"  Temperature: {metrics['calibration_temp']:.3f}")
        print(f"  68% coverage: {metrics['coverage_68_rate']:.1%}")
        print(f"  95% coverage: {metrics['coverage_95_rate']:.1%}")
        print(f"  RMSE: {metrics['mean_rmse']:.4f}")
    
    def _plot_coverage_curves(self):
        """Plot coverage calibration curves"""
        confidence_levels = np.array([50, 60, 70, 80, 90, 95, 99])
        
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_cols = min(3, n_methods)
        n_rows = (n_methods + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
        
        # Handle different subplot configurations
        if n_rows == 1 and n_cols == 1:
            axes = np.array([[axes]])
        elif n_rows == 1:
            axes = axes.reshape(1, -1)
        elif n_cols == 1:
            axes = axes.reshape(-1, 1)
        
        plot_idx = 0
        
        for method_key, result in self.lim.results.items():
            if "error" not in result and plot_idx < n_rows * n_cols:
                row = plot_idx // n_cols
                col = plot_idx % n_cols
                ax = axes[row, col]
                
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                
                test_params_subset = self.test_params[50:150]
                
                # Calculate coverage for different confidence levels
                coverage_rates_orig = []
                coverage_rates_cal = []
                
                temp = result.get('calibration_temp', 1.0)
                
                for conf in confidence_levels:
                    coverage_orig = 0
                    coverage_cal = 0
                    n_samples = min(30, len(test_data))
                    
                    for i in range(n_samples):
                        x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                        samples = result['posterior'].sample((500,), x=x_obs).numpy()
                        
                        lower = (100-conf)/2
                        upper = 100-lower
                        
                        # Original
                        ci_low_orig = np.percentile(samples, lower, axis=0)
                        ci_high_orig = np.percentile(samples, upper, axis=0)
                        
                        # Calibrated
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                        ci_low_cal = np.percentile(samples_cal, lower, axis=0)
                        ci_high_cal = np.percentile(samples_cal, upper, axis=0)
                        
                        if np.all((test_params_subset[i] >= ci_low_orig) & 
                                 (test_params_subset[i] <= ci_high_orig)):
                            coverage_orig += 1
                        if np.all((test_params_subset[i] >= ci_low_cal) & 
                                 (test_params_subset[i] <= ci_high_cal)):
                            coverage_cal += 1
                    
                    coverage_rates_orig.append(100 * coverage_orig/n_samples)
                    coverage_rates_cal.append(100 * coverage_cal/n_samples)
                
                # Plot
                ax.plot(confidence_levels, coverage_rates_orig, 'o-', 
                       label='Original', linewidth=2, color='red')
                ax.plot(confidence_levels, coverage_rates_cal, 's-', 
                       label='Calibrated', linewidth=2, color='green')
                ax.plot([0, 100], [0, 100], 'k--', alpha=0.5, label='Perfect')
                
                ax.set_xlabel('Nominal Coverage (%)')
                ax.set_ylabel('Empirical Coverage (%)')
                ax.set_title(f'{method_key} (T={temp:.2f})')
                ax.legend(fontsize=8)
                ax.grid(True, alpha=0.3)
                ax.set_xlim(45, 100)
                ax.set_ylim(45, 100)
                
                plot_idx += 1
        
        # Hide unused subplots
        for idx in range(plot_idx, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]
            ax.set_visible(False)
        
        plt.suptitle('Coverage Calibration: Before vs After', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def _plot_parameter_recovery(self):
        """Plot parameter recovery with error bars"""
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_params = len(self.param_names)
        
        fig, axes = plt.subplots(n_methods, n_params, figsize=(5*n_params, 4*n_methods))
        
        if n_methods == 1 and n_params == 1:
            axes = np.array([[axes]])
        elif n_methods == 1:
            axes = axes.reshape(1, -1)
        elif n_params == 1:
            axes = axes.reshape(-1, 1)
        
        method_idx = 0
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:150]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:150]
                else:
                    continue
                
                test_params_subset = self.test_params[50:150]
                temp = result.get('calibration_temp', 1.0)
                
                # Collect predictions
                predictions_mean = []
                predictions_16 = []
                predictions_84 = []
                true_vals = []
                n_eval = min(50, len(test_data))
                
                for i in range(n_eval):
                    x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                    samples = result['posterior'].sample((1000,), x=x_obs).numpy()
                    
                    # Apply calibration
                    mean = np.mean(samples, axis=0)
                    samples_cal = mean + (samples - mean) * temp
                    
                    predictions_mean.append(np.mean(samples_cal, axis=0))
                    predictions_16.append(np.percentile(samples_cal, 16, axis=0))
                    predictions_84.append(np.percentile(samples_cal, 84, axis=0))
                    true_vals.append(test_params_subset[i])
                
                predictions_mean = np.array(predictions_mean)
                predictions_16 = np.array(predictions_16)
                predictions_84 = np.array(predictions_84)
                true_vals = np.array(true_vals)
                
                for param_idx in range(n_params):
                    ax = axes[method_idx, param_idx] if n_methods > 1 or n_params > 1 else axes[0, 0]
                    
                    # Error bars
                    yerr_lower = predictions_mean[:, param_idx] - predictions_16[:, param_idx]
                    yerr_upper = predictions_84[:, param_idx] - predictions_mean[:, param_idx]
                    yerr = np.array([yerr_lower, yerr_upper])
                    
                    ax.errorbar(true_vals[:, param_idx], 
                               predictions_mean[:, param_idx],
                               yerr=yerr,
                               fmt='o', alpha=0.6, markersize=4,
                               elinewidth=0.5, capsize=2, capthick=0.5,
                               label='Posterior mean ± 68% CI')
                    
                    # Perfect recovery line
                    lims = [true_vals[:, param_idx].min(), true_vals[:, param_idx].max()]
                    ax.plot(lims, lims, 'r--', alpha=0.7, linewidth=2, 
                           label='Perfect recovery')
                    
                    # Calculate metrics
                    from scipy.stats import pearsonr
                    r2 = pearsonr(true_vals[:, param_idx], 
                                 predictions_mean[:, param_idx])[0]**2
                    rmse = np.sqrt(np.mean((true_vals[:, param_idx] - 
                                           predictions_mean[:, param_idx])**2))
                    
                    ax.set_xlabel(f'True {self.param_labels[self.param_names[param_idx]]}')
                    ax.set_ylabel(f'Predicted {self.param_labels[self.param_names[param_idx]]}')
                    ax.set_title(f'{method_key}\nR²={r2:.3f}, RMSE={rmse:.4f}')
                    ax.grid(True, alpha=0.3)
                    if param_idx == 0:
                        ax.legend(loc='upper left', fontsize=8)
                
                method_idx += 1
        
        plt.suptitle('Parameter Recovery with Uncertainty', fontsize=14)
        plt.tight_layout()
        plt.show()
    
    def _plot_uncertainty_analysis(self):
        """Plot uncertainty analysis"""
        n_methods = len([k for k in self.lim.results.keys() if "error" not in self.lim.results[k]])
        n_params = len(self.param_names)
        
        fig, axes = plt.subplots(n_params, n_methods, figsize=(5*n_methods, 4*n_params))
        
        if n_methods == 1 and n_params == 1:
            axes = np.array([[axes]])
        elif n_methods == 1:
            axes = axes.reshape(-1, 1)
        elif n_params == 1:
            axes = axes.reshape(1, -1)
        
        method_idx = 0
        for method_key, result in self.lim.results.items():
            if "error" not in result:
                # Get test data
                if "power" in method_key:
                    test_data = self.test_pk[50:100]
                elif "pdf" in method_key:
                    test_data = self.test_pdf[50:100]
                else:
                    continue
                
                test_params_subset = self.test_params[50:100]
                temp = result.get('calibration_temp', 1.0)
                
                for param_idx in range(n_params):
                    ax = axes[param_idx, method_idx] if n_params > 1 or n_methods > 1 else axes[0, 0]
                    
                    # Collect errors and uncertainties
                    all_errors = []
                    all_uncertainties = []
                    
                    for i in range(len(test_data)):
                        x_obs = torch.tensor(test_data[i], dtype=torch.float32)
                        samples = result['posterior'].sample((500,), x=x_obs).numpy()
                        
                        # Apply calibration
                        mean = np.mean(samples, axis=0)
                        samples_cal = mean + (samples - mean) * temp
                        
                        pred_mean = np.mean(samples_cal[:, param_idx])
                        pred_std = np.std(samples_cal[:, param_idx])
                        true_val = test_params_subset[i, param_idx]
                        
                        error = pred_mean - true_val
                        all_errors.append(error)
                        all_uncertainties.append(pred_std)
                    
                    all_errors = np.array(all_errors)
                    all_uncertainties = np.array(all_uncertainties)
                    
                    # Plot
                    scatter = ax.scatter(all_uncertainties, all_errors, 
                                       c=np.abs(all_errors), cmap='viridis',
                                       alpha=0.6, s=30)
                    
                    ax.axhline(0, color='red', linestyle='--', alpha=0.5)
                    
                    x_range = np.linspace(all_uncertainties.min(), 
                                        all_uncertainties.max(), 100)
                    ax.fill_between(x_range, -x_range, x_range, 
                                   alpha=0.2, color='gray', label='±1σ')
                    ax.fill_between(x_range, -2*x_range, 2*x_range, 
                                   alpha=0.1, color='gray', label='±2σ')
                    
                    # Calculate z-scores
                    z_scores = all_errors / (all_uncertainties + 1e-10)
                    z_rms = np.sqrt(np.mean(z_scores**2))
                    
                    ax.set_xlabel(f'Posterior Std for {self.param_labels[self.param_names[param_idx]]}')
                    ax.set_ylabel(f'Error in {self.param_labels[self.param_names[param_idx]]}')
                    ax.set_title(f'{method_key}\nZ-RMS={z_rms:.2f}')
                    ax.grid(True, alpha=0.3)
                    
                    if param_idx == 0 and method_idx == 0:
                        ax.legend(loc='upper left', fontsize=8)
                
                method_idx += 1
        
        plt.suptitle('Error vs Uncertainty Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()


# ==============================================================================
# EXAMPLE USAGE
# ==============================================================================

if __name__ == "__main__":
    # Configuration
    SEED = 1
    SIM_ROOT = f"/Users/anirbanroy/Desktop/product_with_noise_{SEED}/"
    OUTPUT_ROOT = "/Users/anirbanroy/Desktop/output_test"
    
    # Example: Testing with flexible parameters
    print("="*80)
    print("TESTING SBI CONVERGENCE WITH FLEXIBLE PARAMETERS")
    print("="*80)
    
    # You can specify any parameters available in your data
    tester = SBIConvergenceTester(
        param_names=["sigma8", "omega_m"],  # Specify which parameters to constrain
        inference_methods=["NPE"],            # Can add: SNPE, NRE, SNLE
        analysis_methods=["pdf", "power_spectrum"],  # Data types to use
        n_train=2000,
        n_test=500,
        n_calibration=50
    )
    
    # Load data
    tester.load_data(SIM_ROOT, SEED)
    
    # Setup LIMference
    tester.setup_limference(OUTPUT_ROOT)
    
    # Train models
    tester.train_models(test_obs_idx=0)
    
    # Calibrate posteriors
    tester.calibrate_posteriors(target_coverage=0.68)
    
    # Analyze convergence
    tester.analyze_convergence()
    
    # Test recovery
    tester.test_recovery()
    
    # Create visualizations
    tester.plot_results()
    
    # Generate final report
    tester.generate_report()
```
